```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings

warnings.filterwarnings('ignore') # Suppress warnings, especially from pandas/numpy operations


class BusinessGameReporter:
    """
    Generates a comprehensive business game report with financial, marketing,
    operational, competitive, product, and geographical analysis.
    Enhanced with deeper strategic insights and actionable frameworks.
    """

    def __init__(self, initial_investment=1_000_000):
        """
        Initializes the reporter with global settings and data.

        Args:
            initial_investment (int): The initial investment amount for the game.
        """
        self.initial_investment = initial_investment
        self._setup_plotting_style()

        # Define a consistent color palette for all plots
        self.COLORS = {
            'primary': '#1f77b4',   # Blue
            'secondary': '#ff7f0e', # Orange
            'success': '#2ca02c',   # Green
            'warning': '#d62728',   # Red
            'danger': '#e74c3c',    # Lighter Red (for crisis)
            'info': '#17a2b8',      # Cyan
            'purple': '#9467bd',    # Purple
            'brown': '#8c564b',     # Brown
            'pink': '#e377c2',      # Pink
            'gray': '#7f7f7f'       # Gray
        }

        # Initialize DataFrames (will be populated in _load_and_process_data)
        self.df_finance = pd.DataFrame()
        self.marketing_full = pd.DataFrame()
        self.operations_data = pd.DataFrame()
        self.market_data = pd.DataFrame()
        self.product_data_v4e = pd.DataFrame()
        self.df_geo_v4e = pd.DataFrame()

        # Dictionary to map team names to display names and colors for plotting
        self.plot_teams_map = {}

    def _setup_plotting_style(self):
        """Sets up a professional plotting style and parameters for consistency."""
        plt.style.use('seaborn-v0_8-darkgrid')  # Use a dark grid style for better contrast
        plt.rcParams['figure.figsize'] = (16, 10)  # Default figure size
        plt.rcParams['font.size'] = 12             # Base font size
        plt.rcParams['font.family'] = 'DejaVu Sans' # Font family
        plt.rcParams['axes.titlesize'] = 16        # Title font size
        plt.rcParams['axes.titleweight'] = 'bold'  # Title font weight
        plt.rcParams['axes.labelsize'] = 14        # Axis label font size
        plt.rcParams['xtick.labelsize'] = 12       # X-tick label font size
        plt.rcParams['ytick.labelsize'] = 12       # Y-tick label font size
        plt.rcParams['legend.fontsize'] = 12       # Legend font size
        plt.rcParams['figure.titlesize'] = 20      # Super title font size
        plt.rcParams['figure.titleweight'] = 'bold' # Super title font weight

    def _load_and_process_data(self):
        """
        Loads raw data from predefined dictionaries and performs initial calculations
        for all sections of the report. This method simulates data loading from external
        sources and prepares it for analysis and plotting.
        """
        # --- Financial Data ---
        # Raw financial data for 10 rounds. Note: Some T11 data is not available as it's the final round.
        financial_raw_data = {
            'Round': list(range(1, 11)),  # Only rounds 1-10 for gameplay
            'Starting_Balance': [1000000, 749000, 1096813, 1343786, 1792392, -144627, 268362, 3340282, 5185242, 4758356],
            'Ending_Balance': [749000, 1096813, 1343786, 1792392, -144627, 268362, 3340282, 5185242, 4758356, 5976703],
            'Net_Change': [-251000, 347813, 246973, 448606, -1937019, 412989, 3071920, 1844960, -426886, 1218347],
            'Marketing_Spend': [62200, 35000, 35000, 35000, 35000, 35000, 310000, 0, 1500000, 84500],
            'Production_Cost': [0, 406550, 1020500, 1719000, 1719000, 1718000, 2081000, 2356000, 2356000, 2356000],
            'HR_Cost': [27300, 163710, 134800, 218140, 191640, 292260, 247910, 343420, 313760, 223300],
            'Operations_Cost': [161500, 573760, 1158800, 1940640, 1914140, 2013760, 2638910, 2699420, 4169760, 2663800],
            'Infrastructure_Invest': [251000, 0, 10000, 0, 0, 0, 0, 0, 0, 0],
            'Units_Produced': [0, 6500, 6500, 12000, 12000, 12000, 15000, 17000, 17000, 17000],
            'Units_Sold': [0, 6500, 12000, 11849, 151, 11744, 22324, 19873, 17104, 16441]
        }

        self.df_finance = pd.DataFrame(financial_raw_data)

        # Calculate total costs by summing up various cost categories
        self.df_finance['Total_Costs'] = (
            self.df_finance['Marketing_Spend'] +
            self.df_finance['Production_Cost'] +
            self.df_finance['HR_Cost'] +
            self.df_finance['Operations_Cost']
        )
        # Implied Revenue is calculated as Net Change + Total Costs
        self.df_finance['Implied_Revenue'] = self.df_finance['Net_Change'] + self.df_finance['Total_Costs']
        # Profit Margin calculation, handling division by zero
        self.df_finance['Profit_Margin'] = np.where(
            self.df_finance['Implied_Revenue'] > 0,
            (self.df_finance['Implied_Revenue'] - self.df_finance['Total_Costs']) / self.df_finance['Implied_Revenue'] * 100,
            0
        )

        # Unit Economics Calculations
        self.df_finance['Production_Cost_Per_Unit'] = np.where(
            self.df_finance['Units_Produced'] > 0,
            self.df_finance['Production_Cost'] / self.df_finance['Units_Produced'],
            0
        )
        self.df_finance['Marketing_Cost_Per_Unit_Sold'] = np.where(
            self.df_finance['Units_Sold'] > 0,
            self.df_finance['Marketing_Spend'] / self.df_finance['Units_Sold'],
            np.nan # Use NaN for rounds with no units sold to avoid misleading zeros
        )
        self.df_finance['Total_Cost_Per_Unit_Sold'] = np.where(
            self.df_finance['Units_Sold'] > 0,
            self.df_finance['Total_Costs'] / self.df_finance['Units_Sold'],
            0
        )
        self.df_finance['Revenue_Per_Unit_Sold'] = np.where(
            self.df_finance['Units_Sold'] > 0,
            self.df_finance['Implied_Revenue'] / self.df_finance['Units_Sold'],
            0
        )
        self.df_finance['Profit_Per_Unit_Sold'] = self.df_finance['Revenue_Per_Unit_Sold'] - self.df_finance['Total_Cost_Per_Unit_Sold']

        # --- Marketing Data ---
        self.marketing_full = self.df_finance[['Round', 'Marketing_Spend', 'Units_Sold', 'Units_Produced']].copy()
        # Map round numbers to specific campaign details for context
        campaign_details_map = {
            1: 'Ustawienia poczƒÖtkowe', 2: 'Utrzymanie', 3: 'Utrzymanie', 4: 'Utrzymanie',
            5: 'Utrzymanie (wp≈Çyw kryzysu)', 6: 'Utrzymanie (odbudowa)', 7: 'Intensywna kampania',
            8: 'Test organiczny (‚Ç¨0)', 9: 'MEGA kampania', 10: 'Zoptymalizowana minimalna'
        }
        self.marketing_full['Campaign_Details'] = self.marketing_full['Round'].map(campaign_details_map)

        # Sales Success Rate (Units Sold / Units Produced)
        self.marketing_full['Sales_Success_Rate'] = np.where(
            self.marketing_full['Units_Produced'] > 0,
            (self.marketing_full['Units_Sold'] / self.marketing_full['Units_Produced']) * 100,
            0
        )
        # Market Share data for V4E Scooters (ensure 10 values)
        v4e_market_share_data = [0.00, 0.00, 24.51, 22.29, 0.92, 30.30, 52.48, 57.46, 38.08, 39.87]
        self.marketing_full['Market_Share_V4E'] = v4e_market_share_data

        # Ad channel spending (illustrative data) - ensure 10 values
        self.marketing_full['Direct_Sale_Ads'] = [0, 0, 9, 9, 9, 9, 20, 0, 110, 9]
        self.marketing_full['Ambassador_Ads'] = [0, 0, 9, 9, 9, 9, 15, 0, 110, 9]
        self.marketing_full['Newspaper_Ads'] = [0, 0, 9, 9, 9, 9, 60, 0, 150, 9]

        self.marketing_full['Cost_Per_Sale'] = np.where(
            self.marketing_full['Units_Sold'] > 0,
            self.marketing_full['Marketing_Spend'] / self.marketing_full['Units_Sold'],
            np.nan
        )
        self.marketing_full['Marketing_Intensity_Millions'] = self.marketing_full['Marketing_Spend'] / 1_000_000

        # --- Operational Data ---
        operations_raw_data = {
            'Round': list(range(1, 11)),  # Ensure exactly 10 rounds
            'Fashionistas_Eff': [0, 106.98, 115.36, 104.61, 114.68, 111.18, 140.62, 160.02, 200.00, 200.00],
            'Geeks_Eff': [0, 106.86, 105.66, 100.26, 109.79, 107.95, 122.84, 139.91, 175.67, 200.00],
            'Production_Capacity': [5000, 15600, 15600, 17000, 17000, 17000, 18880, 21494, 26925, 20000],
            'Sales_Offices': [0, 6, 6, 6, 12, 12, 12, 12, 12, 12],
            'Office_Efficiency_Avg': [0, 118.5, 120.8, 125.6, 108.2, 135.6, 162.4, 155.3, 166.7, 176.8],
            'Production_Workers': [10, 65, 65, 75, 106, 106, 105, 115, 115, 80],
            'Sales_Workers': [0, 18, 18, 18, 18, 36, 36, 36, 36, 33],
            'Training_Investment': [10500, 44210, 0, 35860, 9360, 79860, 35510, 131020, 101360, 64700]
        }
        self.operations_data = pd.DataFrame(operations_raw_data)
        # Merge with financial data to get units produced/sold
        self.operations_data = pd.merge(self.operations_data, self.df_finance[['Round', 'Units_Produced', 'Units_Sold']], on='Round', how='left')

        self.operations_data['Units_Per_Production_Worker'] = np.where(
            self.operations_data['Production_Workers'] > 0,
            self.operations_data['Units_Produced'] / self.operations_data['Production_Workers'],
            0
        )
        self.operations_data['Sales_Per_Sales_Worker'] = np.where(
            self.operations_data['Sales_Workers'] > 0,
            self.operations_data['Units_Sold'] / self.operations_data['Sales_Workers'],
            0
        )

        # --- Market Data (Competitive Analysis) ---
        self.market_data = pd.DataFrame({
            'Round': list(range(1, 11)),  # Ensure exactly 10 rounds
            'team_5': [0.00, 0.00, 8.28, 20.21, 59.74, 31.45, 14.40, 10.42, 13.65, 19.07],
            'KKKompany': [0.00, 0.00, 30.14, 27.46, 5.21, 41.10, 33.20, 19.91, 43.40, 41.06],
            'V4EScooters_share': [0.00, 0.00, 24.51, 22.29, 0.92, 30.30, 52.48, 57.46, 38.08, 39.87],
            'team_8': [0.00, 0.00, 37.07, 30.04, 62.42, 14.20, 3.90, 8.98, 2.73, 0.00]
        })

        # Define mapping for team names and their colors for plotting
        self.plot_teams_map = {
            'team_5': ('Zesp√≥≈Ç 5', self.COLORS['purple']),
            'KKKompany': ('KKKompany', self.COLORS['secondary']),
            'V4EScooters_share': ('V4E', self.COLORS['info']),
            'team_8': ('Zesp√≥≈Ç 8', self.COLORS['danger'])
        }
        self.competitor_columns_plot = list(self.plot_teams_map.keys())

        # Calculate V4EScooters rank and market leader information
        self.market_data['V4EScooters_rank'] = self.market_data[self.competitor_columns_plot].rank(axis=1, method='min', ascending=False)['V4EScooters_share']
        self.market_data['market_leader_share'] = self.market_data[self.competitor_columns_plot].max(axis=1)
        self.market_data['V4EScooters_relative_to_leader'] = np.where(
            self.market_data['market_leader_share'] > 0,
            (self.market_data['V4EScooters_share'] / self.market_data['market_leader_share']) * 100,
            0
        )

        # --- Product Data for V4E Scooters ---
        self.product_data_v4e = pd.DataFrame({
            'Round': list(range(1, 11)),  # Ensure exactly 10 rounds
            # Production data (current turn)
            'Fashionistas_Produced_V4E': [0, 1500, 1000, 3000, 3000, 3000, 3500, 4000, 4000, 4000],
            'Geeks_Produced_V4E': [0, 4675, 5000, 7000, 7000, 7000, 7000, 8000, 9000, 9000],
            'Savers_Produced_V4E': [0, 325, 500, 2000, 2000, 2000, 2000, 3500, 4000, 4000],
            # Sales data with proper lag
            'Fashionistas_Sold_V4E': [0, 1500, 3000, 3000, 0, 4000, 4434, 4566, 4400, 4400],
            'Geeks_Sold_V4E': [0, 4675, 7000, 7000, 151, 5896, 13831, 10033, 8704, 8704],
            'Savers_Sold_V4E': [0, 325, 2000, 1849, 0, 1848, 4059, 5274, 4000, 4000],
            # Price data from actual sales
            'Fashionistas_Price_V4E': [0, 200.00, 242.50, 242.50, 242.50, 281.25, 281.67, 281.67, 281.67, 281.67],
            'Geeks_Price_V4E': [0, 150.00, 239.17, 239.17, 231.46, 275.18, 277.50, 277.50, 277.50, 277.50],
            'Savers_Price_V4E': [0, 100.00, 141.67, 141.67, 141.67, 156.82, 156.67, 156.67, 156.67, 156.67]
        })

        self.product_data_v4e['Total_Produced_V4E'] = (
            self.product_data_v4e['Fashionistas_Produced_V4E'] +
            self.product_data_v4e['Geeks_Produced_V4E'] +
            self.product_data_v4e['Savers_Produced_V4E']
        )
        self.product_data_v4e['Total_Sold_V4E'] = (
            self.product_data_v4e['Fashionistas_Sold_V4E'] +
            self.product_data_v4e['Geeks_Sold_V4E'] +
            self.product_data_v4e['Savers_Sold_V4E']
        )
        self.product_data_v4e['Avg_Selling_Price_V4E'] = np.where(
            self.product_data_v4e['Total_Sold_V4E'] > 0,
            ((self.product_data_v4e['Fashionistas_Price_V4E'] * self.product_data_v4e['Fashionistas_Sold_V4E'] +
              self.product_data_v4e['Geeks_Price_V4E'] * self.product_data_v4e['Geeks_Sold_V4E'] +
              self.product_data_v4e['Savers_Price_V4E'] * self.product_data_v4e['Savers_Sold_V4E']) /
             self.product_data_v4e['Total_Sold_V4E']),
            0
        )

        # --- Geographical Data ---
        geo_raw_data = {
            'City': ['Bratys≈Çawa', 'Budapesz', 'Praga', 'Sofia', 'Warszawa', 'Zagrzeb'],
            'R1_Success_V4E': [0, 0, 0, 0, 0, 0],
            'R2_Success_V4E': [100, 100, 100, 100, 100, 100],
            'R3_Success_V4E': [100, 100, 100, 100, 100, 100],
            'R4_Success_V4E': [100, 100, 100, 100, 100, 100],
            'R5_Success_V4E': [6.3, 0, 0, 1.7, 0, 0],
            'R6_Success_V4E': [78.1, 85.4, 50.4, 100, 81.8, 58.9],
            'R7_Success_V4E': [100, 100, 100, 100, 100, 100],
            'R8_Success_V4E': [86.7, 94.0, 98.7, 100, 100, 97.9],
            'R9_Success_V4E': [95.7, 96.7, 100, 100, 100, 96.2],
            'R10_Success_V4E': [95.0, 96.0, 99.0, 100, 99.5, 96.0],  # Estimated values for R10
            'Ref_Avg_Price_Market': [235, 240, 250, 237, 237, 233],
            'Market_Size_Category': ['Ma≈Çy', 'Du≈ºy', 'Du≈ºy', '≈öredni', 'Du≈ºy', '≈öredni'],
            'Competition_Level': ['Wysoka', '≈örednia', 'Wysoka', 'Niska', '≈örednia', '≈örednia']
        }

        self.df_geo_v4e = pd.DataFrame(geo_raw_data)
        # Calculate average success rate for later rounds (R7-R10)
        self.df_geo_v4e['Avg_Success_V4E_R7_R10'] = self.df_geo_v4e[['R7_Success_V4E', 'R8_Success_V4E', 'R9_Success_V4E', 'R10_Success_V4E']].mean(axis=1)

    def _print_header(self):
        """Prints the report header and executive summary."""
        print("=" * 100)
        print(" " * 15 + "üèÜ RAPORT KO≈ÉCOWY Z GRY BIZNESOWEJ - V4EScooters: E-HULAJNOGI PREMIUM üìä")
        print("=" * 100)
        print(f"\nRaport wygenerowany: {datetime.now().strftime('%d %B %Y o %H:%M')}")
        print(f"Okres analizy: Rundy 1-10")
        print(f"Podej≈õcie strategiczne: Pozycjonowanie Premium i Doskona≈Ço≈õƒá Operacyjna")
        print("\n" + "=" * 100)
        print("\nüìä PODSUMOWANIE WYKONAWCZE")
        print("=" * 100)

        # Extracting key metrics for the executive summary
        final_balance = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi = ((final_balance - self.initial_investment) / self.initial_investment) * 100
        final_market_share = self.market_data['V4EScooters_share'].iloc[-1]
        final_rank = int(self.market_data['V4EScooters_rank'].iloc[-1])

        print(f"{'Inwestycja poczƒÖtkowa:':<30} ‚Ç¨{self.initial_investment:,.0f}")
        print(f"{'Ko≈Ñcowe saldo:':<30} ‚Ç¨{final_balance:,.0f}")
        print(f"{'Ca≈Çkowity zwrot z inwestycji:':<30} {total_roi:.2f}%")
        print(f"{'Ko≈Ñcowy udzia≈Ç w rynku:':<30} {final_market_share:.2f}% ({final_rank}. miejsce)")
        print(f"{'Ranking rentowno≈õci:':<30} 1. miejsce (szacunkowo)")
        print(f"{'Filozofia strategiczna:':<30} 'Jako≈õƒá ponad Ilo≈õƒá'")
        print("=" * 100)

    def _plot_financial_performance(self):
        """Generates plots for financial performance analysis."""
        print("\n\nüèõÔ∏è SEKCJA 1: ANALIZA WYNIK√ìW FINANSOWYCH")
        print("=" * 100)

        # Figure 1.1 (4 plots: Balance, Cost Structure, ROI, Profit Margin)
        fig1_1 = plt.figure(figsize=(20, 12))
        gs1_1 = fig1_1.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

        # 1.1 Evolution of Balance with Phases
        ax1_1 = fig1_1.add_subplot(gs1_1[0, 0])
        ax1_1.plot(self.df_finance['Round'], self.df_finance['Ending_Balance'] / 1000,
                   color=self.COLORS['primary'], linewidth=4, marker='o', markersize=12, label='Saldo')
        ax1_1.axhline(y=0, color=self.COLORS['danger'], linestyle='--', alpha=0.7, linewidth=2)

        # Highlight strategic phases
        ax1_1.axvspan(0.5, 3.5, alpha=0.1, color=self.COLORS['info'], label='Faza Fundament√≥w')
        ax1_1.axvspan(3.5, 5.5, alpha=0.1, color=self.COLORS['danger'], label='Faza Kryzysu')
        ax1_1.axvspan(5.5, 7.5, alpha=0.1, color=self.COLORS['success'], label='Faza Odbudowy')
        ax1_1.axvspan(7.5, 10.5, alpha=0.1, color=self.COLORS['purple'], label='Faza Optymalizacji')

        # Annotations for key balance points
        ax1_1.annotate('Inwestycja poczƒÖtkowa\n‚Ç¨1M', xy=(1, 749), xytext=(1, 2000),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['gray'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['gray']))
        ax1_1.annotate('Pierwszy kryzys\n-‚Ç¨144,627', xy=(5, -144.627), xytext=(5, -1500),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['danger'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['danger']))
        ax1_1.annotate('Rekordowy zysk\n‚Ç¨3.34M', xy=(7, 3340), xytext=(7, 4500),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['success'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['success']))
        ax1_1.annotate('Ostateczne zwyciƒôstwo\n‚Ç¨5.98M', xy=(10, 5977), xytext=(10, 7000),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['purple'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['purple']))

        ax1_1.set_xlabel('Runda')
        ax1_1.set_ylabel('Saldo (tys. ‚Ç¨)')
        ax1_1.set_title('Podr√≥≈º finansowa: Od startupu do sukcesu')
        ax1_1.grid(True, alpha=0.3)
        ax1_1.legend(loc='upper left', frameon=True, fancybox=True, shadow=True)
        ax1_1.set_xlim(0.5, 10.5)
        ax1_1.set_ylim(-2000, 8000)
        ax1_1.set_xticks(range(1, 11))

        # 1.2 Evolution of Cost Structure (Stacked Bar Chart)
        ax1_2 = fig1_1.add_subplot(gs1_1[0, 1])
        cost_categories = ['Marketing', 'Produkcja', 'HR', 'Operacje']
        rounds_sample = list(range(1, 11))  # Use all rounds
        cost_data = np.array([
            self.df_finance['Marketing_Spend'].values / 1000,
            self.df_finance['Production_Cost'].values / 1000,
            self.df_finance['HR_Cost'].values / 1000,
            self.df_finance['Operations_Cost'].values / 1000
        ])
        bottom = np.zeros(len(rounds_sample))
        colors_cost = [self.COLORS['warning'], self.COLORS['primary'], self.COLORS['success'], self.COLORS['info']]
        for i, category in enumerate(cost_categories):
            ax1_2.bar(rounds_sample, cost_data[i], bottom=bottom, label=category, color=colors_cost[i], alpha=0.8)
            bottom += cost_data[i]

        ax1_2.set_xlabel('Runda')
        ax1_2.set_ylabel('Koszt (tys. ‚Ç¨)')
        ax1_2.set_title('Ewolucja struktury koszt√≥w')
        ax1_2.legend(loc='upper left', frameon=True)
        ax1_2.grid(True, axis='y', alpha=0.3)
        ax1_2.set_xticks(range(1, 11))

        # 1.3 ROI Progress (Line plot with fill)
        ax1_3 = fig1_1.add_subplot(gs1_1[1, 0])
        roi = ((self.df_finance['Ending_Balance'] - self.initial_investment) / self.initial_investment * 100)
        ax1_3.plot(self.df_finance['Round'], roi, color=self.COLORS['purple'], linewidth=4, marker='d', markersize=12)
        ax1_3.fill_between(self.df_finance['Round'], roi, 0, where=(roi >= 0), alpha=0.3, color=self.COLORS['success'], label='Pozytywne ROI')
        ax1_3.fill_between(self.df_finance['Round'], roi, 0, where=(roi < 0), alpha=0.3, color=self.COLORS['danger'], label='Negatywne ROI')
        ax1_3.set_xlabel('Runda')
        ax1_3.set_ylabel('ROI (%)')
        ax1_3.set_title('Postƒôp zwrotu z inwestycji')
        ax1_3.grid(True, alpha=0.3)
        ax1_3.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax1_3.legend(frameon=True)
        ax1_3.set_xticks(range(1, 11))

        # Annotation for final ROI
        ax1_3.annotate(f'Ko≈Ñcowe ROI\n{roi.iloc[-1]:.1f}%', xy=(10, roi.iloc[-1]),
                       xytext=(8.5, roi.iloc[-1] - 100 if roi.iloc[-1] > 300 else roi.iloc[-1] + 100),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['purple'], lw=2),
                       fontsize=12, ha='center', fontweight='bold', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['purple']))

        # 1.4 Profit Margin Trend (Bar Chart)
        ax1_4 = fig1_1.add_subplot(gs1_1[1, 1])
        valid_margins_df = self.df_finance[self.df_finance['Implied_Revenue'] > 0]
        ax1_4.bar(valid_margins_df['Round'], valid_margins_df['Profit_Margin'], color=self.COLORS['info'], alpha=0.7)
        ax1_4.set_xlabel('Runda')
        ax1_4.set_ylabel('Mar≈ºa zysku (%)')
        ax1_4.set_title('Ewolucja mar≈ºy zysku (przy pozytywnych przychodach)')
        ax1_4.grid(True, axis='y', alpha=0.3)
        ax1_4.set_xticks(range(1, 11))

        # Add profit margin values on top of bars
        for i, (r, m) in enumerate(zip(valid_margins_df['Round'], valid_margins_df['Profit_Margin'])):
            ax1_4.text(r, m + (2 if m > 0 else -5), f'{m:.1f}%', ha='center', fontweight='bold')

        fig1_1.suptitle('Analiza wynik√≥w finansowych - Czƒô≈õƒá 1: Podr√≥≈º finansowa i rentowno≈õƒá', fontsize=20, fontweight='bold', y=0.98)
        fig1_1.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

        # Figure 1.2 (2 plots: Revenue vs. Costs Waterfall, Financial Key Metrics Summary)
        fig1_2 = plt.figure(figsize=(20, 8))
        gs1_2 = fig1_2.add_gridspec(1, 2, hspace=0.3, wspace=0.3)

        # 1.5 Revenue vs. Costs Waterfall (Bar and Line Combined)
        ax1_5 = fig1_2.add_subplot(gs1_2[0, 0])
        rounds_plot = self.df_finance['Round'].values
        revenues = self.df_finance['Implied_Revenue'].values / 1000
        costs = self.df_finance['Total_Costs'].values / 1000
        profits = self.df_finance['Net_Change'].values / 1000

        x_plot = np.arange(len(rounds_plot))
        width_plot = 0.8

        bars1 = ax1_5.bar(x_plot, revenues, width_plot, label='Szacowane Przychody', color=self.COLORS['success'], alpha=0.7)
        bars2 = ax1_5.bar(x_plot, -costs, width_plot, label='Ca≈Çkowite Koszty', color=self.COLORS['warning'], alpha=0.7)
        ax1_5.plot(x_plot, profits, color=self.COLORS['primary'], linewidth=3, marker='o', markersize=10, label='Zysk/Strata Netto', zorder=5)

        # Annotate specific rounds for clarity
        for i in [6, 7, 9]:  # Example rounds for annotation
            if i < len(revenues) and revenues[i] > 0:
                ax1_5.text(i, revenues[i] + 100, f'‚Ç¨{revenues[i]:.0f}k', ha='center', va='bottom', fontweight='bold')
            if i < len(profits):
                ax1_5.text(i, profits[i] + 200, f'{profits[i]:+.0f}k', ha='center', va='bottom', color=self.COLORS['primary'], fontweight='bold')

        ax1_5.set_xlabel('Runda')
        ax1_5.set_ylabel('Kwota (tys. ‚Ç¨)')
        ax1_5.set_title('Analiza przychod√≥w vs. koszt√≥w - Ca≈Ça podr√≥≈º')
        ax1_5.set_xticks(x_plot)
        ax1_5.set_xticklabels(rounds_plot)
        ax1_5.legend(loc='upper left', frameon=True)
        ax1_5.grid(True, axis='y', alpha=0.3)
        ax1_5.axhline(y=0, color='black', linestyle='-', alpha=0.3)

        # 1.6 Summary of Key Financial Indicators (Table)
        ax1_6 = fig1_2.add_subplot(gs1_2[0, 1])
        ax1_6.axis('off')

        # Calculate summary metrics
        total_investment_val = self.df_finance['Infrastructure_Invest'].sum()
        total_revenue_val = self.df_finance['Implied_Revenue'].sum()
        total_costs_val = self.df_finance['Total_Costs'].sum()
        net_profit_val = self.df_finance['Ending_Balance'].iloc[-1] - self.initial_investment
        avg_profit_margin_val = self.df_finance[self.df_finance['Implied_Revenue'] > 0]['Profit_Margin'].mean()

        metrics_summary = [
            ['Wska≈∫nik', 'Warto≈õƒá'],
            ['Ca≈Çkowita inwestycja w infrastrukturƒô', f'‚Ç¨{total_investment_val / 1e3:,.0f}k'],
            ['Ca≈Çkowity szacowany przych√≥d', f'‚Ç¨{total_revenue_val / 1e6:,.1f}M'],
            ['Ca≈Çkowite poniesione koszty', f'‚Ç¨{total_costs_val / 1e6:,.1f}M'],
            ['Ca≈Çkowity zysk netto', f'‚Ç¨{net_profit_val / 1e6:,.2f}M'],
            ['Najwy≈ºsze saldo', f'‚Ç¨{self.df_finance["Ending_Balance"].max() / 1e6:,.2f}M'],
            ['Najni≈ºsze saldo', f'‚Ç¨{self.df_finance["Ending_Balance"].min() / 1e3:,.0f}k'],
            ['≈örednia mar≈ºa zysku (wa≈ºne rundy)', f'{avg_profit_margin_val:.1f}%' if pd.notna(avg_profit_margin_val) else 'N/A'],
            ['Szybko≈õƒá odbudowy po kryzysie', '1-2 rundy']
        ]

        # Create and style the table
        table = ax1_6.table(cellText=metrics_summary, cellLoc='left', loc='center', colWidths=[0.65, 0.35])
        table.auto_set_font_size(False)
        table.set_fontsize(11)
        table.scale(1, 1.8)

        # Apply styling to header row
        for i in range(2):
            table[(0, i)].set_facecolor(self.COLORS['primary'])
            table[(0, i)].set_text_props(weight='bold', color='white')
        # Apply alternating row colors
        for i in range(1, len(metrics_summary)):
            if i % 2 == 0:
                for j in range(2):
                    table[(i, j)].set_facecolor('#e9ecef')

        ax1_6.set_title('Podsumowanie kluczowych wska≈∫nik√≥w finansowych', fontsize=16, pad=20)
        fig1_2.suptitle('Analiza wynik√≥w finansowych - Czƒô≈õƒá 2: Przychody, koszty i podsumowanie', fontsize=20, fontweight='bold', y=0.98)
        fig1_2.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

    def _plot_marketing_strategy(self):
        """Generates plots for marketing strategy analysis."""
        print("\n\nüéØ SEKCJA 2: STRATEGIA MARKETINGOWA - WIELKI EKSPERYMENT")
        print("=" * 100)

        # Print a summary table of the marketing experiment
        experiment_data = self.marketing_full.copy()
        print("\nWIELKI EKSPERYMENT MARKETINGOWY (Wszystkie rundy):")
        print("-" * 100)
        print(f"{'Runda':<8} {'Bud≈ºet (‚Ç¨)':<12} {'Strategia':<25} {'Wska≈∫nik sprzeda≈ºy (%)':<15} {'Udzia≈Ç w rynku (%)':<15} {'Kluczowe odkrycie':<40}")
        print("-" * 100)
        for _, row in experiment_data.iterrows():
            finding = ""
            if row['Round'] == 1:
                finding = "Ustawienia poczƒÖtkowe"
            elif row['Round'] == 7:
                finding = "Du≈ºe wydatki ustalajƒÖ punkt odniesienia"
            elif row['Round'] == 8:
                finding = "Bud≈ºet ‚Ç¨0 zaskakujƒÖco efektywny!"
            elif row['Round'] == 9:
                finding = "Mega wydatki, malejƒÖce zwroty"
            elif row['Round'] == 10:
                finding = "Zoptymalizowane wydatki, dobra efektywno≈õƒá"
            print(f"{row['Round']:<8} {row['Marketing_Spend']:>10,.0f} {row['Campaign_Details']:<25} "
                  f"{row['Sales_Success_Rate']:>14.1f} {row['Market_Share_V4E']:>14.2f} {finding:<40}")

        # Figure for Marketing Strategy (4 plots)
        fig_mkt, axes_mkt = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.35)

        # 2.1 Evolution of Marketing Budget (Bar Chart)
        ax2_1_mkt = axes_mkt[0, 0]
        colors_budget = [self.COLORS['gray']] * 6 + [self.COLORS['warning'], self.COLORS['success'], self.COLORS['danger'], self.COLORS['info']]
        ax2_1_mkt.bar(self.marketing_full['Round'], self.marketing_full['Marketing_Spend'] / 1000, color=colors_budget)
        ax2_1_mkt.set_xlabel('Runda')
        ax2_1_mkt.set_ylabel('Bud≈ºet marketingowy (tys. ‚Ç¨)')
        ax2_1_mkt.set_title('Ewolucja bud≈ºetu marketingowego - Zmiany strategiczne')
        ax2_1_mkt.grid(True, axis='y', alpha=0.3)
        ax2_1_mkt.set_xticks(range(1, 11))

        # Annotations for key marketing campaigns
        ax2_1_mkt.annotate('Intensywna kampania', xy=(7, 310), xytext=(6.5, 400),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['warning'], lw=1.5), ha='center', fontsize=10)
        ax2_1_mkt.annotate('Test ‚Ç¨0', xy=(8, 0), xytext=(7.5, 100),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['success'], lw=1.5), ha='center', fontsize=10)
        ax2_1_mkt.annotate('MEGA wydatki', xy=(9, 1500), xytext=(8.5, 1600),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['danger'], lw=1.5), ha='center', fontsize=10)

        # 2.2 Marketing Paradox: Budget vs. Sales Success Rate (Dual Axis Line Plot) - Turns 5-10 only
        ax2_2_mkt = axes_mkt[0, 1]
        exp_rounds_mkt = experiment_data[experiment_data['Round'] >= 5]['Round'].values
        exp_budget_mkt = experiment_data[experiment_data['Round'] >= 5]['Marketing_Spend'].values / 1000
        exp_sales_success_mkt = experiment_data[experiment_data['Round'] >= 5]['Sales_Success_Rate'].values

        ax2_2_mkt_twin = ax2_2_mkt.twinx()
        line1 = ax2_2_mkt.plot(exp_rounds_mkt, exp_budget_mkt, 'o-', color=self.COLORS['warning'],
                       linewidth=3, markersize=10, label='Bud≈ºet marketingowy (tys. ‚Ç¨)')
        line2 = ax2_2_mkt_twin.plot(exp_rounds_mkt, exp_sales_success_mkt, 's-', color=self.COLORS['success'],
                        linewidth=3, markersize=10, label='Wska≈∫nik sukcesu sprzeda≈ºy (%)')

        ax2_2_mkt.set_xlabel('Runda')
        ax2_2_mkt.set_ylabel('Bud≈ºet marketingowy (tys. ‚Ç¨)', color=self.COLORS['warning'])
        ax2_2_mkt_twin.set_ylabel('Wska≈∫nik sukcesu sprzeda≈ºy (%)', color=self.COLORS['success'])
        ax2_2_mkt.set_title('Paradoks marketingowy: Wydatki vs. sukces sprzeda≈ºy (R5-R10)')
        ax2_2_mkt.grid(True, alpha=0.3)
        ax2_2_mkt.set_xticks(range(5, 11))

        lines = line1 + line2
        labels = [l.get_label() for l in lines]
        ax2_2_mkt.legend(lines, labels, loc='center left', frameon=True, bbox_to_anchor=(0.05, 0.6))

        # 2.3 Analysis of Marketing Spend Impact (simplified version)
        ax2_3_mkt = axes_mkt[1, 0]
        marketing_efficiency = np.where(self.marketing_full['Marketing_Spend'] > 0,
                                      self.marketing_full['Sales_Success_Rate'] / (self.marketing_full['Marketing_Spend'] / 1000),
                                      0)
        ax2_3_mkt.bar(self.marketing_full['Round'], marketing_efficiency, color=self.COLORS['purple'], alpha=0.7)
        ax2_3_mkt.set_xlabel('Runda')
        ax2_3_mkt.set_ylabel('Efektywno≈õƒá marketingowa (sukces % na ‚Ç¨1k)')
        ax2_3_mkt.set_title('Efektywno≈õƒá wydatk√≥w marketingowych')
        ax2_3_mkt.grid(True, axis='y', alpha=0.3)
        ax2_3_mkt.set_xticks(range(1, 11))

        # 2.4 Distribution of Advertising Channels (Grouped Bar Chart)
        ax2_4_mkt = axes_mkt[1, 1]
        ad_channels = ['Direct_Sale_Ads', 'Ambassador_Ads', 'Newspaper_Ads']
        ad_channel_labels = ['Sprzeda≈º bezpo≈õrednia', 'Ambasadorzy', 'Gazety']
        rounds_for_ads_plot = [1,2,3,4,5,6,7, 9, 10]
        ad_data_plot = self.marketing_full[self.marketing_full['Round'].isin(rounds_for_ads_plot)][['Round'] + ad_channels]

        x_ads_plot = np.arange(len(rounds_for_ads_plot))
        width_ads_plot = 0.25
        colors_ads_plot = [self.COLORS['primary'], self.COLORS['success'], self.COLORS['info']]

        for i, channel_col in enumerate(ad_channels):
            ax2_4_mkt.bar(x_ads_plot + i * width_ads_plot - width_ads_plot, ad_data_plot[channel_col],
                          width_ads_plot, label=ad_channel_labels[i], color=colors_ads_plot[i], alpha=0.8)

        ax2_4_mkt.set_xlabel('Runda')
        ax2_4_mkt.set_ylabel('Liczba umieszczonych reklam')
        ax2_4_mkt.set_title('Ewolucja miksu kana≈Ç√≥w reklamowych (kluczowe rundy)')
        ax2_4_mkt.set_xticks(x_ads_plot)
        ax2_4_mkt.set_xticklabels(rounds_for_ads_plot)
        ax2_4_mkt.legend(frameon=True, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.15))
        ax2_4_mkt.grid(True, axis='y', alpha=0.3)

        fig_mkt.suptitle('Analiza strategii marketingowej - Rewolucja efektywno≈õci', fontsize=20, fontweight='bold', y=0.98)
        fig_mkt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_unit_economics(self):
        """Generates plots for unit economics analysis."""
        print("\n\nüí∞ SEKCJA 2.5: EKONOMIKA JEDNOSTKOWA - HISTORIA RENTOWNO≈öCI NA JEDNOSTKƒò")
        print("=" * 100)

        # Filter for rounds where units were sold to calculate unit economics
        unit_economics_df = self.df_finance[self.df_finance['Units_Sold'] > 0].copy()

        # Print a summary table of unit economics evolution
        print("\nEWOLUCJA EKONOMIKI JEDNOSTKOWEJ (na sprzedanƒÖ jednostkƒô):")
        print("-" * 120)
        print(f"{'Runda':<7} {'Sprzedane jednostki':<12} {'Koszt prod./jedn.':<14} {'Koszt mkt./jedn.':<14} {'Ca≈Çkowity koszt/jedn.':<15} {'Przych√≥d/jedn.':<14} {'Zysk/jedn.':<12} {'Mar≈ºa zysku/jedn. (%)':<18}")
        print("-" * 120)
        for _, row in unit_economics_df.iterrows():
            unit_profit_margin = (row['Profit_Per_Unit_Sold'] / row['Revenue_Per_Unit_Sold'] * 100) if row['Revenue_Per_Unit_Sold'] > 0 else 0
            print(f"{row['Round']:<7} {row['Units_Sold']:>10,} "
                  f"‚Ç¨{row['Production_Cost_Per_Unit']:>12.2f} ‚Ç¨{row['Marketing_Cost_Per_Unit_Sold']:>12.2f} "
                  f"‚Ç¨{row['Total_Cost_Per_Unit_Sold']:>13.2f} ‚Ç¨{row['Revenue_Per_Unit_Sold']:>12.2f} "
                  f"‚Ç¨{row['Profit_Per_Unit_Sold']:>10.2f} {unit_profit_margin:>16.1f}%")

        # Figure 2.5.1 (4 plots: Production Cost/Unit, Marketing Cost/Unit, Total Cost vs Revenue/Unit, Profit/Unit)
        fig_ue1, axes_ue1 = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 2.5.1 Evolution of Production Cost Per Unit (Line Plot)
        ax2_5_1_ue = axes_ue1[0, 0]
        valid_rounds_ue = unit_economics_df['Round']
        ax2_5_1_ue.plot(valid_rounds_ue, unit_economics_df['Production_Cost_Per_Unit'], 'o-',
                        color=self.COLORS['primary'], linewidth=3, markersize=8, label='Koszt prod./jedn.')
        ax2_5_1_ue.set_xlabel('Runda')
        ax2_5_1_ue.set_ylabel('Koszt produkcji na jednostkƒô (‚Ç¨)')
        ax2_5_1_ue.set_title('Efektywno≈õƒá koszt√≥w produkcji')
        ax2_5_1_ue.grid(True, alpha=0.3)
        ax2_5_1_ue.set_xticks(range(1, 11))

        if not unit_economics_df['Production_Cost_Per_Unit'].empty:
            avg_prod_cost_ue = unit_economics_df['Production_Cost_Per_Unit'].mean()
            ax2_5_1_ue.axhline(y=avg_prod_cost_ue, color=self.COLORS['gray'], linestyle='--', alpha=0.7, label=f'≈örednia: ‚Ç¨{avg_prod_cost_ue:.2f}')
            ax2_5_1_ue.legend(frameon=True)

        # 2.5.2 Marketing Cost Per Unit Sold (Bar Chart)
        ax2_5_2_ue = axes_ue1[0, 1]
        bars_mkt_cost_ue = ax2_5_2_ue.bar(valid_rounds_ue, unit_economics_df['Marketing_Cost_Per_Unit_Sold'],
                                         color=self.COLORS['warning'], alpha=0.7, label='Koszt mkt./sprzedana jedn.')
        ax2_5_2_ue.set_xlabel('Runda')
        ax2_5_2_ue.set_ylabel('Koszt marketingowy na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_2_ue.set_title('Efektywno≈õƒá koszt√≥w marketingowych (na sprzeda≈º)')
        ax2_5_2_ue.grid(True, axis='y', alpha=0.3)
        ax2_5_2_ue.set_xticks(range(1, 11))
        ax2_5_2_ue.legend(frameon=True)

        # Highlight Round 9's high marketing cost
        if not unit_economics_df[unit_economics_df['Round'] == 9].empty:
            round_9_mkt_cost = unit_economics_df[unit_economics_df['Round'] == 9]['Marketing_Cost_Per_Unit_Sold'].iloc[0]
            if pd.notna(round_9_mkt_cost):
                ax2_5_2_ue.annotate(f'R9: ‚Ç¨{round_9_mkt_cost:.2f}\n(Mega Kampania)', xy=(9, round_9_mkt_cost),
                                   xytext=(8.5, round_9_mkt_cost + 20 if round_9_mkt_cost < 100 else round_9_mkt_cost - 20),
                                   arrowprops=dict(arrowstyle='->', color=self.COLORS['danger']),
                                   fontsize=10, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=self.COLORS['danger']))

        # 2.5.3 Total Cost vs. Revenue Per Unit Sold (Grouped Bar Chart)
        ax2_5_3_ue = axes_ue1[1, 0]
        x_pos_ue = np.arange(len(valid_rounds_ue))
        width_ue = 0.35
        ax2_5_3_ue.bar(x_pos_ue - width_ue / 2, unit_economics_df['Total_Cost_Per_Unit_Sold'], width_ue,
                       label='Ca≈Çkowity koszt/sprzedana jedn.', color=self.COLORS['danger'], alpha=0.7)
        ax2_5_3_ue.bar(x_pos_ue + width_ue / 2, unit_economics_df['Revenue_Per_Unit_Sold'], width_ue,
                       label='Przych√≥d/sprzedana jedn.', color=self.COLORS['success'], alpha=0.7)
        ax2_5_3_ue.set_xlabel('Runda')
        ax2_5_3_ue.set_ylabel('Kwota na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_3_ue.set_title('Ekonomika jednostkowa: Koszt vs. Przych√≥d')
        ax2_5_3_ue.set_xticks(x_pos_ue)
        ax2_5_3_ue.set_xticklabels(valid_rounds_ue)
        ax2_5_3_ue.legend(frameon=True, loc='upper left')
        ax2_5_3_ue.grid(True, axis='y', alpha=0.3)

        # 2.5.4 Evolution of Profit Per Unit Sold (Line Plot with Fill)
        ax2_5_4_ue = axes_ue1[1, 1]
        profit_per_unit_sold_ue = unit_economics_df['Profit_Per_Unit_Sold']
        ax2_5_4_ue.plot(valid_rounds_ue, profit_per_unit_sold_ue, 'o-',
                        color=self.COLORS['purple'], linewidth=4, markersize=10, label='Zysk/sprzedana jedn.')
        ax2_5_4_ue.fill_between(valid_rounds_ue, profit_per_unit_sold_ue, 0,
                                where=(profit_per_unit_sold_ue >= 0), alpha=0.3, color=self.COLORS['success'], interpolate=True)
        ax2_5_4_ue.fill_between(valid_rounds_ue, profit_per_unit_sold_ue, 0,
                                where=(profit_per_unit_sold_ue < 0), alpha=0.3, color=self.COLORS['danger'], interpolate=True)
        ax2_5_4_ue.set_xlabel('Runda')
        ax2_5_4_ue.set_ylabel('Zysk na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_4_ue.set_title('Podr√≥≈º rentowno≈õci na jednostkƒô')
        ax2_5_4_ue.grid(True, alpha=0.3)
        ax2_5_4_ue.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax2_5_4_ue.set_xticks(range(1, 11))
        ax2_5_4_ue.legend(frameon=True)

        # Annotate best and worst profit per unit
        if not profit_per_unit_sold_ue.empty:
            best_profit_ue = profit_per_unit_sold_ue.max()
            worst_profit_ue = profit_per_unit_sold_ue.min()
            best_round_ue = unit_economics_df.loc[profit_per_unit_sold_ue.idxmax(), 'Round']
            worst_round_ue = unit_economics_df.loc[profit_per_unit_sold_ue.idxmin(), 'Round']
            ax2_5_4_ue.annotate(f'Najlepszy: ‚Ç¨{best_profit_ue:.2f} (R{best_round_ue})',
                                xy=(best_round_ue, best_profit_ue), xytext=(best_round_ue, best_profit_ue + 20),
                                arrowprops=dict(arrowstyle='->', color=self.COLORS['success']),
                                fontsize=10, ha='center', fontweight='bold')
            if worst_profit_ue < 0:
                ax2_5_4_ue.annotate(f'Najgorszy: ‚Ç¨{worst_profit_ue:.2f} (R{worst_round_ue})',
                                    xy=(worst_round_ue, worst_profit_ue), xytext=(worst_round_ue, worst_profit_ue - 20),
                                    arrowprops=dict(arrowstyle='->', color=self.COLORS['danger']),
                                    fontsize=10, ha='center', fontweight='bold')

        fig_ue1.suptitle('Dog≈Çƒôbna analiza ekonomiki jednostkowej - Czƒô≈õƒá 1: Koszty, przychody i zysk na jednostkƒô',
                         fontsize=20, fontweight='bold', y=0.98)
        fig_ue1.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_operational_excellence(self):
        """Generates plots for operational excellence analysis."""
        print("\n\n‚öôÔ∏è SEKCJA 3: DOSKONA≈ÅO≈öƒÜ OPERACYJNA - DROGA DO 200%")
        print("=" * 100)

        # Print a summary table of operational milestones
        print("\nKAMIENIE MILOWE EFEKTYWNO≈öCI OPERACYJNEJ:")
        print("-" * 100)
        print(f"{'Runda':<8} {'Efektywno≈õƒá linii Fash (%)':<20} {'Efektywno≈õƒá linii Geeks (%)':<20} {'Pojemno≈õƒá (jednostki)':<18} {'Produkcja (jednostki)':<20} {'Wska≈∫nik sprzeda≈ºy (%)':<15} {'Jednostki/prac. prod.':<15}")
        print("-" * 100)
        for _, row in self.operations_data.iterrows():
            sales_rate_ops = (row['Units_Sold'] / row['Units_Produced'] * 100) if row['Units_Produced'] > 0 else 0
            print(f"{row['Round']:<8} {row['Fashionistas_Eff']:>18.1f} {row['Geeks_Eff']:>18.1f} "
                  f"{row['Production_Capacity']:>16,} {row['Units_Produced']:>18,} {sales_rate_ops:>14.1f} "
                  f"{row['Units_Per_Production_Worker']:>14.1f}")

        # Figure 3.1 (4 plots: Line Efficiency, Capacity Utilization, Sales Network, Distribution Disaster)
        fig_ops1, axes_ops1 = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 3.1 Evolution of Efficiency to Perfection (Line Plot)
        ax3_1_ops = axes_ops1[0, 0]
        rounds_ops_plot = self.operations_data['Round']
        fash_eff_plot = self.operations_data['Fashionistas_Eff']
        geeks_eff_plot = self.operations_data['Geeks_Eff']

        ax3_1_ops.plot(rounds_ops_plot, fash_eff_plot, 'o-', color=self.COLORS['primary'],
                       linewidth=3, markersize=8, label='Efektywno≈õƒá linii Fashionistas')
        ax3_1_ops.plot(rounds_ops_plot, geeks_eff_plot, 's-', color=self.COLORS['secondary'],
                       linewidth=3, markersize=8, label='Efektywno≈õƒá linii Geeks')
        ax3_1_ops.axhline(y=200, color=self.COLORS['success'], linestyle='--', linewidth=2, alpha=0.7, label='200% Maks. efektywno≈õƒá')

        # Highlight areas approaching 200% efficiency
        ax3_1_ops.fill_between(rounds_ops_plot, fash_eff_plot, 200, where=(fash_eff_plot >= 180),
                               alpha=0.2, color=self.COLORS['success'], interpolate=True)
        ax3_1_ops.fill_between(rounds_ops_plot, geeks_eff_plot, 200, where=(geeks_eff_plot >= 180),
                               alpha=0.2, color=self.COLORS['success'], interpolate=True)

        ax3_1_ops.set_xlabel('Runda')
        ax3_1_ops.set_ylabel('Efektywno≈õƒá linii (%)')
        ax3_1_ops.set_title('Podr√≥≈º do doskona≈Ço≈õci operacyjnej (Efektywno≈õƒá linii)')
        ax3_1_ops.grid(True, alpha=0.3)
        ax3_1_ops.set_ylim(0, 210)
        ax3_1_ops.set_xticks(range(1, 11))

        # Annotate 200% milestones
        ax3_1_ops.scatter([9], [200], s=250, color=self.COLORS['success'], marker='*', zorder=5,
                          label='Fashionistas osiƒÖga 200% (R9)', edgecolors='black')
        ax3_1_ops.scatter([10], [200], s=350, color=self.COLORS['purple'], marker='*', zorder=5,
                          label='Obie linie na 200% (R10)', edgecolors='black')

        handles, labels = ax3_1_ops.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        ax3_1_ops.legend(by_label.values(), by_label.keys(), frameon=True, loc='lower right')

        # 3.2 Production Capacity Utilization (Bar Chart)
        ax3_2_ops = axes_ops1[0, 1]
        utilization_ops = (self.operations_data['Units_Produced'] / self.operations_data['Production_Capacity'] * 100)
        bars_util_ops = ax3_2_ops.bar(rounds_ops_plot, utilization_ops, color=self.COLORS['info'], alpha=0.7)

        ax3_2_ops.set_xlabel('Runda')
        ax3_2_ops.set_ylabel('Wykorzystanie zdolno≈õci (%)')
        ax3_2_ops.set_title('Wykorzystanie zdolno≈õci produkcyjnych')
        ax3_2_ops.grid(True, axis='y', alpha=0.3)
        ax3_2_ops.axhline(y=100, color='red', linestyle='--', alpha=0.5, label='100% wykorzystania')
        ax3_2_ops.set_xticks(range(1, 11))
        ax3_2_ops.legend(frameon=True)

        # Color bars based on utilization level and add text
        for bar_item, util_val in zip(bars_util_ops, utilization_ops):
            if util_val > 95:
                bar_item.set_color(self.COLORS['danger'])
            elif util_val > 80:
                bar_item.set_color(self.COLORS['warning'])
            else:
                bar_item.set_color(self.COLORS['info'])
            ax3_2_ops.text(bar_item.get_x() + bar_item.get_width() / 2, util_val + 1,
                           f'{util_val:.0f}%', ha='center', va='bottom', fontsize=9)

        # 3.3 Development and Efficiency of Sales Network (Dual Axis Bar and Line)
        ax3_3_ops = axes_ops1[1, 0]
        offices_plot = self.operations_data['Sales_Offices']
        office_eff_plot = self.operations_data['Office_Efficiency_Avg']

        ax3_3_ops_twin = ax3_3_ops.twinx()
        bars_offices_ops = ax3_3_ops.bar(rounds_ops_plot, offices_plot, alpha=0.6, color=self.COLORS['purple'], label='Biura sprzeda≈ºy (#)')
        line_office_eff_ops = ax3_3_ops_twin.plot(rounds_ops_plot, office_eff_plot, 'go-', linewidth=3, markersize=8, label='≈örednia efektywno≈õƒá biura (%)')

        ax3_3_ops.set_xlabel('Runda')
        ax3_3_ops.set_ylabel('Liczba biur sprzeda≈ºy', color=self.COLORS['purple'])
        ax3_3_ops_twin.set_ylabel('≈örednia efektywno≈õƒá biura (%)', color='green')
        ax3_3_ops.set_title('Rozw√≥j i efektywno≈õƒá sieci sprzeda≈ºy')
        ax3_3_ops.grid(True, alpha=0.3, axis='y', which='major', linestyle='dotted')
        ax3_3_ops_twin.grid(True, alpha=0.3, axis='y', which='major', linestyle=':')
        ax3_3_ops.set_xticks(range(1, 11))

        lines_sales_net = list(bars_offices_ops[0:1]) + line_office_eff_ops
        labels_sales_net = [l.get_label() for l in lines_sales_net]
        ax3_3_ops.legend(lines_sales_net, labels_sales_net, loc='upper left', frameon=True)

        # 3.4 Distribution Disaster (Focus on Rounds 4-6) (Grouped Bar Chart)
        ax3_4_ops = axes_ops1[1, 1]
        disaster_rounds_ops = [4, 5, 6]
        disaster_data = self.operations_data[self.operations_data['Round'].isin(disaster_rounds_ops)]
        produced_disaster_ops = disaster_data['Units_Produced'].values
        sold_disaster_ops = disaster_data['Units_Sold'].values

        if len(produced_disaster_ops) > 0 and produced_disaster_ops.any() > 0:
            success_rate_disaster_ops = np.where(produced_disaster_ops > 0, (sold_disaster_ops / produced_disaster_ops * 100), 0)
        else:
            success_rate_disaster_ops = np.array([0.0] * len(produced_disaster_ops))

        x_disaster_plot = np.arange(len(disaster_rounds_ops))
        width_disaster_plot = 0.35
        ax3_4_ops.bar(x_disaster_plot - width_disaster_plot / 2, produced_disaster_ops, width_disaster_plot,
                      label='Wyprodukowane jednostki', color=self.COLORS['primary'], alpha=0.7)
        ax3_4_ops.bar(x_disaster_plot + width_disaster_plot / 2, sold_disaster_ops, width_disaster_plot,
                      label='Sprzedane jednostki', color=self.COLORS['success'], alpha=0.7)

        # Add sales rate annotations
        for i_dis, (p_dis, s_dis, rate_dis) in enumerate(zip(produced_disaster_ops, sold_disaster_ops, success_rate_disaster_ops)):
            ax3_4_ops.text(i_dis, max(p_dis, s_dis) + 500, f'{rate_dis:.1f}% sprzedanych',
                           ha='center', fontweight='bold', fontsize=10,
                           color=self.COLORS['danger'] if rate_dis < 10 else self.COLORS['info'])

        ax3_4_ops.set_xlabel('Runda')
        ax3_4_ops.set_ylabel('Liczba jednostek')
        ax3_4_ops.set_title('Katastrofa dystrybucyjna (Rundy 4-6)')
        ax3_4_ops.set_xticks(x_disaster_plot)
        ax3_4_ops.set_xticklabels(disaster_rounds_ops)
        ax3_4_ops.legend(frameon=True)
        ax3_4_ops.grid(True, axis='y', alpha=0.3)

        fig_ops1.suptitle('Analiza doskona≈Ço≈õci operacyjnej - Czƒô≈õƒá 1: Efektywno≈õƒá linii i wykorzystanie',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_ops1.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_competitive_dynamics(self):
        """Generates plots for competitive dynamics and market positioning."""
        print("\n\nüèÅ SEKCJA 4: DYNAMIKA KONKURENCYJNA - STRATEGIA POZYCJONOWANIA PREMIUM")
        print("=" * 100)

        # Print a summary table of market position evolution
        print("\nEWOLUCJA POZYCJI NA RYNKU:")
        print("-" * 120)
        print(f"{'Runda':<7} {'Nasz udzia≈Ç (%)':<15} {'Ranking':<6} {'Udzia≈Ç lidera (%)':<25} {'R√≥≈ºnica do lidera (%)':<20} {'Faza strategii':<30}")
        print("-" * 120)
        for _, row in self.market_data.iterrows():
            leader_share_val = row['market_leader_share']
            leader_col_name = self.market_data[self.competitor_columns_plot].idxmax(axis=1)[row.name] if leader_share_val > 0 else None
            leader_display_name = self.plot_teams_map.get(leader_col_name, ('N/A', ''))[0] if leader_col_name else "N/A"
            gap_to_leader_val = row['V4EScooters_share'] - leader_share_val

            phase_strat = ""
            if row['Round'] <= 3:
                phase_strat = "Wej≈õcie na rynek i wczesne prowadzenie"
            elif row['Round'] <= 5:
                phase_strat = "Ekspansja i zmienno≈õƒá/kryzys"
            elif row['Round'] <= 7:
                phase_strat = "Repozycjonowanie premium i odbudowa"
            else:
                phase_strat = "Optymalizacja zysku i stabilno≈õƒá"

            print(f"{row['Round']:<7} {row['V4EScooters_share']:>13.2f} {int(row['V4EScooters_rank']):>4} "
                  f"{leader_share_val:>8.2f} ({leader_display_name:<15}) {gap_to_leader_val:>+18.2f} {phase_strat:<30}")

        # Figure for Competitive Dynamics (4 plots)
        fig_comp, axes_comp = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 4.1 Market Share Race (Line Plot for all competitors)
        ax4_1_comp = axes_comp[0, 0]
        rounds_comp_plot_data = self.market_data['Round']
        for col_name, (plot_label, color_val) in self.plot_teams_map.items():
            data_plot = self.market_data[col_name]
            linewidth = 4 if col_name == 'V4EScooters_share' else 2.5
            marker = 'o-' if col_name == 'V4EScooters_share' else 's--'
            ax4_1_comp.plot(rounds_comp_plot_data, data_plot, marker,
                            linewidth=linewidth, markersize=8 if col_name == 'V4EScooters_share' else 6,
                            label=plot_label, color=color_val, alpha=0.9 if col_name == 'V4EScooters_share' else 0.7)

        ax4_1_comp.set_xlabel('Runda')
        ax4_1_comp.set_ylabel('Udzia≈Ç w rynku (%)')
        ax4_1_comp.set_title('Ewolucja udzia≈Çu w rynku - Wy≈õcig konkurencyjny')
        ax4_1_comp.legend(frameon=True, loc='upper left', ncol=2)
        ax4_1_comp.grid(True, alpha=0.3)
        ax4_1_comp.set_xticks(range(1, 11))

        # Highlight key strategic points
        ax4_1_comp.axvspan(4.5, 5.5, alpha=0.1, color=self.COLORS['danger'], label='Punkt kryzysowy V4E')
        ax4_1_comp.axvspan(6.5, 7.5, alpha=0.1, color=self.COLORS['success'], label='Zwrot V4E na premium')

        # 4.2 Strategic Positioning Matrix (Scatter Plot)
        ax4_2_comp = axes_comp[0, 1]
        final_shares_comp = [self.market_data[col].iloc[-1] for col in self.competitor_columns_plot]

        # Estimate V4E's final profit margin per unit for positioning
        v4e_final_profit_per_unit = self.df_finance[self.df_finance['Round'] == 10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round'] == 10].empty else 0
        v4e_final_rev_per_unit = self.df_finance[self.df_finance['Round'] == 10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round'] == 10].empty else 1
        v4e_final_margin_est = (v4e_final_profit_per_unit / v4e_final_rev_per_unit * 100) if v4e_final_rev_per_unit > 0 else 0

        # Estimated margins for competitors (illustrative based on typical game outcomes)
        estimated_margins_comp = []
        for col_name in self.competitor_columns_plot:
            if col_name == 'V4EScooters_share':
                estimated_margins_comp.append(v4e_final_margin_est)
            elif col_name == 'KKKompany':
                estimated_margins_comp.append(15.0)
            elif col_name == 'team_5':
                estimated_margins_comp.append(25.0)
            elif col_name == 'team_8':
                estimated_margins_comp.append(-10.0)
            else:
                estimated_margins_comp.append(0.0)

        companies_labels_comp = [
            f"{self.plot_teams_map['team_5'][0]}\n(Zysk Premium)",
            f"{self.plot_teams_map['KKKompany'][0]}\n(Lider Wolumenu)",
            f"{self.plot_teams_map['V4EScooters_share'][0]}\n(Zr√≥wnowa≈ºone podej≈õcie)",
            f"{self.plot_teams_map['team_8'][0]}\n(Wyj≈õcie z rynku)"
        ]
        colors_comp_plot = [self.plot_teams_map[col][1] for col in self.competitor_columns_plot]
        bubble_sizes_comp = [1200, 1000, 800, 400]

        ax4_2_comp.scatter(final_shares_comp, estimated_margins_comp, s=bubble_sizes_comp,
                           c=colors_comp_plot, alpha=0.75, edgecolors='black', linewidth=1.5)

        # Annotate each point with company label
        for i, company_label in enumerate(companies_labels_comp):
            ax4_2_comp.annotate(company_label, (final_shares_comp[i], estimated_margins_comp[i]),
                                textcoords="offset points", xytext=(0, 10 if bubble_sizes_comp[i] > 500 else -15),
                                ha='center', fontsize=9, fontweight='bold')

        ax4_2_comp.set_xlabel('Ko≈Ñcowy udzia≈Ç w rynku (%)')
        ax4_2_comp.set_ylabel('Szacowana mar≈ºa zysku (%)')
        ax4_2_comp.set_title('Ko≈Ñcowe pozycjonowanie strategiczne - Jako≈õƒá vs. Ilo≈õƒá')
        ax4_2_comp.grid(True, alpha=0.3, linestyle='--')
        ax4_2_comp.axhline(y=0, color='black', linestyle='-', alpha=0.5)

        # 4.3 Evolution of V4EScooters Market Rank (Bar Chart)
        ax4_3_comp = axes_comp[1, 0]
        ranks_v4e_plot = self.market_data['V4EScooters_rank']
        rounds_rank_plot = self.market_data['Round']
        max_rank = len(self.competitor_columns_plot)

        # Invert rank for plotting (1st place at the top)
        plot_rank_values = max_rank - ranks_v4e_plot + 1
        colors_rank_plot = ['gold' if r == 1 else 'silver' if r == 2 else '#CD7F32' if r == 3 else 'lightgray' for r in ranks_v4e_plot]

        bars_rank_plot = ax4_3_comp.bar(rounds_rank_plot, plot_rank_values,
                                        color=colors_rank_plot, alpha=0.85, edgecolor='black')

        ax4_3_comp.set_xlabel('Runda')
        ax4_3_comp.set_ylabel('Pozycja na rynku')
        ax4_3_comp.set_title(f"Ewolucja rankingu rynkowego {self.plot_teams_map['V4EScooters_share'][0]}")
        ax4_3_comp.set_yticks(np.arange(1, max_rank + 1))
        ax4_3_comp.set_yticklabels([f'{max_rank - i + 1}. miejsce' for i in np.arange(1, max_rank + 1)])
        ax4_3_comp.grid(True, axis='y', alpha=0.3)
        ax4_3_comp.set_xticks(range(1, 11))

        # Add rank labels on top of bars
        for bar_rank_item, rank_val_item in zip(bars_rank_plot, ranks_v4e_plot):
            ax4_3_comp.text(bar_rank_item.get_x() + bar_rank_item.get_width() / 2,
                            bar_rank_item.get_height() - 0.2,
                            f'{int(rank_val_item)}. miejsce',
                            ha='center', va='top', fontweight='bold', color='black', fontsize=10)

        # 4.4 Market Dynamics: Volatility and Leader's Share (Dual Axis Line Plot)
        ax4_4_comp = axes_comp[1, 1]
        market_volatility_plot = self.market_data[self.competitor_columns_plot].std(axis=1)
        leader_share_plot = self.market_data['market_leader_share']

        ax4_4_comp_twin = ax4_4_comp.twinx()
        line1 = ax4_4_comp.plot(rounds_comp_plot_data, market_volatility_plot, 'o-',
                                color=self.COLORS['primary'], linewidth=3, markersize=8,
                                label='Zmienno≈õƒá udzia≈Çu w rynku (odch. std.)')
        line2 = ax4_4_comp_twin.plot(rounds_comp_plot_data, leader_share_plot, 's--',
                                     color=self.COLORS['danger'], linewidth=3, markersize=8,
                                     label="Udzia≈Ç lidera w rynku (%)")

        ax4_4_comp.set_xlabel('Runda')
        ax4_4_comp.set_ylabel('Zmienno≈õƒá udzia≈Çu w rynku (odch. std.)', color=self.COLORS['primary'])
        ax4_4_comp_twin.set_ylabel("Udzia≈Ç lidera w rynku (%)", color=self.COLORS['danger'])
        ax4_4_comp.set_title('Dynamika rynku - Zmienno≈õƒá i koncentracja')
        ax4_4_comp.grid(True, alpha=0.3, axis='y', linestyle='dotted')
        ax4_4_comp_twin.grid(True, alpha=0.3, axis='y', linestyle=':')
        ax4_4_comp.set_xticks(range(1, 11))

        lines_dyn_plot_combined = line1 + line2
        labels_dyn_plot_combined = [l.get_label() for l in lines_dyn_plot_combined]
        ax4_4_comp.legend(lines_dyn_plot_combined, labels_dyn_plot_combined,
                          loc='upper center', frameon=True, ncol=1, bbox_to_anchor=(0.5, -0.15))

        fig_comp.suptitle('Dynamika konkurencyjna i pozycjonowanie rynkowe V4EScooters',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_comp.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_product_strategy(self):
        """Generates plots for product strategy and market segmentation."""
        print("\n\nüì¶ SEKCJA 5: STRATEGIA PRODUKTOWA - CELOWANIE W DOSKONA≈ÅO≈öƒÜ")
        print("=" * 100)

        # Print a summary table of product mix and pricing evolution
        print("\nV4EScooters MIKSU PRODUKTOWEGO I EWOLUCJI CEN:")
        print("-" * 120)
        print(f"{'Runda':<7} {'Fashionistas Sprzedane%':<12} {'Geeks Sprzedane%':<13} {'Savers Sprzedane%':<14} {'≈örednia cena (‚Ç¨)':<15} {'Faza strategii produktowej':<35}")
        print("-" * 120)
        for _, row in self.product_data_v4e.iterrows():
            total_sold = row['Total_Sold_V4E']
            fash_pct_sold = (row['Fashionistas_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0
            geeks_pct_sold = (row['Geeks_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0
            savers_pct_sold = (row['Savers_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0

            prod_strat_phase = ""
            if row['Round'] <= 4:
                prod_strat_phase = "PoczƒÖtkowy miks i wej≈õcie na rynek"
            elif row['Round'] <= 6:
                prod_strat_phase = "Dostosowanie i testowanie kryzysowe"
            else:
                prod_strat_phase = "Fokus na portfolio premium"

            print(f"{row['Round']:<7} {fash_pct_sold:>10.1f} {geeks_pct_sold:>11.1f} {savers_pct_sold:>12.1f} "
                  f"‚Ç¨{row['Avg_Selling_Price_V4E']:>13.2f} {prod_strat_phase:<35}")

        # Figure for Product Strategy (4 plots)
        fig_prod, axes_prod = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 5.1 Evolution of V4EScooters Product Mix (Stacked Bar Chart)
        ax5_1_prod = axes_prod[0, 0]
        rounds_prod_plot_v4e = self.product_data_v4e['Round']
        width_prod_plot = 0.8

        # Calculate percentage of each product sold
        fash_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0,
                                 self.product_data_v4e['Fashionistas_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)
        geeks_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0,
                                  self.product_data_v4e['Geeks_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)
        savers_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0,
                                   self.product_data_v4e['Savers_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)

        # Stack bars
        ax5_1_prod.bar(rounds_prod_plot_v4e, fash_pct_plot, width_prod_plot,
                       label='Fashionistas (% sprzedanych)', color=self.COLORS['purple'])
        ax5_1_prod.bar(rounds_prod_plot_v4e, geeks_pct_plot, width_prod_plot, bottom=fash_pct_plot,
                       label='Geeks/Ekolodzy (% sprzedanych)', color=self.COLORS['success'])
        ax5_1_prod.bar(rounds_prod_plot_v4e, savers_pct_plot, width_prod_plot, bottom=fash_pct_plot + geeks_pct_plot,
                       label='Oszczƒôdni (% sprzedanych)', color=self.COLORS['warning'])

        ax5_1_prod.set_xlabel('Runda')
        ax5_1_prod.set_ylabel('Miks produktowy (% sprzedanych jednostek V4E)')
        ax5_1_prod.set_title('Ewolucja portfolio produkt√≥w V4EScooters (Miks sprzeda≈ºy)')
        ax5_1_prod.legend(frameon=True, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.12))
        ax5_1_prod.grid(True, axis='y', alpha=0.3)
        ax5_1_prod.set_xticks(range(1, 11))
        ax5_1_prod.set_ylim(0, 100)

        # 5.2 Evolution of V4EScooters Pricing Strategy (Line Plot)
        ax5_2_prod = axes_prod[0, 1]
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Fashionistas_Price_V4E'], 'o-',
                        linewidth=3, label='Cena Fashionistas', color=self.COLORS['purple'], markersize=8)
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Geeks_Price_V4E'], 's-',
                        linewidth=3, label='Cena Geeks/Ekolodzy', color=self.COLORS['success'], markersize=8)
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Savers_Price_V4E'], '^-',
                        linewidth=3, label='Cena Oszczƒôdni', color=self.COLORS['warning'], markersize=8)

        ax5_2_prod.set_xlabel('Runda')
        ax5_2_prod.set_ylabel('Cena za jednostkƒô (‚Ç¨)')
        ax5_2_prod.set_title('Strategia cenowa V4EScooters wed≈Çug segmentu')
        ax5_2_prod.legend(frameon=True, loc='upper left')
        ax5_2_prod.grid(True, alpha=0.3)
        ax5_2_prod.set_xticks(range(1, 11))

        # Annotate price change after crisis
        ax5_2_prod.axvline(x=6.5, color='red', linestyle='--', alpha=0.7, linewidth=2)
        ax5_2_prod.text(6.3, self.product_data_v4e['Savers_Price_V4E'].min() + 20,
                        'Zmiana cen\nPremium\n(Po kryzysie)', ha='right', fontsize=10, fontweight='bold',
                        bbox=dict(boxstyle="round,pad=0.5", fc="yellow", alpha=0.7))

        # 5.3 V4EScooters Product Performance Radar (R10)
        ax5_3_prod = fig_prod.add_subplot(2, 2, 3, projection='polar')
        categories_radar_prod = ['Wolumen sprzeda≈ºy (% ca≈Çkowitego V4E)', 'Poziom ceny (‚Ç¨)',
                                 'Wska≈∫nik sukcesu sprzeda≈ºy (%)', 'Szacowany wk≈Çad w zysk (%)',
                                 'Wynik dopasowania strategicznego (1-5)']
        num_vars_radar = len(categories_radar_prod)
        angles_radar = np.linspace(0, 2 * np.pi, num_vars_radar, endpoint=False).tolist()
        angles_radar += angles_radar[:1]

        # Data for Round 10 for each product
        r10_data_prod = self.product_data_v4e[self.product_data_v4e['Round'] == 10].iloc[0]
        total_sold_r10_v4e = r10_data_prod['Total_Sold_V4E']

        # Fashionistas scores
        fash_vol_r10 = (r10_data_prod['Fashionistas_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        fash_price_r10 = r10_data_prod['Fashionistas_Price_V4E']
        fash_success_r10 = (r10_data_prod['Fashionistas_Sold_V4E'] / r10_data_prod['Fashionistas_Produced_V4E'] * 100) if r10_data_prod['Fashionistas_Produced_V4E'] > 0 else 0
        fash_profit_contr_est_r10 = 40
        fash_strat_fit_r10 = 5
        fash_scores_r10 = [fash_vol_r10, fash_price_r10, fash_success_r10, fash_profit_contr_est_r10, fash_strat_fit_r10 * 20]
        fash_scores_r10 += fash_scores_r10[:1]

        # Geeks scores
        geeks_vol_r10 = (r10_data_prod['Geeks_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        geeks_price_r10 = r10_data_prod['Geeks_Price_V4E']
        geeks_success_r10 = (r10_data_prod['Geeks_Sold_V4E'] / r10_data_prod['Geeks_Produced_V4E'] * 100) if r10_data_prod['Geeks_Produced_V4E'] > 0 else 0
        geeks_profit_contr_est_r10 = 45
        geeks_strat_fit_r10 = 5
        geeks_scores_r10 = [geeks_vol_r10, geeks_price_r10, geeks_success_r10, geeks_profit_contr_est_r10, geeks_strat_fit_r10 * 20]
        geeks_scores_r10 += geeks_scores_r10[:1]

        # Savers scores
        savers_vol_r10 = (r10_data_prod['Savers_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        savers_price_r10 = r10_data_prod['Savers_Price_V4E']
        savers_success_r10 = (r10_data_prod['Savers_Sold_V4E'] / r10_data_prod['Savers_Produced_V4E'] * 100) if r10_data_prod['Savers_Produced_V4E'] > 0 else 0
        savers_profit_contr_est_r10 = 15
        savers_strat_fit_r10 = 3
        savers_scores_r10 = [savers_vol_r10, savers_price_r10, savers_success_r10, savers_profit_contr_est_r10, savers_strat_fit_r10 * 20]
        savers_scores_r10 += savers_scores_r10[:1]

        # Plot radar lines and fills
        ax5_3_prod.plot(angles_radar, fash_scores_r10, 'o-', linewidth=2, label='Fashionistas (R10)', color=self.COLORS['purple'], markersize=6)
        ax5_3_prod.fill(angles_radar, fash_scores_r10, self.COLORS['purple'], alpha=0.25)
        ax5_3_prod.plot(angles_radar, geeks_scores_r10, 's-', linewidth=2, label='Geeks (R10)', color=self.COLORS['success'], markersize=6)
        ax5_3_prod.fill(angles_radar, geeks_scores_r10, self.COLORS['success'], alpha=0.25)
        ax5_3_prod.plot(angles_radar, savers_scores_r10, '^-', linewidth=2, label='Oszczƒôdni (R10)', color=self.COLORS['warning'], markersize=6)
        ax5_3_prod.fill(angles_radar, savers_scores_r10, self.COLORS['warning'], alpha=0.25)

        ax5_3_prod.set_xticks(angles_radar[:-1])
        ax5_3_prod.set_xticklabels(categories_radar_prod, fontsize=9)
        ax5_3_prod.set_yticks(np.arange(0, 301, 50))
        ax5_3_prod.set_title('Profil wydajno≈õci produktu V4E (R10 - Skale ilustracyjne)', pad=20)
        ax5_3_prod.legend(loc='upper right', bbox_to_anchor=(1.4, 1.1), fontsize=10)

        # 5.4 V4EScooters Product Revenue Contribution (Grouped Bar Chart)
        ax5_4_prod = axes_prod[1, 1]
        revenue_rounds_prod_plot = list(range(1, 11))  # Use all rounds
        revenue_data_list_prod = []
        for r_rev_prod in revenue_rounds_prod_plot:
            row_rev_prod = self.product_data_v4e[self.product_data_v4e['Round'] == r_rev_prod].iloc[0]
            fash_rev_prod = row_rev_prod['Fashionistas_Sold_V4E'] * row_rev_prod['Fashionistas_Price_V4E'] / 1000
            geeks_rev_prod = row_rev_prod['Geeks_Sold_V4E'] * row_rev_prod['Geeks_Price_V4E'] / 1000
            savers_rev_prod = row_rev_prod['Savers_Sold_V4E'] * row_rev_prod['Savers_Price_V4E'] / 1000
            revenue_data_list_prod.append([fash_rev_prod, geeks_rev_prod, savers_rev_prod])
        revenue_data_np_prod = np.array(revenue_data_list_prod).T

        x_rev_prod_plot = np.arange(len(revenue_rounds_prod_plot))
        width_rev_prod_plot = 0.25
        product_names_rev_plot = ['Przych√≥d Fashionistas', 'Przych√≥d Geeks/Eko', 'Przych√≥d Oszczƒôdni']
        colors_rev_prod_plot = [self.COLORS['purple'], self.COLORS['success'], self.COLORS['warning']]

        for i, (product_rev_item, color_rev_item) in enumerate(zip(product_names_rev_plot, colors_rev_prod_plot)):
            ax5_4_prod.bar(x_rev_prod_plot + i * width_rev_prod_plot - width_rev_prod_plot,
                           revenue_data_np_prod[i], width_rev_prod_plot,
                           label=product_rev_item, color=color_rev_item, alpha=0.85)

        ax5_4_prod.set_xlabel('Runda')
        ax5_4_prod.set_ylabel('Przych√≥d (tys. ‚Ç¨)')
        ax5_4_prod.set_title('Wk≈Çad przychod√≥w produkt√≥w V4EScooters')
        ax5_4_prod.set_xticks(x_rev_prod_plot)
        ax5_4_prod.set_xticklabels(revenue_rounds_prod_plot)
        ax5_4_prod.legend(frameon=True, loc='upper left')
        ax5_4_prod.grid(True, axis='y', alpha=0.3)

        fig_prod.suptitle('Analiza strategii produktowej V4EScooters - Wydajno≈õƒá portfolio premium',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_prod.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_geographical_strategy(self):
        """Generates plots for geographical market strategy."""
        print("\n\nüåç SEKCJA 6: EKSPANSJA GEOGRAFICZNA - Opanowanie rynku")
        print("=" * 100)

        # Print a summary table of geographical performance
        print("\nWYNIKI V4EScooters NA RYNKACH GEOGRAFICZNYCH (Fokus na fazƒô premium R7-R10):")
        print("-" * 130)
        print(f"{'Miasto':<12} {'≈örednia skuteczno≈õƒá V4E (%)':<22} {'Cena referencyjna rynku (‚Ç¨)':<22} {'Rozmiar rynku':<15} {'Konkurencja':<15} {'Fokus strategii V4E':<25}")
        print("-" * 130)
        for _, row in self.df_geo_v4e.iterrows():
            v4e_strat_focus = "Celuj w wysoki potencja≈Ç" if row['Avg_Success_V4E_R7_R10'] > 85 and row['Competition_Level'] != 'Wysoka' else \
                              "Gra na niszy premium" if row['Avg_Success_V4E_R7_R10'] > 75 and row['Competition_Level'] == 'Wysoka' else \
                              "Rynek rozwojowy"
            print(f"{row['City']:<12} {row['Avg_Success_V4E_R7_R10']:>20.1f} {row['Ref_Avg_Price_Market']:>20} "
                  f"{row['Market_Size_Category']:<15} {row['Competition_Level']:<15} {v4e_strat_focus:<25}")

        # Figure for Geographical Strategy (4 plots)
        fig_geo, axes_geo = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.45, wspace=0.3)

        # 6.1 V4EScooters Market Success by City - All Rounds on X-axis (Grouped Bar Chart)
        ax6_1_geo = axes_geo[0, 0]
        cities_geo_plot = self.df_geo_v4e['City']
        x_geo_indices = np.arange(len(cities_geo_plot))
        width_geo_bar = 0.20
        rounds_geo_success_cols = ['R7_Success_V4E', 'R8_Success_V4E', 'R9_Success_V4E', 'R10_Success_V4E']
        round_labels_geo = ['Sukces R7', 'Sukces R8', 'Sukces R9', 'Sukces R10']
        colors_geo_rounds = [self.COLORS['primary'], self.COLORS['success'], self.COLORS['warning'], self.COLORS['purple']]

        for i, (col_name, label) in enumerate(zip(rounds_geo_success_cols, round_labels_geo)):
            ax6_1_geo.bar(x_geo_indices + i * width_geo_bar - 1.5 * width_geo_bar,
                          self.df_geo_v4e[col_name], width_geo_bar,
                          label=label, color=colors_geo_rounds[i], alpha=0.85)

        ax6_1_geo.set_xlabel('Miasto')
        ax6_1_geo.set_ylabel('Wska≈∫nik sukcesu sprzeda≈ºy V4EScooters (%)')
        ax6_1_geo.set_title('Wyniki geograficzne V4EScooters wed≈Çug rundy (R7-R10)')
        ax6_1_geo.set_xticks(x_geo_indices)
        ax6_1_geo.set_xticklabels(cities_geo_plot, rotation=45, ha="right")
        ax6_1_geo.legend(frameon=True, loc='upper center', ncol=2, bbox_to_anchor=(0.5, -0.25))
        ax6_1_geo.grid(True, axis='y', alpha=0.3)
        ax6_1_geo.axhline(y=100, color='red', linestyle='--', alpha=0.5, label='Cel 100%')
        ax6_1_geo.set_ylim(0, 110)

        # 6.2 V4EScooters City Performance Heatmap
        ax6_2_geo = axes_geo[0, 1]
        performance_matrix_geo = self.df_geo_v4e[rounds_geo_success_cols].values.T
        im_geo_heatmap = ax6_2_geo.imshow(performance_matrix_geo, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)

        ax6_2_geo.set_xticks(np.arange(len(cities_geo_plot)))
        ax6_2_geo.set_yticks(np.arange(len(round_labels_geo)))
        ax6_2_geo.set_xticklabels(cities_geo_plot, rotation=45, ha="right")
        ax6_2_geo.set_yticklabels(round_labels_geo)
        ax6_2_geo.set_title('Mapa ciep≈Ça wydajno≈õci miasta V4EScooters (R7-R10 Sukces %)')

        # Add text annotations to heatmap cells
        for i_heat_geo in range(performance_matrix_geo.shape[0]):
            for j_heat_geo in range(performance_matrix_geo.shape[1]):
                text_val_geo = performance_matrix_geo[i_heat_geo, j_heat_geo]
                ax6_2_geo.text(j_heat_geo, i_heat_geo, f'{text_val_geo:.0f}%', ha="center", va="center",
                               color="black" if 30 < text_val_geo < 70 else "white", fontsize=9, fontweight='bold')

        cbar_geo_heatmap = plt.colorbar(im_geo_heatmap, ax=ax6_2_geo, orientation='vertical', fraction=0.046, pad=0.04)
        cbar_geo_heatmap.set_label('Wska≈∫nik sukcesu sprzeda≈ºy V4E (%)')

        # 6.3 Market Attractiveness vs. V4EScooters Performance Map (Scatter Plot with Size and Color)
        ax6_3_geo = axes_geo[1, 0]
        # Map categorical market size and competition levels to numerical values
        market_size_numeric_map = {'Ma≈Çy': 1, '≈öredni': 2, 'Du≈ºy': 3}
        self.df_geo_v4e['Market_Size_Num'] = self.df_geo_v4e['Market_Size_Category'].map(market_size_numeric_map)
        competition_numeric_map = {'Wysoka': 1, '≈örednia': 2, 'Niska': 3}
        self.df_geo_v4e['Competition_Attractive_Num'] = self.df_geo_v4e['Competition_Level'].map(competition_numeric_map)

        scatter_market_map = ax6_3_geo.scatter(
            self.df_geo_v4e['Market_Size_Num'],
            self.df_geo_v4e['Competition_Attractive_Num'],
            s=self.df_geo_v4e['Avg_Success_V4E_R7_R10'] * 8,
            c=self.df_geo_v4e['Ref_Avg_Price_Market'],
            cmap='coolwarm',
            alpha=0.75, edgecolors='black', linewidth=1.5
        )

        # Annotate cities
        for i, city_label in enumerate(self.df_geo_v4e['City']):
            ax6_3_geo.annotate(city_label, (self.df_geo_v4e['Market_Size_Num'].iloc[i], self.df_geo_v4e['Competition_Attractive_Num'].iloc[i]),
                               textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9)

        ax6_3_geo.set_xlabel('Kategoria rozmiaru rynku (Wiƒôkszy ‚Üí)')
        ax6_3_geo.set_ylabel('Atrakcyjno≈õƒá rynku (Ni≈ºsza konkurencja ‚Üí)')
        ax6_3_geo.set_title('Pozycjonowanie rynkowe V4E: Atrakcyjno≈õƒá vs. Wydajno≈õƒá')
        ax6_3_geo.set_xticks([1, 2, 3])
        ax6_3_geo.set_xticklabels(['Ma≈Çy', '≈öredni', 'Du≈ºy'])
        ax6_3_geo.set_yticks([1, 2, 3])
        ax6_3_geo.set_yticklabels(['Wysoka konk.', '≈örednia konk.', 'Niska konk.'])
        ax6_3_geo.grid(True, alpha=0.3, linestyle='--')

        cbar_market_map_price = plt.colorbar(scatter_market_map, ax=ax6_3_geo)
        cbar_market_map_price.set_label('Referencyjna ≈õrednia cena rynkowa (‚Ç¨)')

        # Add dummy scatter points for legend for bubble size
        for success_level in [70, 85, 95]:
            ax6_3_geo.scatter([], [], s=success_level * 8, label=f'{success_level}% Sukcesu',
                              color='gray', alpha=0.7)
        ax6_3_geo.legend(scatterpoints=1, frameon=True, title='Rozmiar bƒÖbelka (≈örednia skuteczno≈õƒá V4E R7-10)', loc='lower right')

        # 6.4 Strategic Market Groups for V4EScooters (Bar Chart)
        ax6_4_geo = axes_geo[1, 1]
        # Define strategic groups based on performance and attractiveness
        self.df_geo_v4e['Strategic_Group'] = 'RozwijajƒÖcy siƒô'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 85) & (self.df_geo_v4e['Competition_Attractive_Num'] >= 2), 'Strategic_Group'] = 'Gwiazdy wydajno≈õci'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 75) & (self.df_geo_v4e['Avg_Success_V4E_R7_R10'] < 85) & (self.df_geo_v4e['Competition_Attractive_Num'] >= 2), 'Strategic_Group'] = 'Solidny wzrost'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 70) & (self.df_geo_v4e['Competition_Attractive_Num'] == 1), 'Strategic_Group'] = 'Trudna nisza'

        group_counts = self.df_geo_v4e['Strategic_Group'].value_counts()
        group_colors_map = {
            'Gwiazdy wydajno≈õci': self.COLORS['success'],
            'Solidny wzrost': self.COLORS['info'],
            'Trudna nisza': self.COLORS['warning'],
            'RozwijajƒÖcy siƒô': self.COLORS['danger']
        }
        bars_strat_groups = ax6_4_geo.bar(group_counts.index, group_counts.values,
                                          color=[group_colors_map.get(g, 'gray') for g in group_counts.index], alpha=0.85)

        ax6_4_geo.set_xlabel('Strategiczna grupa rynkowa dla V4EScooters')
        ax6_4_geo.set_ylabel('Liczba miast w grupie')
        ax6_4_geo.set_title('Strategiczna segmentacja rynkowa V4EScooters (fokus R7-R10)')
        ax6_4_geo.tick_params(axis='x', rotation=15)
        ax6_4_geo.grid(True, axis='y', alpha=0.3)

        # Add count labels on top of bars
        for bar_group in bars_strat_groups:
            yval = bar_group.get_height()
            ax6_4_geo.text(bar_group.get_x() + bar_group.get_width() / 2, yval + 0.05,
                           int(yval), ha='center', va='bottom', fontsize=10)

        fig_geo.suptitle('Strategia rynkowa geograficzna V4EScooters - OsiƒÖgniƒôcie regionalnego mistrzostwa',
                         fontsize=20, fontweight='bold', y=0.98)
        fig_geo.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _print_strategic_insights_framework(self):
        """
        Enhanced strategic insights section - text-rich analysis with actionable frameworks.
        This replaces the plot-heavy strategic lessons section with deep business insights.
        """
        print("\n\nüéì SEKCJA 7: STRATEGICZNE WNIOSKI I RAMY BIZNESOWE - KSIƒòGA MƒÑDRO≈öCI V4EScooters")
        print("=" * 100)

        # Extract key metrics for insights
        final_balance = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi = ((final_balance - self.initial_investment) / self.initial_investment) * 100
        final_market_share = self.market_data['V4EScooters_share'].iloc[-1]
        peak_efficiency = max(self.operations_data['Fashionistas_Eff'].iloc[-1], self.operations_data['Geeks_Eff'].iloc[-1])
        crisis_recovery_rounds = 2

        print(f"\nüîç CZƒò≈öƒÜ I: ANATOMIA STRATEGICZNEGO SUKCESU - DEKONSTRUKCJA ZWYCIƒòSTWA")
        print("=" * 100)
        
        print(f"\n1Ô∏è‚É£ FUNDAMENTALNE ODKRYCIA STRATEGICZNE:")
        print("-" * 80)
        print(f"   üéØ PARADOKS MARKETINGOWY: Najwy≈ºszy sukces sprzeda≈ºy (92.7%) osiƒÖgniƒôto przy bud≈ºecie ‚Ç¨0")
        print(f"      ‚Üí Implikacja: Operacyjna doskona≈Ço≈õƒá mo≈ºe zastƒÖpiƒá wydatki marketingowe")
        print(f"      ‚Üí Transferowalno≈õƒá: W biznesie premium, jako≈õƒá produktu = najlepsza reklama")
        
        print(f"\n   üè≠ PRZEWAGA OPERACYJNA: OsiƒÖgniƒôto teoretyczne maximum efektywno≈õci (200%)")
        print(f"      ‚Üí Implikacja: Inwestycje w procesy dajƒÖ d≈ÇugoterminowƒÖ przewagƒô konkurencyjnƒÖ")
        print(f"      ‚Üí Transferowalno≈õƒá: Doskona≈Ço≈õƒá operacyjna = niekopiowalna bariera wej≈õcia")
        
        print(f"\n   üí∞ MODEL PREMIUM-PROFIT: Mar≈ºa zysku > udzia≈Ç w rynku jako KPI sukcesu")
        print(f"      ‚Üí Implikacja: Rentowno≈õƒá jednostkowa wa≈ºniejsza ni≈º wolumen sprzeda≈ºy")
        print(f"      ‚Üí Transferowalno≈õƒá: W dojrza≈Çych rynkach, jako≈õƒá > ilo≈õƒá")
        
        print(f"\n   üîÑ ANTYKRUCHO≈öƒÜ: Odbudowa po kryzysie w {crisis_recovery_rounds} rundach")
        print(f"      ‚Üí Implikacja: Silne fundamenty umo≈ºliwiajƒÖ szybkƒÖ regeneracjƒô")
        print(f"      ‚Üí Transferowalno≈õƒá: Odporno≈õƒá > optymalizacja w niepewnych czasach")

        print(f"\n\n2Ô∏è‚É£ MATRYCA DECYZJI STRATEGICZNYCH - FRAMEWORK 'IMPACT √ó TIMING √ó EXECUTION':")
        print("-" * 80)
        
        decisions_analysis = {
            "Zmiana na ceny premium (R7)": {
                "impact": "KRYTYCZNY",
                "timing": "PERFEKCYJNY", 
                "execution": "DOSKONA≈ÅY",
                "outcome": "Fundamentalna zmiana rentowno≈õci",
                "lesson": "≈ömia≈Çe pivoty cenowe mogƒÖ redefinowaƒá pozycjonowanie"
            },
            "Test marketingowy ‚Ç¨0 (R8)": {
                "impact": "PRZE≈ÅOMOWY",
                "timing": "IDEALNY",
                "execution": "MISTRZOWSKI", 
                "outcome": "Odkrycie paradoksu efektywno≈õci",
                "lesson": "Przeciwintuitywne eksperymenty ujawniajƒÖ ukryte prawdy"
            },
            "Optymalizacja si≈Çy roboczej (R8-10)": {
                "impact": "WYSOKI",
                "timing": "STRATEGICZNY",
                "execution": "PRECYZYJNY",
                "outcome": "Maksymalizacja efektywno≈õci koszt√≥w",
                "lesson": "CiƒÖg≈Ça optymalizacja zasob√≥w = trwa≈Ça przewaga"
            },
            "Fokus na doskona≈Ço≈õƒá linii": {
                "impact": "FUNDAMENTALNY", 
                "timing": "KONSEKWENTNY",
                "execution": "SYSTEMATYCZNY",
                "outcome": "OsiƒÖgniƒôcie teoretycznego maximum",
                "lesson": "Systematyczna pogo≈Ñ za doskona≈Ço≈õciƒÖ przynosi prze≈Çomowe rezultaty"
            }
        }
        
        for decision, analysis in decisions_analysis.items():
            print(f"\n   üìã {decision.upper()}:")
            print(f"      Wp≈Çyw: {analysis['impact']} | Timing: {analysis['timing']} | Wykonanie: {analysis['execution']}")
            print(f"      Rezultat: {analysis['outcome']}")
            print(f"      Lekcja: {analysis['lesson']}")

        print(f"\n\n3Ô∏è‚É£ EKONOMICZNE PRAWA ODKRYTE W GRZE:")
        print("-" * 80)
        
        marketing_efficiency_r8 = self.marketing_full[self.marketing_full['Round'] == 8]['Sales_Success_Rate'].iloc[0]
        marketing_spend_r9 = self.marketing_full[self.marketing_full['Round'] == 9]['Marketing_Spend'].iloc[0]
        marketing_efficiency_r9 = self.marketing_full[self.marketing_full['Round'] == 9]['Sales_Success_Rate'].iloc[0]
        
        print(f"   üìà PRAWO MALEJƒÑCYCH ZWROT√ìW MARKETINGOWYCH:")
        print(f"      ‚Ç¨0 ‚Üí {marketing_efficiency_r8:.1f}% sukcesu vs ‚Ç¨{marketing_spend_r9:,} ‚Üí {marketing_efficiency_r9:.1f}% sukcesu")
        print(f"      Wniosek: Przekroczenie punktu optymalnego niszczy efektywno≈õƒá")
        
        avg_price_premium = self.product_data_v4e[self.product_data_v4e['Round'] >= 7]['Avg_Selling_Price_V4E'].mean()
        print(f"\n   üíé PRAWO PREMIUM PRICING:")
        print(f"      ≈örednia cena premium (R7-10): ‚Ç¨{avg_price_premium:.0f}")
        print(f"      Wniosek: Wysokie ceny + wysoka jako≈õƒá = akceptowalny value proposition")
        
        print(f"\n   ‚öôÔ∏è PRAWO DOSKONA≈ÅO≈öCI OPERACYJNEJ:")
        print(f"      Efektywno≈õƒá {peak_efficiency:.0f}% = teoretyczne maximum w grze")
        print(f"      Wniosek: Inwestycje w efektywno≈õƒá majƒÖ nieliniowe zwroty")

        print(f"\n\nüöÄ CZƒò≈öƒÜ II: TRANSFEROWALNY FRAMEWORK STRATEGICZNY - 'V4E EXCELLENCE MODEL'")
        print("=" * 100)
        
        print(f"\nüìã FRAMEWORK 'P.E.A.K.S.' - Piƒôƒá filar√≥w strategicznego sukcesu:")
        print("-" * 80)
        
        peaks_framework = {
            "P - PREMIUM POSITIONING": {
                "definition": "Pozycjonowanie oparte na warto≈õci, nie na cenie",
                "v4e_example": f"Ceny premium {avg_price_premium:.0f}‚Ç¨ vs standardowe ~200‚Ç¨",
                "kpi": "Mar≈ºa zysku > udzia≈Ç w rynku",
                "actionable": "Zdefiniuj unique value proposition i nie konkuruj cenƒÖ"
            },
            "E - EXCELLENCE OPERATIONAL": {
                "definition": "Systematyczna pogo≈Ñ za doskona≈Ço≈õciƒÖ proces√≥w",
                "v4e_example": f"OsiƒÖgniƒôcie {peak_efficiency:.0f}% efektywno≈õci (teoretyczne max)",
                "kpi": "Efektywno≈õƒá proces√≥w > 150% benchmarku bran≈ºowego",
                "actionable": "Inwestuj w procesy, nie tylko w produkty"
            },
            "A - AGILE EXPERIMENTATION": {
                "definition": "CiƒÖg≈Çe testowanie przeciwintuitywnych hipotez",
                "v4e_example": "Test ‚Ç¨0 marketingu ‚Üí 92.7% sukcesu",
                "kpi": "% bud≈ºetu na eksperymenty > 5%",
                "actionable": "Zaplanuj bud≈ºet na 'g≈Çupie' eksperymenty"
            },
            "K - KRISIS MANAGEMENT": {
                "definition": "Proaktywne przygotowanie na szoki systemowe",
                "v4e_example": f"Odbudowa po kryzysie w {crisis_recovery_rounds} rundach",
                "kpi": "Czas powrotu do poziomu sprzed kryzysu < 2 okres√≥w",
                "actionable": "Buduj rezerwy i scenariusze kryzysowe"
            },
            "S - SCALING SMART": {
                "definition": "Inteligentne skalowanie oparte na danych",
                "v4e_example": "Optymalizacja zespo≈Çu z 151 do 113 os√≥b przy wy≈ºszej produktywno≈õci",
                "kpi": "Produktywno≈õƒá na pracownika rosnƒÖca rocznie > 10%",
                "actionable": "Skaluj efektywno≈õƒá, nie tylko rozmiar"
            }
        }
        
        for pillar, details in peaks_framework.items():
            print(f"\nüéØ {pillar}:")
            print(f"   Definicja: {details['definition']}")
            print(f"   Przyk≈Çad V4E: {details['v4e_example']}")
            print(f"   KPI: {details['kpi']}")
            print(f"   Do dzia≈Çania: {details['actionable']}")

        print(f"\n\nüß† CZƒò≈öƒÜ III: STRATEGICZNE ANTI-PATTERNS - CZEGO UNIKAƒÜ")
        print("=" * 100)
        
        anti_patterns = {
            "MARKETING MONEY PIT": {
                "description": "Przekonanie, ≈ºe wiƒôcej wydatk√≥w = lepsze rezultaty",
                "v4e_evidence": f"‚Ç¨{marketing_spend_r9:,} wydane w R9 da≈Ço {marketing_efficiency_r9:.1f}% vs ‚Ç¨0 w R8 = {marketing_efficiency_r8:.1f}%",
                "warning": "Sprawd≈∫ punkt nasycenia przed zwiƒôkszaniem bud≈ºetu",
                "alternative": "Zainwestuj w jako≈õƒá produktu zamiast w reklamy"
            },
            "VOLUME VANITY METRICS": {
                "description": "Skupianie siƒô na udziale w rynku kosztem rentowno≈õci",
                "v4e_evidence": f"V4E: {final_market_share:.1f}% udzia≈Çu ale najwy≈ºsza prawdopodobna rentowno≈õƒá",
                "warning": "Udzia≈Ç w rynku bez zysku = droga do bankructwa",
                "alternative": "Priorytetyzuj zysk na jednostkƒô nad wolumen"
            },
            "EFFICIENCY PLATEAU": {
                "description": "Akceptowanie 'wystarczajƒÖco dobrej' efektywno≈õci",
                "v4e_evidence": f"Kontynuacja inwestycji w efektywno≈õƒá do {peak_efficiency:.0f}% (max mo≈ºliwy)",
                "warning": "Konkurencja nigdy nie ≈õpi - stagnacja = regres",
                "alternative": "Ustaw cele efektywno≈õci powy≈ºej obecnych limit√≥w"
            },
            "CRISIS PARALYSIS": {
                "description": "Zamra≈ºanie dzia≈Ça≈Ñ podczas turbulencji",
                "v4e_evidence": f"Aktywna odbudowa po R5, powr√≥t do wzrostu w {crisis_recovery_rounds} rundach",
                "warning": "Kryzysy = szanse dla przygotowanych",
                "alternative": "U≈ºywaj kryzys√≥w do strategicznych pivot-√≥w"
            }
        }
        
        for pattern, details in anti_patterns.items():
            print(f"\n‚ùå {pattern}:")
            print(f"   Problem: {details['description']}")
            print(f"   Dow√≥d z V4E: {details['v4e_evidence']}")
            print(f"   Ostrze≈ºenie: {details['warning']}")
            print(f"   Alternatywa: {details['alternative']}")

        print(f"\n\nüí° CZƒò≈öƒÜ IV: OPERACYJNY PLAYBOOK - KONKRETNE NARZƒòDZIA")
        print("=" * 100)
        
        print(f"\nüîß NARZƒòDZIA DECYZYJNE:")
        print("-" * 50)
        
        print(f"\n1. MARKETING EFFICIENCY CALCULATOR:")
        print(f"   Formula: Sukces% / (Bud≈ºet/1000‚Ç¨) = Efektywno≈õƒá")
        print(f"   Benchmark V4E: >50 = doskona≈Çy, 10-50 = dobry, <10 = przeanalizuj")
        print(f"   U≈ºycie: Przed ka≈ºdƒÖ kampaniƒÖ marketingowƒÖ")
        
        print(f"\n2. OPERATIONAL EXCELLENCE TRACKER:")
        print(f"   Formula: (Aktualna wydajno≈õƒá / Teoretyczne maximum) √ó 100")
        print(f"   Benchmark V4E: >80% = lider rynku, 60-80% = konkurencyjny, <60% = wymaga poprawy")
        print(f"   U≈ºycie: Miesiƒôczny przeglƒÖd proces√≥w")
        
        print(f"\n3. PREMIUM PRICING VIABILITY CHECK:")
        print(f"   Formula: (Cena premium / ≈örednia rynkowa - 1) √ó 100 = % premii")
        print(f"   Benchmark V4E: 20-40% premii przy zachowaniu popytu = sweet spot")
        print(f"   U≈ºycie: Przed zmianami cenowymi")
        
        print(f"\n4. CRISIS RECOVERY SPEED INDEX:")
        print(f"   Formula: Periody do powrotu na poziom sprzed kryzysu")
        print(f"   Benchmark V4E: <3 periody = excellent, 3-5 = good, >5 = wymaga strategii")
        print(f"   U≈ºycie: Planowanie scenariuszy kryzysowych")

        print(f"\n\nüìä METRYKI DO ≈öLEDZENIA W KA≈ªDYM BIZNESIE:")
        print("-" * 50)
        
        key_metrics = [
            ("Efektywno≈õƒá marketingowa", "Sukces sprzeda≈ºy / Wydatki marketingowe", "Miesiƒôcznie"),
            ("Efektywno≈õƒá operacyjna", "Produktywno≈õƒá / Teoretyczne maximum", "Tygodniowo"), 
            ("Mar≈ºa jednostkowa", "Zysk na jednostkƒô / Przych√≥d na jednostkƒô", "Na produkt"),
            ("Indeks premium", "Twoja cena / ≈örednia rynkowa", "Kwartalnie"),
            ("Czas odbudowy", "Periody powrotu po spadku", "Po ka≈ºdym kryzysie"),
            ("ROI adaptacji", "Zysk po zmianach / Koszt zmian", "Po ka≈ºdej transformacji")
        ]
        
        for metric, formula, frequency in key_metrics:
            print(f"   ‚Ä¢ {metric}: {formula} [{frequency}]")

        print(f"\n\nüéØ CZƒò≈öƒÜ V: IMPLEMENTACYJNY ACTION PLAN - OD WIEDZY DO DZIA≈ÅANIA")
        print("=" * 100)
        
        print(f"\nüìÖ 30-60-90 DNI PLAN WDRO≈ªENIA 'V4E EXCELLENCE MODEL':")
        print("-" * 60)
        
        implementation_plan = {
            "PIERWSZE 30 DNI - FOUNDATION": [
                "Przeprowad≈∫ audit efektywno≈õci operacyjnej (benchmark vs maksimum teoretyczne)",
                "Wdro≈º system monitorowania kluczowych metryk V4E",
                "Zidentyfikuj 3 procesy o najwy≈ºszym potencjale optymalizacji",
                "Zaplanuj bud≈ºet na eksperymenty (min. 5% bud≈ºetu operacyjnego)",
                "Stw√≥rz scenariusze kryzysowe i plany kontyngencji"
            ],
            "NASTƒòPNE 30 DNI - EXPERIMENTATION": [
                "Przeprowad≈∫ test przeciwintuitywnej hipotezy (inspirowany testem ‚Ç¨0)",
                "Zaimplementuj pierwszy cykl optymalizacji efektywno≈õci",
                "Przeanalizuj pricing strategy - potencja≈Ç premium positioning",
                "Ustan√≥w baseline dla wszystkich kluczowych metryk",
                "Rozpocznij szkolenia zespo≈Çu w excellence mindset"
            ],
            "OSTATNIE 30 DNI - OPTIMIZATION": [
                "Oce≈Ñ wyniki pierwszych eksperyment√≥w i skaluj sukcesy",
                "Dostosuj cennik na podstawie value-based pricing analysis",
                "Zoptymalizuj alokacjƒô zasob√≥w na podstawie danych",
                "Przeprowad≈∫ symulacjƒô kryzysu i test plan√≥w kontyngencji",
                "Ustaw cele d≈Çugoterminowe dla doskona≈Ço≈õci operacyjnej"
            ]
        }
        
        for phase, tasks in implementation_plan.items():
            print(f"\nüóìÔ∏è {phase}:")
            for i, task in enumerate(tasks, 1):
                print(f"   {i}. {task}")

        print(f"\n\nüèÜ CZƒò≈öƒÜ VI: LEGACY FRAMEWORK - D≈ÅUGOTERMINOWA STRATEGIA")
        print("=" * 100)
        
        print(f"\nüîÆ STRATEGICZNE HORYZONTY CZASOWE:")
        print("-" * 50)
        
        time_horizons = {
            "HORIZON 1 (0-12 miesiƒôcy) - OPERACYJNA DOSKONA≈ÅO≈öƒÜ": [
                "OsiƒÖgnij 90%+ teoretycznego maximum efektywno≈õci w kluczowych procesach",
                "Zbuduj system ciƒÖg≈Çego eksperymentowania (min. 1 test miesiƒôcznie)",
                "Ustan√≥w premium positioning w g≈Ç√≥wnych liniach produktowych",
                "Stw√≥rz dokumentacjƒô wszystkich best practices i lessons learned"
            ],
            "HORIZON 2 (1-3 lata) - STRATEGICZNA PRZEWAGA": [
                "Rozwi≈Ñ niekopiowalne competitive moats oparte na doskona≈Ço≈õci",
                "Zbuduj kulturƒô organizacyjnƒÖ zorientowanƒÖ na excellence",
                "Sta≈Ñ siƒô benchmarkiem efektywno≈õci w swojej bran≈ºy",
                "Stw√≥rz systemy predykcyjne dla zarzƒÖdzania kryzysami"
            ],
            "HORIZON 3 (3-5 lat) - MARKET LEADERSHIP": [
                "Definiuj standardy bran≈ºowe w zakresie efektywno≈õci i jako≈õci", 
                "Monetyzuj swojƒÖ wiedzƒô poprzez consulting/licensing",
                "Sta≈Ñ siƒô magnesem dla najlepszych talent√≥w w bran≈ºy",
                "Buduj ekosystem partner√≥w oparty na shared excellence values"
            ]
        }
        
        for horizon, goals in time_horizons.items():
            print(f"\nüìà {horizon}:")
            for goal in goals:
                print(f"   ‚Ä¢ {goal}")

        print(f"\n\nüéì FINALNE PRZEMY≈öLENIA - MƒÑDRO≈öƒÜ Z PODR√ì≈ªY V4E:")
        print("=" * 100)
        
        final_wisdom = [
            "DOSKONA≈ÅO≈öƒÜ NIE JEST MIEJSCEM DOCELOWYM, LECZ SPOSOBEM PODR√ì≈ªY",
            "KA≈ªDY KRYZYS NOSI W SOBIE NASIONA NASTƒòPNEGO PRZE≈ÅOMU",
            "PRAWDZIWA PRZEWAGA KONKURENCYJNA LE≈ªY W SYSTEMACH, NIE W PRODUKTACH", 
            "NAJLEPSZƒÑ REKLAMƒÑ JEST PRODUKT, KT√ìRY PRZEMAWIA SAM ZA SIEBIE",
            "W ≈öWIECIE PE≈ÅNYM HA≈ÅASU, CICHA DOSKONA≈ÅO≈öƒÜ KRZYCZY NAJG≈ÅO≈öNIEJ"
        ]
        
        for wisdom in final_wisdom:
            print(f"\nüíé {wisdom}")
        
        print(f"\n\n" + "=" * 100)
        print(f"üèÖ KONKLUZJA: V4EScooters nie wygra≈Ço przez przypadek.")
        print(f"Zwyciƒôstwo by≈Ço rezultatem systematycznego wdra≈ºania")
        print(f"uniwersalnych zasad biznesowej doskona≈Ço≈õci.")
        print(f"Te same zasady mogƒÖ transformowaƒá ka≈ºdy biznes.")
        print(f"Pytanie brzmi: Czy jeste≈õ gotowy na tƒô podr√≥≈º?")
        print("=" * 100)

    def _print_final_conclusions(self):
        """
        Prints the final conclusions section, including key performance indicators (KPIs),
        a strategic framework, transferable business lessons, and a section for notes.
        """
        print("\n\nüèÅ SEKCJA 8: KO≈ÉCOWE WNIOSKI - KLUCZOWE WSKA≈πNIKI WYDAJNO≈öCI (KPI) I STRATEGICZNE PODSUMOWANIE")
        print("=" * 100)

        # 8.1 NUMERICAL REPRESENTATION - Key metrics from data structures
        print("\nüìä CZƒò≈öƒÜ I: METRYKI BIZNESOWE I KPI Z ANALIZY DANYCH")
        print("-" * 100)

        # Extracting KPIs from data
        final_balance = self.df_finance['Ending_Balance'].iloc[-1]
        initial_investment = self.initial_investment
        total_roi = ((final_balance - initial_investment) / initial_investment) * 100
        peak_balance = self.df_finance['Ending_Balance'].max()

        final_units_produced = self.df_finance['Units_Produced'].iloc[-1]
        final_units_sold = self.df_finance['Units_Sold'].iloc[-1]
        final_production_efficiency = (self.operations_data['Fashionistas_Eff'].iloc[-1] + self.operations_data['Geeks_Eff'].iloc[-1]) / 2

        final_market_share = self.market_data['V4EScooters_share'].iloc[-1]
        peak_market_share = self.market_data['V4EScooters_share'].max()
        crisis_minimum_balance = self.df_finance['Ending_Balance'].min()
        crisis_rounds_to_recovery = 2

        # Marketing ROI from "great experiment"
        marketing_spend_r9 = self.df_finance[self.df_finance['Round'] == 9]['Marketing_Spend'].iloc[0]
        marketing_spend_r8 = self.df_finance[self.df_finance['Round'] == 8]['Marketing_Spend'].iloc[0]
        sales_success_r8 = self.marketing_full[self.marketing_full['Round'] == 8]['Sales_Success_Rate'].iloc[0]
        sales_success_r9 = self.marketing_full[self.marketing_full['Round'] == 9]['Sales_Success_Rate'].iloc[0]

        # Unit Economics
        final_profit_per_unit = self.df_finance[self.df_finance['Round'] == 10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round'] == 10].empty else 0
        final_revenue_per_unit = self.df_finance[self.df_finance['Round'] == 10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round'] == 10].empty else 1
        unit_profit_margin = (final_profit_per_unit / final_revenue_per_unit * 100) if final_revenue_per_unit > 0 else 0

        # Geographical Excellence
        top_city_performance = self.df_geo_v4e['Avg_Success_V4E_R7_R10'].max()
        avg_geographical_performance = self.df_geo_v4e['Avg_Success_V4E_R7_R10'].mean()

        print(f"üí∞ FINANSE I ROI:")
        print(f" ‚Ä¢ Inwestycja poczƒÖtkowa: ‚Ç¨{initial_investment:,.0f}")
        print(f" ‚Ä¢ Saldo ko≈Ñcowe: ‚Ç¨{final_balance:,.0f}")
        print(f" ‚Ä¢ Szczytowe saldo: ‚Ç¨{peak_balance:,.0f} (Runda {self.df_finance['Ending_Balance'].idxmax() + 1})")
        print(f" ‚Ä¢ Ca≈Çkowity ROI: {total_roi:.1f}%")
        print(f" ‚Ä¢ Najni≈ºszy punkt (kryzys R5): ‚Ç¨{crisis_minimum_balance:,.0f}")
        print(f" ‚Ä¢ Szybko≈õƒá odbudowy: {crisis_rounds_to_recovery} rundy (od dna do silnego zysku)")

        print(f"\n‚öôÔ∏è DOSKONA≈ÅO≈öƒÜ OPERACYJNA:")
        print(f" ‚Ä¢ Ko≈Ñcowa efektywno≈õƒá linii produkcyjnych: {final_production_efficiency:.1f}% (cel: 200%)")
        print(f" ‚Ä¢ Status efektywno≈õci maksymalnej: {'OSIƒÑGNIƒòTO' if final_production_efficiency >= 195 else 'W TRAKCIE'}")
        print(f" ‚Ä¢ Ko≈Ñcowa zdolno≈õƒá produkcyjna: {self.operations_data['Production_Capacity'].iloc[-1]:,} jednostek")
        print(f" ‚Ä¢ Wykorzystanie zdolno≈õci R10: {(final_units_produced / self.operations_data['Production_Capacity'].iloc[-1] * 100):.1f}%")
        print(f" ‚Ä¢ Optymalizacja si≈Çy roboczej: {self.operations_data['Production_Workers'].iloc[-1] + self.operations_data['Sales_Workers'].iloc[-1]} pracownik√≥w (vs szczyt {(self.operations_data['Production_Workers'] + self.operations_data['Sales_Workers']).max()})")

        print(f"\nüìà MARKETING I RYNEK:")
        print(f" ‚Ä¢ Ko≈Ñcowy udzia≈Ç w rynku: {final_market_share:.2f}%")
        print(f" ‚Ä¢ Szczytowy udzia≈Ç w rynku: {peak_market_share:.2f}% (Runda {self.market_data['V4EScooters_share'].idxmax() + 1})")
        print(f" ‚Ä¢ Strategiczne odkrycie: Bud≈ºet ‚Ç¨{marketing_spend_r8:,} ‚Üí {sales_success_r8:.1f}% sukcesu vs ‚Ç¨{marketing_spend_r9:,} ‚Üí {sales_success_r9:.1f}% sukcesu")
        # Handle division by zero for marketing efficiency if spend is 0
        mkt_eff_r8 = sales_success_r8 / marketing_spend_r8 * 1000 if marketing_spend_r8 > 0 else 'NIESKO≈ÉCZONA'
        mkt_eff_r9 = sales_success_r9 / marketing_spend_r9 * 1000 if marketing_spend_r9 > 0 else 'NIESKO≈ÉCZONA'
        print(f" ‚Ä¢ Efektywno≈õƒá marketingowa (sukces % na ‚Ç¨1k): {mkt_eff_r8:.3f}" if isinstance(mkt_eff_r8, float) else f" ‚Ä¢ Efektywno≈õƒá marketingowa (sukces % na ‚Ç¨1k): {mkt_eff_r8} (R8) vs {mkt_eff_r9:.3f} (R9)")

        print(f"\nüéØ EKONOMIKA JEDNOSTKOWA:")
        print(f" ‚Ä¢ Przych√≥d/jednostka (R10): ‚Ç¨{final_revenue_per_unit:.2f}")
        print(f" ‚Ä¢ Zysk/jednostka (R10): ‚Ç¨{final_profit_per_unit:.2f}")
        print(f" ‚Ä¢ Mar≈ºa zysku jednostkowego: {unit_profit_margin:.1f}%")
        print(f" ‚Ä¢ Sprzeda≈º ostatnia runda: {final_units_sold:,} jednostek")

        print(f"\nüåç DOSKONA≈ÅO≈öƒÜ GEOGRAFICZNA:")
        print(f" ‚Ä¢ Najlepsza wydajno≈õƒá miasta: {top_city_performance:.1f}%")
        print(f" ‚Ä¢ ≈örednia wydajno≈õƒá geograficzna (R7-R10): {avg_geographical_performance:.1f}%")
        print(f" ‚Ä¢ Miasta z >90% wydajno≈õciƒÖ: {len(self.df_geo_v4e[self.df_geo_v4e['Avg_Success_V4E_R7_R10'] > 90])}/6")

        # 8.2 STRATEGIC FRAMEWORK - Business interpretation of KPIs
        print(f"\nüéì CZƒò≈öƒÜ II: STRATEGICZNY FRAMEWORK SUKCESU - ANALIZA KPI W KONTEK≈öCIE BIZNESOWYM")
        print("-" * 100)

        # Rating based on KPIs
        roi_rating = "WYBITNY" if total_roi > 400 else "BARDZO DOBRY" if total_roi > 200 else "DOBRY" if total_roi > 100 else "PRZECIƒòTNY"
        operational_rating = "MAKSYMALNY" if final_production_efficiency >= 195 else "WYSOKI" if final_production_efficiency >= 150 else "≈öREDNI"
        market_strategy_rating = "EFEKTYWNY PREMIUM" if final_market_share > 30 and unit_profit_margin > 20 else "ROZWIJAJƒÑCY SIƒò"
        crisis_mgmt_rating = "EKSPERCKI" if crisis_rounds_to_recovery <= 2 else "DOBRY" if crisis_rounds_to_recovery <= 4 else "POWOLNY"

        print(f"üìä MATRYCA OCENY WYDAJNO≈öCI STRATEGICZNEJ:")
        print(f" ‚Ä¢ Rentowno≈õƒá finansowa: {roi_rating} ({total_roi:.1f}% ROI)")
        print(f" ‚Ä¢ Mistrzostwo operacyjne: {operational_rating} ({final_production_efficiency:.1f}% efektywno≈õci)")
        print(f" ‚Ä¢ Pozycjonowanie rynkowe: {market_strategy_rating} ({final_market_share:.1f}% udzia≈Çu, {unit_profit_margin:.1f}% mar≈ºy)")
        print(f" ‚Ä¢ ZarzƒÖdzanie kryzysowe: {crisis_mgmt_rating} (odbudowa w {crisis_rounds_to_recovery} rundach)")
        print(f" ‚Ä¢ Innowacja marketingowa: PRZE≈ÅOMOWA (odkrycie paradoksu wydatk√≥w ‚Ç¨0)")

        print(f"\nüèÜ KLUCZOWE WSKA≈πNIKI PRZEWAGI KONKURENCYJNEJ (vs rynku):")
        v4e_final_pos = int(self.market_data['V4EScooters_rank'].iloc[-1])
        market_leader_share = self.market_data['market_leader_share'].iloc[-1]
        competitive_gap = final_market_share - market_leader_share

        print(f" ‚Ä¢ Pozycja rynkowa: {v4e_final_pos}. miejsce z 4 graczy")
        print(f" ‚Ä¢ R√≥≈ºnica do lidera: {competitive_gap:+.1f} punkt√≥w procentowych")
        print(f" ‚Ä¢ Strategia r√≥≈ºnicowania: PREMIUM-PROFIT vs VOLUME-FOCUSED konkurencji")
        print(f" ‚Ä¢ Przewaga operacyjna: 200% efektywno≈õci (prawdopodobnie lider rynku)")
        print(f" ‚Ä¢ Przewaga adaptacyjna: Jedyna firma z udokumentowanƒÖ zdolno≈õciƒÖ szybkiej odbudowy po kryzysie")

        # 8.3 BUSINESS LESSONS - Strategic takeaways
        print(f"\nüí° CZƒò≈öƒÜ III: TRANSFEROWALNE LEKCJE BIZNESOWE I BEST PRACTICES")
        print("-" * 100)
        print("üîë FRAMEWORK STRATEGICZNY '4W+A' SUKCESU V4EScooters:")
        print(" 1Ô∏è‚É£ WYDAJNO≈öƒÜ (Performance): Maksymalizacja efektywno≈õci operacyjnej jako fundament")
        print(" 2Ô∏è‚É£ WARTO≈öƒÜ (Value): Pozycjonowanie premium oparte na jako≈õci i mar≈ºy, nie wolumenie")
        print(" 3Ô∏è‚É£ WGLƒÑD (Insight): Eksperymentowanie marketingowe prowadzƒÖce do prze≈Çomowych odkryƒá")
        print(" 4Ô∏è‚É£ WYTRWA≈ÅO≈öƒÜ (Persistence): ZarzƒÖdzanie kryzysowe i szybka adaptacja strategiczna")
        print(" ‚ûï ALOKACJA (Allocation): Optymalna alokacja zasob√≥w w oparciu o dane i rezultaty")

        print("\nüìà METRYKI SUKCESU DO REPLIKACJI W INNYCH BIZNESACH:")
        conversion_rate = (final_units_sold / final_units_produced * 100) if final_units_produced > 0 else 0
        print(f" ‚Ä¢ Model konwersji: {conversion_rate:.1f}% (produkcja ‚Üí sprzeda≈º)")
        print(f" ‚Ä¢ Benchmark efektywno≈õci: {final_production_efficiency:.0f}% (osiƒÖgniƒôcie teoretycznego maximum)")
        print(f" ‚Ä¢ Wzorzec ROI marketingowego: ‚Ç¨{marketing_spend_r8:,} input ‚Üí {sales_success_r8:.1f}% output (paradoks efektywno≈õci)")
        print(f" ‚Ä¢ Standard odbudowy: {crisis_rounds_to_recovery} okresy od kryzysu do wzrostu")
        print(f" ‚Ä¢ Target mar≈ºy jednostkowej: {unit_profit_margin:.1f}% w strategii premium")

        print("\nüéØ STRATEGICZNE PYTANIA DLA KA≈ªDEGO BIZNESU (oparte na V4E experience):")
        print(" ‚ùì Czy Twoja firma osiƒÖgnƒô≈Ça maksimum mo≈ºliwej efektywno≈õci operacyjnej?")
        print(" ‚ùì Czy testujesz niekonwencjonalne podej≈õcia marketingowe (jak eksperyment ‚Ç¨0)?")
        print(" ‚ùì Czy Twoja strategia priorytetyzuje d≈ÇugoterminowƒÖ rentowno≈õƒá nad kr√≥tkookresowym udzia≈Çem w rynku?")
        print(" ‚ùì Czy masz udokumentowany plan zarzƒÖdzania kryzysowego z metrykami odbudowy?")
        print(" ‚ùì Czy regularnie optymalizujesz alokacjƒô zasob√≥w w oparciu o rzeczywiste rezultaty?")

        # 8.4 FREE SPACE FOR NOTES AND REFLECTIONS
        print(f"\nüìù SEKCJA NOTATEK I DALSZYCH PRZEMY≈öLE≈É:")
        print("=" * 100)
        print("\n" + " " * 20 + "[ PRZESTRZE≈É NA W≈ÅASNE WNIOSKI I PRZEMY≈öLENIA ]")
        print("\n" + "_" * 90)
        print("KLUCZOWE OBSERWACJE, KT√ìRE CHCE ZAPAMIƒòTAƒÜ:")
        print("_" * 90)
        for i in range(5):
            print(f"{i + 1}. ________________________________________________")
            print(" ________________________________________________\n")
        print("_" * 90)
        print("POMYS≈ÅY NA IMPLEMENTACJƒò W MOIM BIZNESIE/PROJEKCIE:")
        print("_" * 90)
        for i in range(5):
            print(f"{i + 1}. ________________________________________________")
            print(" ________________________________________________\n")
        print("_" * 90)
        print("PYTANIA DO DALSZEGO BADANIA:")
        print("_" * 90)
        for i in range(3):
            print(f"{i + 1}. ________________________________________________")
            print(" ________________________________________________\n")
        print("_" * 90)
        print("METRYKI, KT√ìRE CHCƒò ≈öLEDZIƒÜ W MOIM KONTEK≈öCIE:")
        print("_" * 90)
        for i in range(4):
            print(f"‚Ä¢ KPI {i + 1}: _______________________________________")
            print(" Target: _________________ Deadline: _____________\n")

        print("\n" + "=" * 100)
        print(" " * 15 + "‚úÖ V4EScooters - KOMPLETNA ANALIZA ZAKO≈ÉCZONA! ‚úÖ")
        print(" " * 10 + "üéØ Od startupu przez kryzys do mistrzostwa strategicznego üéØ")
        print("=" * 100)

    def generate_report(self):
        """
        Generates the complete business game report by calling all
        data loading, processing, printing, and plotting methods.
        Enhanced version with deep strategic insights.
        """
        self._load_and_process_data()
        self._print_header()
        self._plot_financial_performance()
        self._plot_marketing_strategy()
        self._plot_unit_economics()
        self._plot_operational_excellence()
        self._plot_competitive_dynamics()
        self._plot_product_strategy()
        self._plot_geographical_strategy()
        self._print_strategic_insights_framework()  # Enhanced strategic section
        self._print_final_conclusions()


# To run the report:
if __name__ == "__main__":
    # Create an instance of the reporter and generate the report
    reporter = BusinessGameReporter()
    reporter.generate_report()
```
