```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')

class BusinessGameReporter:
    """
    Generates a comprehensive business game report with financial, marketing,
    operational, competitive, product, and geographical analysis.
    """

    def __init__(self, initial_investment=1_000_000):
        """
        Initializes the reporter with global settings and data.

        Args:
            initial_investment (int): The initial investment amount for the game.
        """
        self.initial_investment = initial_investment
        self._setup_plotting_style()
        self.COLORS = {
            'primary': '#1f77b4',
            'secondary': '#ff7f0e',
            'success': '#2ca02c',
            'warning': '#d62728',
            'danger': '#e74c3c',
            'info': '#17a2b8',
            'purple': '#9467bd',
            'brown': '#8c564b',
            'pink': '#e377c2',
            'gray': '#7f7f7f'
        }
        self.df_finance = pd.DataFrame()
        self.marketing_full = pd.DataFrame()
        self.operations_data = pd.DataFrame()
        self.market_data = pd.DataFrame()
        self.product_data_v4e = pd.DataFrame()
        self.df_geo_v4e = pd.DataFrame()
        self.plot_teams_map = {} # Will be populated in _load_data

    def _setup_plotting_style(self):
        """Sets up professional plotting style and parameters."""
        plt.style.use('seaborn-v0_8-darkgrid')
        plt.rcParams['figure.figsize'] = (16, 10)
        plt.rcParams['font.size'] = 12
        plt.rcParams['font.family'] = 'DejaVu Sans'
        plt.rcParams['axes.titlesize'] = 16
        plt.rcParams['axes.titleweight'] = 'bold'
        plt.rcParams['axes.labelsize'] = 14
        plt.rcParams['xtick.labelsize'] = 12
        plt.rcParams['ytick.labelsize'] = 12
        plt.rcParams['legend.fontsize'] = 12
        plt.rcParams['figure.titlesize'] = 20
        plt.rcParams['figure.titleweight'] = 'bold'

    def _load_and_process_data(self):
        """Loads raw data and performs initial calculations for all sections."""
        # Financial Data
        financial_raw_data = {
            'Round': list(range(1, 11)),  # Extended to include Turn 11
            'Starting_Balance': [1000000, 749000, 1096813, 1343786, 1792392, -144627, 268362, 3340282, 5185242, 4758356],
            'Ending_Balance': [749000, 1096813, 1343786, 1792392, -144627, 268362, 3340282, 5185242, 4758356, 5976703],  # T11 ending balance not provided
            'Net_Change': [-251000, 347813, 246973, 448606, -1937019, 412989, 3071920, 1844960, -426886, 1218347],  # T11 net change not available yet
            'Marketing_Spend': [62200, 3500, 3500, 3500, 3500, 3500, 310000, 0, 1500000, 84500],  # T11 data not available
            'Production_Cost': [0, 406550, 1020500, 1719000, 1719000, 1718000, 2081000, 2356000, 2356000, 2356000],  # T11 data not available
            'HR_Cost': [27300, 163710, 134800, 218140, 191640, 292260, 247910, 343420, 313760, 223300],  # T11 data not available
            'Operations_Cost': [161500, 573760, 1158800, 1940640, 1914140, 2013760, 2638910, 2699420, 4169760, 2663800],  # T11 data not available
            'Infrastructure_Invest': [251000, 0, 0, 10000, 0, 0, 0, 0, 0, 0],  # T11 data not available
            'Units_Produced': [0, 6500, 6500, 12000, 12000, 12000, 15000, 17000, 17000, 17000],  # T11 data not available
            'Units_Sold': [0, 6500, 12000, 11849, 151, 11744, 22324, 17873, 17104, 15000]  # T11 data not available
        }

        self.df_finance = pd.DataFrame(financial_raw_data)

        self.df_finance['Total_Costs'] = (self.df_finance['Marketing_Spend'] + self.df_finance['Production_Cost'] +
                                            self.df_finance['HR_Cost'] + self.df_finance['Operations_Cost'])
        self.df_finance['Implied_Revenue'] = self.df_finance['Net_Change'] + self.df_finance['Total_Costs']
        self.df_finance['Profit_Margin'] = np.where(self.df_finance['Implied_Revenue'] > 0,
                                                     (self.df_finance['Implied_Revenue'] - self.df_finance['Total_Costs']) /
                                                     self.df_finance['Implied_Revenue'] * 100, 0)

        # Unit Economics Calculations
        self.df_finance['Production_Cost_Per_Unit'] = np.where(self.df_finance['Units_Produced'] > 0,
                                                                self.df_finance['Production_Cost'] / self.df_finance['Units_Produced'], 0)
        self.df_finance['Marketing_Cost_Per_Unit_Sold'] = np.where(self.df_finance['Units_Sold'] > 0,
                                                                    self.df_finance['Marketing_Spend'] / self.df_finance['Units_Sold'], np.nan)
        self.df_finance['Total_Cost_Per_Unit_Sold'] = np.where(self.df_finance['Units_Sold'] > 0,
                                                                self.df_finance['Total_Costs'] / self.df_finance['Units_Sold'], 0)
        self.df_finance['Revenue_Per_Unit_Sold'] = np.where(self.df_finance['Units_Sold'] > 0,
                                                             self.df_finance['Implied_Revenue'] / self.df_finance['Units_Sold'], 0)
        self.df_finance['Profit_Per_Unit_Sold'] = self.df_finance['Revenue_Per_Unit_Sold'] - self.df_finance['Total_Cost_Per_Unit_Sold']

        # Marketing Data
        self.marketing_full = self.df_finance[['Round', 'Marketing_Spend', 'Units_Sold', 'Units_Produced']].copy()
        campaign_details_map = {
            1: 'Ustawienia poczƒÖtkowe', 2: 'Utrzymanie', 3: 'Utrzymanie', 4: 'Utrzymanie',
            5: 'Utrzymanie (wp≈Çyw kryzysu)', 6: 'Utrzymanie (odbudowa)', 7: 'Intensywna kampania',
            8: 'Test organiczny (‚Ç¨0)', 9: 'MEGA kampania', 10: 'Zoptymalizowana minimalna'
        }
        self.marketing_full['Campaign_Details'] = self.marketing_full['Round'].map(campaign_details_map)
        self.marketing_full['Sales_Success_Rate'] = np.where(self.marketing_full['Units_Produced'] > 0,
                                                             (self.marketing_full['Units_Sold'] / self.marketing_full['Units_Produced']) * 100, 0)
        v4e_market_share_data = [0.00, 0.00, 24.51, 22.29, 0.92, 30.30, 52.48, 57.46, 38.08, 39.87]
        self.marketing_full['Market_Share_V4E'] = v4e_market_share_data
        self.marketing_full['Direct_Sale_Ads'] = [0, 0, 9, 9, 9, 9, 20, 0, 110, 9]
        self.marketing_full['Ambassador_Ads'] = [0, 0, 9, 9, 9, 9, 15, 0, 110, 9]
        self.marketing_full['Newspaper_Ads'] = [0, 0, 9, 9, 9, 9, 60, 0, 150, 9]
        self.marketing_full['Cost_Per_Sale'] = np.where(self.marketing_full['Units_Sold'] > 0,
                                                         self.marketing_full['Marketing_Spend'] / self.marketing_full['Units_Sold'], np.nan)
        self.marketing_full['Marketing_Intensity_Millions'] = self.marketing_full['Marketing_Spend'] / 1_000_000

        # Operational Data
        operations_raw_data = {
            'Round': list(range(1, 11)),
            'Fashionistas_Eff': [0, 106.98, 115.36, 104.61, 114.68, 111.18, 140.62, 160.02, 200.00, 200.00],
            'Geeks_Eff': [0, 106.86, 105.66, 100.26, 109.79, 107.95, 122.84, 139.91, 175.67, 200.00],
            'Production_Capacity': [5000, 15600, 15600, 17000, 17000, 17000, 18880, 21494, 26925, 20000],
            'Sales_Offices': [0, 6, 6, 6, 12, 12, 12, 12, 12, 12],
            'Office_Efficiency_Avg': [0, 118.5, 120.8, 125.6, 108.2, 135.6, 162.4, 155.3, 166.7, 176.8],
            'Production_Workers': [10, 65, 65, 75, 106, 106, 105, 115, 115, 80],
            'Sales_Workers': [0, 18, 18, 18, 18, 36, 36, 36, 36, 33],
            'Training_Investment': [10500, 44210, 0, 35860, 9360, 79860, 35510, 131020, 101360, 64700]
        }
        self.operations_data = pd.DataFrame(operations_raw_data)
        self.operations_data = pd.merge(self.operations_data, self.df_finance[['Round', 'Units_Produced', 'Units_Sold']], on='Round', how='left')
        self.operations_data['Units_Per_Production_Worker'] = np.where(self.operations_data['Production_Workers'] > 0,
                                                                         self.operations_data['Units_Produced'] / self.operations_data['Production_Workers'], 0)
        self.operations_data['Sales_Per_Sales_Worker'] = np.where(self.operations_data['Sales_Workers'] > 0,
                                                                   self.operations_data['Units_Sold'] / self.operations_data['Sales_Workers'], 0)

        # Market Data
        self.market_data = pd.DataFrame({
            'Round': list(range(1, 11)),
            'team_5': [0.00, 0.00, 8.28, 20.21, 59.74, 31.45, 14.40, 10.42, 13.65, 19.07],
            'KKKompany': [0.00, 0.00, 30.14, 27.46, 5.21, 41.10, 33.20, 19.91, 43.40, 41.06],
            'V4EScooters_share': [0.00, 0.00, 24.51, 22.29, 0.92, 30.30, 52.48, 57.46, 38.08, 39.87],
            'team_8': [0.00, 0.00, 37.07, 30.04, 62.42, 14.20, 3.90, 8.98, 2.73, 0.00]
        })
        # Correction for V4EScooters_share based on marketing_full data
        self.market_data['V4EScooters_share'] = self.marketing_full['Market_Share_V4E'].values

        self.plot_teams_map = {
            'team_5': ('Zesp√≥≈Ç 5', self.COLORS['purple']),
            'KKKompany': ('KKKompany', self.COLORS['secondary']),
            'V4EScooters_share': ('V4E', self.COLORS['info']),
            'team_8': ('Zesp√≥≈Ç 8', self.COLORS['danger'])
        }
        self.competitor_columns_plot = list(self.plot_teams_map.keys())

        self.market_data['V4EScooters_rank'] = self.market_data[self.competitor_columns_plot].rank(axis=1, method='min', ascending=False)['V4EScooters_share']
        self.market_data['market_leader_share'] = self.market_data[self.competitor_columns_plot].max(axis=1)
        self.market_data['V4EScooters_relative_to_leader'] = np.where(self.market_data['market_leader_share'] > 0,
                                                                        (self.market_data['V4EScooters_share'] / self.market_data['market_leader_share']) * 100, 0)

        # Product Data
        self.product_data_v4e = pd.DataFrame({
            'Round': [3, 4, 5, 6, 7, 8, 9, 10],
            'Fashionistas_Produced_V4E': [1000, 3000, 3000, 3000, 3500, 4000, 4000, 4000],
            'Geeks_Produced_V4E': [5000, 7000, 7000, 7000, 8000, 9000, 9000, 9000],
            'Savers_Produced_V4E': [500, 2000, 2000, 2000, 3500, 4000, 4000, 4000],
            'Fashionistas_Sold_V4E': [1000, 3000, 0, 1000, 3000, 4000, 3600, 3400],
            'Geeks_Sold_V4E': [5000, 7000, 0, 5000, 7000, 9000, 8100, 8100],
            'Savers_Sold_V4E': [500, 1849, 151, 2160, 3244, 4000, 3600, 3400],
            'Fashionistas_Price_V4E': [155, 160, 160, 150, 275, 280, 280, 280],
            'Geeks_Price_V4E': [165, 154, 161, 156, 275, 275, 275, 275],
            'Savers_Price_V4E': [91, 88, 91, 88, 155, 155, 155, 155]
        })
        self.product_data_v4e['Total_Produced_V4E'] = (self.product_data_v4e['Fashionistas_Produced_V4E'] +
                                                         self.product_data_v4e['Geeks_Produced_V4E'] +
                                                         self.product_data_v4e['Savers_Produced_V4E'])
        self.product_data_v4e['Total_Sold_V4E'] = (self.product_data_v4e['Fashionistas_Sold_V4E'] +
                                                     self.product_data_v4e['Geeks_Sold_V4E'] +
                                                     self.product_data_v4e['Savers_Sold_V4E'])
        self.product_data_v4e['Avg_Selling_Price_V4E'] = np.where(
            self.product_data_v4e['Total_Sold_V4E'] > 0,
            ((self.product_data_v4e['Fashionistas_Price_V4E'] * self.product_data_v4e['Fashionistas_Sold_V4E'] +
              self.product_data_v4e['Geeks_Price_V4E'] * self.product_data_v4e['Geeks_Sold_V4E'] +
              self.product_data_v4e['Savers_Price_V4E'] * self.product_data_v4e['Savers_Sold_V4E']) /
             self.product_data_v4e['Total_Sold_V4E']),
            0
        )

        # Geographical Data
        geo_raw_data = {
            'City': ['Bratys≈Çawa', 'Budapeszt', 'Praga', 'Sofia', 'Warszawa', 'Zagrzeb'],
            'R3_Success_V4E': [100, 100, 100, 100, 100, 100],
            'R4_Success_V4E': [90, 95, 92, 98, 96, 93],
            'R5_Success_V4E': [5, 0, 2, 10, 3, 1],
            'R6_Success_V4E': [50, 60, 55, 70, 65, 58],
            'R7_Success_V4E': [70, 85, 75, 90, 88, 72],
            'R8_Success_V4E': [75, 90, 80, 95, 92, 78],
            'R9_Success_V4E': [80, 92, 85, 97, 94, 82],
            'R10_Success_V4E': [85, 95, 88, 99, 96, 86],
            'Ref_Avg_Price_Market': [235, 240, 250, 237, 237, 233],
            'Market_Size_Category': ['Ma≈Çy', 'Du≈ºy', 'Du≈ºy', '≈öredni', 'Du≈ºy', '≈öredni'],
            'Competition_Level': ['Wysoka', '≈örednia', 'Wysoka', 'Niska', '≈örednia', '≈örednia']
        }
        self.df_geo_v4e = pd.DataFrame(geo_raw_data)
        self.df_geo_v4e['Avg_Success_V4E_R7_R10'] = self.df_geo_v4e[['R7_Success_V4E', 'R8_Success_V4E',
                                                                     'R9_Success_V4E', 'R10_Success_V4E']].mean(axis=1)

    def _print_header(self):
        """Prints the report header and executive summary."""
        print("="*100)
        print(" " * 15 + "üèÜ RAPORT KO≈ÉCOWY Z GRY BIZNESOWEJ - V4EScooters: E-HULAJNOGI PREMIUM ÔøΩ")
        print("="*100)
        print(f"\nRaport wygenerowany: {datetime.now().strftime('%d %B %Y o %H:%M')}")
        print(f"Okres analizy: Rundy 1-10")
        print(f"Podej≈õcie strategiczne: Pozycjonowanie Premium i Doskona≈Ço≈õƒá Operacyjna")
        print("\n" + "="*100)

        print("\nüìä PODSUMOWANIE WYKONAWCZE")
        print("="*100)
        final_balance = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi = ((final_balance - self.initial_investment) / self.initial_investment) * 100
        final_market_share = self.market_data['V4EScooters_share'].iloc[-1]
        final_rank = int(self.market_data['V4EScooters_rank'].iloc[-1])

        print(f"{'Inwestycja poczƒÖtkowa:':<30} ‚Ç¨{self.initial_investment:,.0f}")
        print(f"{'Ko≈Ñcowe saldo:':<30} ‚Ç¨{final_balance:,.0f}")
        print(f"{'Ca≈Çkowity zwrot z inwestycji:':<30} {total_roi:.2f}%")
        print(f"{'Ko≈Ñcowy udzia≈Ç w rynku:':<30} {final_market_share:.2f}% ({final_rank}. miejsce)")
        print(f"{'Ranking rentowno≈õci:':<30} 1. miejsce (szacunkowo)")
        print(f"{'Filozofia strategiczna:':<30} 'Jako≈õƒá ponad Ilo≈õƒá'")
        print("="*100)

    def _plot_financial_performance(self):
        """Generates plots for financial performance analysis."""
        print("\n\nüèõÔ∏è SEKCJA 1: ANALIZA WYNIK√ìW FINANSOWYCH")
        print("="*100)

        # Figure 1.1 (4 plots)
        fig1_1 = plt.figure(figsize=(20, 12))
        gs1_1 = fig1_1.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

        # 1.1 Ewolucja salda z fazami
        ax1_1 = fig1_1.add_subplot(gs1_1[0, 0])
        ax1_1.plot(self.df_finance['Round'], self.df_finance['Ending_Balance']/1000,
                   color=self.COLORS['primary'], linewidth=4, marker='o', markersize=12, label='Saldo')
        ax1_1.axhline(y=0, color=self.COLORS['danger'], linestyle='--', alpha=0.7, linewidth=2)

        ax1_1.axvspan(0.5, 3.5, alpha=0.1, color=self.COLORS['info'], label='Faza Fundament√≥w')
        ax1_1.axvspan(3.5, 5.5, alpha=0.1, color=self.COLORS['danger'], label='Faza Kryzysu')
        ax1_1.axvspan(5.5, 7.5, alpha=0.1, color=self.COLORS['success'], label='Faza Odbudowy')
        ax1_1.axvspan(7.5, 10.5, alpha=0.1, color=self.COLORS['purple'], label='Faza Optymalizacji')

        ax1_1.annotate('Inwestycja poczƒÖtkowa\n‚Ç¨1M', xy=(1, 749), xytext=(1, 2000),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['gray'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['gray']))
        ax1_1.annotate('Pierwszy kryzys\n-‚Ç¨144,627', xy=(5, -144.627), xytext=(5, -1500),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['danger'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['danger']))
        ax1_1.annotate('Rekordowy zysk\n‚Ç¨3.34M', xy=(7, 3340), xytext=(7, 4500),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['success'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['success']))
        ax1_1.annotate('Ostateczne zwyciƒôstwo\n‚Ç¨5.98M', xy=(10, 5977), xytext=(10, 7000),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['purple'], lw=2),
                       fontsize=11, ha='center', bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['purple']))

        ax1_1.set_xlabel('Runda')
        ax1_1.set_ylabel('Saldo (tys. ‚Ç¨)')
        ax1_1.set_title('Podr√≥≈º finansowa: Od startupu do sukcesu')
        ax1_1.grid(True, alpha=0.3)
        ax1_1.legend(loc='upper left', frameon=True, fancybox=True, shadow=True)
        ax1_1.set_xlim(0.5, 10.5)
        ax1_1.set_ylim(-2000, 8000)
        ax1_1.set_xticks(self.df_finance['Round'])

        # 1.2 Ewolucja struktury koszt√≥w
        ax1_2 = fig1_1.add_subplot(gs1_1[0, 1])
        cost_categories = ['Marketing', 'Produkcja', 'HR', 'Operacje']
        rounds_sample = [2, 4, 7, 10]
        cost_data = np.array([
            [self.df_finance.loc[self.df_finance['Round'] == r, 'Marketing_Spend'].values[0]/1000 if r in self.df_finance['Round'].values else 0 for r in rounds_sample],
            [self.df_finance.loc[self.df_finance['Round'] == r, 'Production_Cost'].values[0]/1000 if r in self.df_finance['Round'].values else 0 for r in rounds_sample],
            [self.df_finance.loc[self.df_finance['Round'] == r, 'HR_Cost'].values[0]/1000 if r in self.df_finance['Round'].values else 0 for r in rounds_sample],
            [self.df_finance.loc[self.df_finance['Round'] == r, 'Operations_Cost'].values[0]/1000 if r in self.df_finance['Round'].values else 0 for r in rounds_sample]
        ])

        bottom = np.zeros(len(rounds_sample))
        colors_cost = [self.COLORS['warning'], self.COLORS['primary'], self.COLORS['success'], self.COLORS['info']]
        for i, category in enumerate(cost_categories):
            ax1_2.bar(rounds_sample, cost_data[i], bottom=bottom, label=category,
                      color=colors_cost[i], alpha=0.8)
            bottom += cost_data[i]

        ax1_2.set_xlabel('Runda')
        ax1_2.set_ylabel('Koszt (tys. ‚Ç¨)')
        ax1_2.set_title('Ewolucja struktury koszt√≥w')
        ax1_2.legend(loc='upper left', frameon=True)
        ax1_2.grid(True, axis='y', alpha=0.3)
        ax1_2.set_xticks(rounds_sample)

        # 1.3 Postƒôp ROI
        ax1_3 = fig1_1.add_subplot(gs1_1[1, 0])
        roi = ((self.df_finance['Ending_Balance'] - self.initial_investment) / self.initial_investment * 100)
        ax1_3.plot(self.df_finance['Round'], roi, color=self.COLORS['purple'],
                   linewidth=4, marker='d', markersize=12)
        ax1_3.fill_between(self.df_finance['Round'], roi, 0, where=(roi >= 0),
                           alpha=0.3, color=self.COLORS['success'], label='Pozytywne ROI')
        ax1_3.fill_between(self.df_finance['Round'], roi, 0, where=(roi < 0),
                           alpha=0.3, color=self.COLORS['danger'], label='Negatywne ROI')
        ax1_3.set_xlabel('Runda')
        ax1_3.set_ylabel('ROI (%)')
        ax1_3.set_title('Postƒôp zwrotu z inwestycji')
        ax1_3.grid(True, alpha=0.3)
        ax1_3.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax1_3.legend(frameon=True)
        ax1_3.set_xticks(self.df_finance['Round'])

        ax1_3.annotate(f'Ko≈Ñcowe ROI\n{roi.iloc[-1]:.1f}%',
                       xy=(10, roi.iloc[-1]), xytext=(8.5, roi.iloc[-1] - 100 if roi.iloc[-1] > 300 else roi.iloc[-1] + 100),
                       arrowprops=dict(arrowstyle='->', color=self.COLORS['purple'], lw=2),
                       fontsize=12, ha='center', fontweight='bold',
                       bbox=dict(boxstyle="round,pad=0.5", facecolor='white', edgecolor=self.COLORS['purple']))

        # 1.4 Trend mar≈ºy zysku
        ax1_4 = fig1_1.add_subplot(gs1_1[1, 1])
        valid_margins_df = self.df_finance[self.df_finance['Implied_Revenue'] > 0]
        ax1_4.bar(valid_margins_df['Round'], valid_margins_df['Profit_Margin'], color=self.COLORS['info'], alpha=0.7)
        ax1_4.set_xlabel('Runda')
        ax1_4.set_ylabel('Mar≈ºa zysku (%)')
        ax1_4.set_title('Ewolucja mar≈ºy zysku (przy pozytywnych przychodach)')
        ax1_4.grid(True, axis='y', alpha=0.3)
        ax1_4.set_xticks(valid_margins_df['Round'])
        for i, (r, m) in enumerate(zip(valid_margins_df['Round'], valid_margins_df['Profit_Margin'])):
            ax1_4.text(r, m + (2 if m > 0 else -5), f'{m:.1f}%', ha='center', fontweight='bold')

        fig1_1.suptitle('Analiza wynik√≥w finansowych - Czƒô≈õƒá 1: Podr√≥≈º finansowa i rentowno≈õƒá',
                         fontsize=20, fontweight='bold', y=0.98)
        fig1_1.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

        # Figure 1.2 (2 plots)
        fig1_2 = plt.figure(figsize=(20, 8))
        gs1_2 = fig1_2.add_gridspec(1, 2, hspace=0.3, wspace=0.3)

        # 1.5 Wodospad przychod√≥w vs. koszt√≥w
        ax1_5 = fig1_2.add_subplot(gs1_2[0, 0])
        rounds_plot = self.df_finance['Round'].values
        revenues = self.df_finance['Implied_Revenue'].values / 1000
        costs = self.df_finance['Total_Costs'].values / 1000
        profits = self.df_finance['Net_Change'].values / 1000

        x_plot = np.arange(len(rounds_plot))
        width_plot = 0.8

        bars1 = ax1_5.bar(x_plot, revenues, width_plot, label='Szacowane Przychody',
                          color=self.COLORS['success'], alpha=0.7)
        bars2 = ax1_5.bar(x_plot, -costs, width_plot, label='Ca≈Çkowite Koszty',
                          color=self.COLORS['warning'], alpha=0.7)
        ax1_5.plot(x_plot, profits, color=self.COLORS['primary'], linewidth=3,
                   marker='o', markersize=10, label='Zysk/Strata Netto', zorder=5)

        for i in [6, 7, 9]:
            round_index = i
            if round_index < len(revenues) and revenues[round_index] > 0:
                ax1_5.text(round_index, revenues[round_index] + 100, f'‚Ç¨{revenues[round_index]:.0f}k',
                           ha='center', va='bottom', fontweight='bold')
            if round_index < len(profits):
                ax1_5.text(round_index, profits[round_index] + 200, f'{profits[round_index]:+.0f}k',
                           ha='center', va='bottom', color=self.COLORS['primary'], fontweight='bold')

        ax1_5.set_xlabel('Runda')
        ax1_5.set_ylabel('Kwota (tys. ‚Ç¨)')
        ax1_5.set_title('Analiza przychod√≥w vs. koszt√≥w - Ca≈Ça podr√≥≈º')
        ax1_5.set_xticks(x_plot)
        ax1_5.set_xticklabels(rounds_plot)
        ax1_5.legend(loc='upper left', frameon=True)
        ax1_5.grid(True, axis='y', alpha=0.3)
        ax1_5.axhline(y=0, color='black', linestyle='-', alpha=0.3)

        # 1.6 Podsumowanie kluczowych wska≈∫nik√≥w finansowych
        ax1_6 = fig1_2.add_subplot(gs1_2[0, 1])
        ax1_6.axis('off')
        total_investment_val = self.df_finance['Infrastructure_Invest'].sum()
        total_revenue_val = self.df_finance['Implied_Revenue'].sum()
        total_costs_val = self.df_finance['Total_Costs'].sum()
        net_profit_val = self.df_finance['Ending_Balance'].iloc[-1] - self.initial_investment
        avg_profit_margin_val = self.df_finance[self.df_finance['Implied_Revenue']>0]['Profit_Margin'].mean()

        metrics_summary = [
            ['Wska≈∫nik', 'Warto≈õƒá'],
            ['Ca≈Çkowita inwestycja w infrastrukturƒô', f'‚Ç¨{total_investment_val/1e3:,.0f}k'],
            ['Ca≈Çkowity szacowany przych√≥d', f'‚Ç¨{total_revenue_val/1e6:,.1f}M'],
            ['Ca≈Çkowite poniesione koszty', f'‚Ç¨{total_costs_val/1e6:,.1f}M'],
            ['Ca≈Çkowity zysk netto', f'‚Ç¨{net_profit_val/1e6:,.2f}M'],
            ['Najwy≈ºsze saldo', f'‚Ç¨{self.df_finance["Ending_Balance"].max()/1e6:,.2f}M'],
            ['Najni≈ºsze saldo', f'‚Ç¨{self.df_finance["Ending_Balance"].min()/1e3:,.0f}k'],
            ['≈örednia mar≈ºa zysku (wa≈ºne rundy)', f'{avg_profit_margin_val:.1f}%' if pd.notna(avg_profit_margin_val) else 'N/A'],
            ['Szybko≈õƒá odbudowy po kryzysie', '1-2 rundy']
        ]

        table = ax1_6.table(cellText=metrics_summary, cellLoc='left', loc='center', colWidths=[0.65, 0.35])
        table.auto_set_font_size(False)
        table.set_fontsize(11)
        table.scale(1, 1.8)

        for i in range(2):
            table[(0, i)].set_facecolor(self.COLORS['primary'])
            table[(0, i)].set_text_props(weight='bold', color='white')

        for i in range(1, len(metrics_summary)):
            if i % 2 == 0:
                for j in range(2):
                    table[(i, j)].set_facecolor('#e9ecef')

        ax1_6.set_title('Podsumowanie kluczowych wska≈∫nik√≥w finansowych', fontsize=16, pad=20)

        fig1_2.suptitle('Analiza wynik√≥w finansowych - Czƒô≈õƒá 2: Przychody, koszty i podsumowanie',
                         fontsize=20, fontweight='bold', y=0.98)
        fig1_2.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

    def _plot_marketing_strategy(self):
        """Generates plots for marketing strategy analysis."""
        print("\n\nüéØ SEKCJA 2: STRATEGIA MARKETINGOWA - WIELKI EKSPERYMENT")
        print("="*100)

        experiment_data = self.marketing_full[self.marketing_full['Round'].isin([7, 8, 9, 10])].copy()
        print("\nWIELKI EKSPERYMENT MARKETINGOWY (Rundy 7-10):")
        print("-" * 100)
        print(f"{'Runda':<8} {'Bud≈ºet (‚Ç¨)':<12} {'Strategia':<25} {'Wska≈∫nik sprzeda≈ºy (%)':<15} {'Udzia≈Ç w rynku (%)':<15} {'Kluczowe odkrycie':<40}")
        print("-" * 100)
        for _, row in experiment_data.iterrows():
            finding = ""
            if row['Round'] == 7: finding = "Du≈ºe wydatki ustalajƒÖ punkt odniesienia"
            elif row['Round'] == 8: finding = "Bud≈ºet ‚Ç¨0 zaskakujƒÖco efektywny!"
            elif row['Round'] == 9: finding = "Mega wydatki, malejƒÖce zwroty"
            elif row['Round'] == 10: finding = "Zoptymalizowane wydatki, dobra efektywno≈õƒá"
            print(f"{row['Round']:<8} {row['Marketing_Spend']:>10,.0f} {row['Campaign_Details']:<25} "
                  f"{row['Sales_Success_Rate']:>14.1f} {row['Market_Share_V4E']:>14.2f} {finding:<40}")

        fig_mkt, axes_mkt = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.35)

        # 2.1 Ewolucja bud≈ºetu marketingowego
        ax2_1_mkt = axes_mkt[0, 0]
        colors_budget = [self.COLORS['gray']]*6 + [self.COLORS['warning'], self.COLORS['success'], self.COLORS['danger'], self.COLORS['info']]
        ax2_1_mkt.bar(self.marketing_full['Round'], self.marketing_full['Marketing_Spend']/1000, color=colors_budget)
        ax2_1_mkt.set_xlabel('Runda')
        ax2_1_mkt.set_ylabel('Bud≈ºet marketingowy (tys. ‚Ç¨)')
        ax2_1_mkt.set_title('Ewolucja bud≈ºetu marketingowego - Zmiany strategiczne')
        ax2_1_mkt.grid(True, axis='y', alpha=0.3)
        ax2_1_mkt.set_xticks(self.marketing_full['Round'])
        ax2_1_mkt.annotate('Intensywna kampania', xy=(7, 310), xytext=(6.5, 400),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['warning'], lw=1.5), ha='center', fontsize=10)
        ax2_1_mkt.annotate('Test ‚Ç¨0', xy=(8, 0), xytext=(7.5, 100),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['success'], lw=1.5), ha='center', fontsize=10)
        ax2_1_mkt.annotate('MEGA wydatki', xy=(9, 1500), xytext=(8.5, 1600),
                           arrowprops=dict(arrowstyle='->', color=self.COLORS['danger'], lw=1.5), ha='center', fontsize=10)

        # 2.2 Paradoks marketingowy: Bud≈ºet vs. wska≈∫nik sukcesu sprzeda≈ºy
        ax2_2_mkt = axes_mkt[0, 1]
        exp_rounds_mkt = experiment_data['Round'].values
        exp_budget_mkt = experiment_data['Marketing_Spend'].values / 1000
        exp_sales_success_mkt = experiment_data['Sales_Success_Rate'].values

        ax2_2_mkt_twin = ax2_2_mkt.twinx()
        line1 = ax2_2_mkt.plot(exp_rounds_mkt, exp_budget_mkt, 'o-', color=self.COLORS['warning'],
                                 linewidth=3, markersize=10, label='Bud≈ºet marketingowy (tys. ‚Ç¨)')
        line2 = ax2_2_mkt_twin.plot(exp_rounds_mkt, exp_sales_success_mkt, 's-', color=self.COLORS['success'],
                                     linewidth=3, markersize=10, label='Wska≈∫nik sukcesu sprzeda≈ºy (%)')

        ax2_2_mkt.set_xlabel('Runda eksperymentu (7-10)')
        ax2_2_mkt.set_ylabel('Bud≈ºet marketingowy (tys. ‚Ç¨)', color=self.COLORS['warning'])
        ax2_2_mkt_twin.set_ylabel('Wska≈∫nik sukcesu sprzeda≈ºy (%)', color=self.COLORS['success'])
        ax2_2_mkt.set_title('Paradoks marketingowy: Wydatki vs. sukces sprzeda≈ºy')
        ax2_2_mkt.grid(True, alpha=0.3)
        ax2_2_mkt.set_xticks(exp_rounds_mkt)
        lines = line1 + line2
        labels = [l.get_label() for l in lines]
        ax2_2_mkt.legend(lines, labels, loc='center left', frameon=True, bbox_to_anchor=(0.05, 0.6))
        ax2_2_mkt.annotate('Punkt paradoksu:\nBud≈ºet ‚Ç¨0 ‚Üí 92.7% sukcesu (R8)\nvs\nBud≈ºet ‚Ç¨1.5M ‚Üí 90.5% sukcesu (R9)',
                            xy=(8, exp_sales_success_mkt[exp_rounds_mkt == 8][0]),
                            xytext=(8.5, 60), ha='center', fontsize=10, fontweight='bold',
                            bbox=dict(boxstyle="round,pad=0.5", facecolor='yellow', alpha=0.7),
                            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='black'))

        # 2.3 Analiza wp≈Çywu wydatk√≥w marketingowych na zyski w nastƒôpnej rundzie
        ax2_3_mkt = axes_mkt[1, 0]
        
        # We want to see the impact of marketing spend in round N on earnings (Net_Change) in round N+1.
        # We will calculate this for rounds where marketing spend was significant and a next round exists.
        # Rounds 7 and 9 are good candidates from the "experiment_data" context.
        
        impact_data_mkt = []
        
        # For Round 7's marketing spend, observe Net_Change in Round 8
        marketing_spend_r7 = self.df_finance.loc[self.df_finance['Round'] == 7, 'Marketing_Spend'].iloc[0]
        net_change_r8 = self.df_finance.loc[self.df_finance['Round'] == 8, 'Net_Change'].iloc[0]
        
        if marketing_spend_r7 > 0:
            # Multiplier: Net_Change in next round / Marketing_Spend in current round
            earnings_multiplier_r7 = net_change_r8 / marketing_spend_r7
            impact_data_mkt.append({
                'Round': 7,
                'Earnings_Multiplier': earnings_multiplier_r7,
                'Marketing_Spend_k': marketing_spend_r7 / 1000,
                'Next_Round_Earnings_k': net_change_r8 / 1000
            })

        # For Round 9's marketing spend, observe Net_Change in Round 10
        marketing_spend_r9 = self.df_finance.loc[self.df_finance['Round'] == 9, 'Marketing_Spend'].iloc[0]
        net_change_r10 = self.df_finance.loc[self.df_finance['Round'] == 10, 'Net_Change'].iloc[0]

        if marketing_spend_r9 > 0:
            earnings_multiplier_r9 = net_change_r10 / marketing_spend_r9
            impact_data_mkt.append({
                'Round': 9,
                'Earnings_Multiplier': earnings_multiplier_r9,
                'Marketing_Spend_k': marketing_spend_r9 / 1000,
                'Next_Round_Earnings_k': net_change_r10 / 1000
            })
            
        impact_df_mkt = pd.DataFrame(impact_data_mkt)

        # Plotting
        if not impact_df_mkt.empty:
            bars_impact_mkt = ax2_3_mkt.bar(impact_df_mkt['Round'], impact_df_mkt['Earnings_Multiplier'], 
                                             color=self.COLORS['purple'], alpha=0.7)
            ax2_3_mkt.set_xlabel('Runda inwestycji marketingowej') 
            ax2_3_mkt.set_ylabel('Mno≈ºnik zysku (Zysk nast. rundy / Wydatki mkt.)')
            ax2_3_mkt.set_title('Wp≈Çyw wydatk√≥w marketingowych na zyski w nastƒôpnej rundzie')

            ax2_3_mkt.grid(True, axis='y', alpha=0.3)
            ax2_3_mkt.axhline(y=1, color='black', linestyle='--', alpha=0.5, label='Punkt rentowno≈õci (Mno≈ºnik = 1)') # Multiplier of 1 means break-even
            ax2_3_mkt.set_xticks(impact_df_mkt['Round'])
            for bar_item, row_item in zip(bars_impact_mkt, impact_df_mkt.itertuples()):
                height = bar_item.get_height()
                label = f'x{height:.2f}\n(‚Ç¨{row_item.Marketing_Spend_k:.0f}k wydatki\n‚Ç¨{row_item.Next_Round_Earnings_k:+.0f}k zysku w nast. rundzie)'
                ax2_3_mkt.text(bar_item.get_x() + bar_item.get_width()/2,
                               height + (np.sign(height) * 0.1 if height != 0 else 0.1), # Adjust text position slightly
                               label, ha='center', va='bottom' if height >=0 else 'top', fontsize=9)
            ax2_3_mkt.legend(frameon=True)
        else:
            ax2_3_mkt.text(0.5, 0.5, 'Brak danych do analizy wp≈Çywu marketingowego na zyski.', 
                           horizontalalignment='center', verticalalignment='center', 
                           transform=ax2_3_mkt.transAxes, fontsize=12, color='gray')


        # 2.4 Rozk≈Çad kana≈Ç√≥w reklamowych
        ax2_4_mkt = axes_mkt[1, 1]
        ad_channels = ['Direct_Sale_Ads', 'Ambassador_Ads', 'Newspaper_Ads']
        ad_channel_labels = ['Sprzeda≈º bezpo≈õrednia', 'Ambasadorzy', 'Gazety']
        rounds_for_ads_plot = [7, 9, 10]
        ad_data_plot = self.marketing_full[self.marketing_full['Round'].isin(rounds_for_ads_plot)][['Round'] + ad_channels]

        x_ads_plot = np.arange(len(rounds_for_ads_plot))
        width_ads_plot = 0.25
        colors_ads_plot = [self.COLORS['primary'], self.COLORS['success'], self.COLORS['info']]

        for i, channel_col in enumerate(ad_channels):
            ax2_4_mkt.bar(x_ads_plot + i*width_ads_plot - width_ads_plot,
                          ad_data_plot[channel_col], width_ads_plot,
                          label=ad_channel_labels[i], color=colors_ads_plot[i], alpha=0.8)

        ax2_4_mkt.set_xlabel('Runda')
        ax2_4_mkt.set_ylabel('Liczba umieszczonych reklam')
        ax2_4_mkt.set_title('Ewolucja miksu kana≈Ç√≥w reklamowych (kluczowe rundy)')
        ax2_4_mkt.set_xticks(x_ads_plot)
        ax2_4_mkt.set_xticklabels(rounds_for_ads_plot)
        ax2_4_mkt.legend(frameon=True, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.15))
        ax2_4_mkt.grid(True, axis='y', alpha=0.3)

        fig_mkt.suptitle('Analiza strategii marketingowej - Rewolucja efektywno≈õci',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_mkt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_unit_economics(self):
        """Generates plots for unit economics analysis."""
        print("\n\nüí∞ SEKCJA 2.5: EKONOMIKA JEDNOSTKOWA - HISTORIA RENTOWNO≈öCI NA JEDNOSTKƒò")
        print("="*100)

        unit_economics_df = self.df_finance[self.df_finance['Units_Sold'] > 0].copy()

        print("\nEWOLUCJA EKONOMIKI JEDNOSTKOWEJ (na sprzedanƒÖ jednostkƒô):")
        print("-" * 120)
        print(f"{'Runda':<7} {'Sprzedane jednostki':<12} {'Koszt prod./jedn.':<14} {'Koszt mkt./jedn.':<14} {'Ca≈Çkowity koszt/jedn.':<15} {'Przych√≥d/jedn.':<14} {'Zysk/jedn.':<12} {'Mar≈ºa zysku/jedn. (%)':<18}")
        print("-" * 120)
        for _, row in unit_economics_df.iterrows():
            unit_profit_margin = (row['Profit_Per_Unit_Sold'] / row['Revenue_Per_Unit_Sold'] * 100) if row['Revenue_Per_Unit_Sold'] > 0 else 0
            print(f"{row['Round']:<7} {row['Units_Sold']:>10,} "
                  f"‚Ç¨{row['Production_Cost_Per_Unit']:>12.2f} ‚Ç¨{row['Marketing_Cost_Per_Unit_Sold']:>12.2f} "
                  f"‚Ç¨{row['Total_Cost_Per_Unit_Sold']:>13.2f} ‚Ç¨{row['Revenue_Per_Unit_Sold']:>12.2f} "
                  f"‚Ç¨{row['Profit_Per_Unit_Sold']:>10.2f} {unit_profit_margin:>16.1f}%")

        # Figure 2.5.1 (4 plots)
        fig_ue1, axes_ue1 = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 2.5.1 Ewolucja kosztu produkcji na jednostkƒô
        ax2_5_1_ue = axes_ue1[0, 0]
        valid_rounds_ue = unit_economics_df['Round']
        ax2_5_1_ue.plot(valid_rounds_ue, unit_economics_df['Production_Cost_Per_Unit'], 'o-',
                         color=self.COLORS['primary'], linewidth=3, markersize=8, label='Koszt prod./jedn.')
        ax2_5_1_ue.set_xlabel('Runda')
        ax2_5_1_ue.set_ylabel('Koszt produkcji na jednostkƒô (‚Ç¨)')
        ax2_5_1_ue.set_title('Efektywno≈õƒá koszt√≥w produkcji')
        ax2_5_1_ue.grid(True, alpha=0.3)
        ax2_5_1_ue.set_xticks(valid_rounds_ue)

        if not unit_economics_df['Production_Cost_Per_Unit'].empty:
            avg_prod_cost_ue = unit_economics_df['Production_Cost_Per_Unit'].mean()
            ax2_5_1_ue.axhline(y=avg_prod_cost_ue, color=self.COLORS['gray'], linestyle='--', alpha=0.7, label=f'≈örednia: ‚Ç¨{avg_prod_cost_ue:.2f}')
        ax2_5_1_ue.legend(frameon=True)

        # 2.5.2 Koszt marketingowy na sprzedanƒÖ jednostkƒô
        ax2_5_2_ue = axes_ue1[0, 1]
        bars_mkt_cost_ue = ax2_5_2_ue.bar(valid_rounds_ue, unit_economics_df['Marketing_Cost_Per_Unit_Sold'],
                                           color=self.COLORS['warning'], alpha=0.7, label='Koszt mkt./sprzedana jedn.')
        ax2_5_2_ue.set_xlabel('Runda')
        ax2_5_2_ue.set_ylabel('Koszt marketingowy na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_2_ue.set_title('Efektywno≈õƒá koszt√≥w marketingowych (na sprzeda≈º)')
        ax2_5_2_ue.grid(True, axis='y', alpha=0.3)
        ax2_5_2_ue.set_xticks(valid_rounds_ue)
        ax2_5_2_ue.legend(frameon=True)

        if not unit_economics_df[unit_economics_df['Round']==9].empty:
            round_9_mkt_cost = unit_economics_df[unit_economics_df['Round']==9]['Marketing_Cost_Per_Unit_Sold'].iloc[0]
            if pd.notna(round_9_mkt_cost):
                ax2_5_2_ue.annotate(f'R9: ‚Ç¨{round_9_mkt_cost:.2f}\n(Mega Kampania)',
                                     xy=(9, round_9_mkt_cost), xytext=(8.5, round_9_mkt_cost + 20 if round_9_mkt_cost < 100 else round_9_mkt_cost - 20),
                                     arrowprops=dict(arrowstyle='->', color=self.COLORS['danger']),
                                     fontsize=10, ha='center', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec=self.COLORS['danger']))

        # 2.5.3 Ca≈Çkowity koszt vs. przych√≥d na sprzedanƒÖ jednostkƒô
        ax2_5_3_ue = axes_ue1[1, 0]
        x_pos_ue = np.arange(len(valid_rounds_ue))
        width_ue = 0.35

        ax2_5_3_ue.bar(x_pos_ue - width_ue/2, unit_economics_df['Total_Cost_Per_Unit_Sold'],
                         width_ue, label='Ca≈Çkowity koszt/sprzedana jedn.', color=self.COLORS['danger'], alpha=0.7)
        ax2_5_3_ue.bar(x_pos_ue + width_ue/2, unit_economics_df['Revenue_Per_Unit_Sold'],
                         width_ue, label='Przych√≥d/sprzedana jedn.', color=self.COLORS['success'], alpha=0.7)

        ax2_5_3_ue.set_xlabel('Runda')
        ax2_5_3_ue.set_ylabel('Kwota na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_3_ue.set_title('Ekonomika jednostkowa: Koszt vs. Przych√≥d')
        ax2_5_3_ue.set_xticks(x_pos_ue)
        ax2_5_3_ue.set_xticklabels(valid_rounds_ue)
        ax2_5_3_ue.legend(frameon=True, loc='upper left')
        ax2_5_3_ue.grid(True, axis='y', alpha=0.3)

        # 2.5.4 Ewolucja zysku na sprzedanƒÖ jednostkƒô
        ax2_5_4_ue = axes_ue1[1, 1]
        profit_per_unit_sold_ue = unit_economics_df['Profit_Per_Unit_Sold']
        ax2_5_4_ue.plot(valid_rounds_ue, profit_per_unit_sold_ue, 'o-',
                         color=self.COLORS['purple'], linewidth=4, markersize=10, label='Zysk/sprzedana jedn.')
        ax2_5_4_ue.fill_between(valid_rounds_ue, profit_per_unit_sold_ue, 0,
                                 where=(profit_per_unit_sold_ue >= 0),
                                 alpha=0.3, color=self.COLORS['success'], interpolate=True)
        ax2_5_4_ue.fill_between(valid_rounds_ue, profit_per_unit_sold_ue, 0,
                                 where=(profit_per_unit_sold_ue < 0),
                                 alpha=0.3, color=self.COLORS['danger'], interpolate=True)

        ax2_5_4_ue.set_xlabel('Runda')
        ax2_5_4_ue.set_ylabel('Zysk na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_4_ue.set_title('Podr√≥≈º rentowno≈õci na jednostkƒô')
        ax2_5_4_ue.grid(True, alpha=0.3)
        ax2_5_4_ue.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax2_5_4_ue.set_xticks(valid_rounds_ue)
        ax2_5_4_ue.legend(frameon=True)

        if not profit_per_unit_sold_ue.empty:
            best_profit_ue = profit_per_unit_sold_ue.max()
            worst_profit_ue = profit_per_unit_sold_ue.min()
            best_round_ue = unit_economics_df.loc[profit_per_unit_sold_ue.idxmax(), 'Round']
            worst_round_ue = unit_economics_df.loc[profit_per_unit_sold_ue.idxmin(), 'Round']

            ax2_5_4_ue.annotate(f'Najlepszy: ‚Ç¨{best_profit_ue:.2f} (R{best_round_ue})',
                                 xy=(best_round_ue, best_profit_ue), xytext=(best_round_ue, best_profit_ue + 20),
                                 arrowprops=dict(arrowstyle='->', color=self.COLORS['success']),
                                 fontsize=10, ha='center', fontweight='bold')
            if worst_profit_ue < 0:
                ax2_5_4_ue.annotate(f'Najgorszy: ‚Ç¨{worst_profit_ue:.2f} (R{worst_round_ue})',
                                     xy=(worst_round_ue, worst_profit_ue), xytext=(worst_round_ue, worst_profit_ue - 20),
                                     arrowprops=dict(arrowstyle='->', color=self.COLORS['danger']),
                                     fontsize=10, ha='center', fontweight='bold')

        fig_ue1.suptitle('Dog≈Çƒôbna analiza ekonomiki jednostkowej - Czƒô≈õƒá 1: Koszty, przychody i zysk na jednostkƒô',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_ue1.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

        # Figure 2.5.2 (2 plots)
        fig_ue2, axes_ue2 = plt.subplots(1, 2, figsize=(18, 8))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 2.5.5 Rozk≈Çad koszt√≥w jednostkowych (s≈Çupkowy skumulowany)
        ax2_5_5_ue = axes_ue2[0]
        unit_economics_df['HR_Cost_Per_Unit_Sold'] = np.where(unit_economics_df['Units_Sold'] > 0, unit_economics_df['HR_Cost'] / unit_economics_df['Units_Sold'], 0)
        unit_economics_df['Ops_Excl_Prod_Cost_Per_Unit_Sold'] = np.where(unit_economics_df['Units_Sold'] > 0,
                                                                          (unit_economics_df['Operations_Cost'] - unit_economics_df['Production_Cost']) / unit_economics_df['Units_Sold'], 0)

        cost_components_ue = ['Production_Cost_Per_Unit', 'Marketing_Cost_Per_Unit_Sold',
                               'HR_Cost_Per_Unit_Sold', 'Ops_Excl_Prod_Cost_Per_Unit_Sold']
        cost_labels_ue = ['Produkcja', 'Marketing', 'HR', 'Inne operacje']
        colors_stack_ue = [self.COLORS['primary'], self.COLORS['warning'], self.COLORS['success'], self.COLORS['info']]
        bottom_stack_ue = np.zeros(len(valid_rounds_ue))

        for i, component_col in enumerate(cost_components_ue):
            ax2_5_5_ue.bar(valid_rounds_ue, unit_economics_df[component_col], bottom=bottom_stack_ue,
                            label=cost_labels_ue[i], color=colors_stack_ue[i], alpha=0.8)
            bottom_stack_ue += unit_economics_df[component_col].fillna(0)

        ax2_5_5_ue.set_xlabel('Runda')
        ax2_5_5_ue.set_ylabel('Koszt na sprzedanƒÖ jednostkƒô (‚Ç¨)')
        ax2_5_5_ue.set_title('Struktura koszt√≥w jednostkowych')
        ax2_5_5_ue.legend(frameon=True, loc='upper left', ncol=2)
        ax2_5_5_ue.grid(True, axis='y', alpha=0.3)
        ax2_5_5_ue.set_xticks(valid_rounds_ue)

        # 2.5.6 Karta wynik√≥w efektywno≈õci ekonomiki jednostkowej (mapa ciep≈Ça)
        ax2_5_6_ue = axes_ue2[1]
        if not unit_economics_df['Production_Cost_Per_Unit'].empty and unit_economics_df['Production_Cost_Per_Unit'].min() > 0:
            min_prod_cost_ue = unit_economics_df['Production_Cost_Per_Unit'].min()
            unit_economics_df['Prod_Efficiency_Score'] = (min_prod_cost_ue / unit_economics_df['Production_Cost_Per_Unit']) * 100
        else:
            unit_economics_df['Prod_Efficiency_Score'] = 0

        if not unit_economics_df['Marketing_Cost_Per_Unit_Sold'].empty:
            max_mkt_cost_ue = unit_economics_df['Marketing_Cost_Per_Unit_Sold'].max()
            min_mkt_cost_ue = unit_economics_df['Marketing_Cost_Per_Unit_Sold'].min()
            if (max_mkt_cost_ue - min_mkt_cost_ue) > 0:
                unit_economics_df['Mkt_Efficiency_Score'] = ((max_mkt_cost_ue - unit_economics_df['Marketing_Cost_Per_Unit_Sold']) / (max_mkt_cost_ue - min_mkt_cost_ue)) * 100
            else:
                unit_economics_df['Mkt_Efficiency_Score'] = 50
            unit_economics_df['Mkt_Efficiency_Score'] = unit_economics_df['Mkt_Efficiency_Score'].fillna(0).clip(0,100)
        else:
            unit_economics_df['Mkt_Efficiency_Score'] = 0

        unit_economics_df['Unit_Profit_Margin_Score'] = np.where(unit_economics_df['Revenue_Per_Unit_Sold'] > 0,
                                                                  (unit_economics_df['Profit_Per_Unit_Sold'] / unit_economics_df['Revenue_Per_Unit_Sold']) * 100, -100).clip(-100,100)
        if not unit_economics_df['Unit_Profit_Margin_Score'].empty:
            min_margin_score = unit_economics_df['Unit_Profit_Margin_Score'].min()
            max_margin_score = unit_economics_df['Unit_Profit_Margin_Score'].max()
            if (max_margin_score - min_margin_score) > 0:
                unit_economics_df['Unit_Profit_Margin_Heatmap_Score'] = ((unit_economics_df['Unit_Profit_Margin_Score'] - min_margin_score) / (max_margin_score - min_margin_score)) * 100
            else:
                unit_economics_df['Unit_Profit_Margin_Heatmap_Score'] = 50
            unit_economics_df['Unit_Profit_Margin_Heatmap_Score'] = unit_economics_df['Unit_Profit_Margin_Heatmap_Score'].fillna(0).clip(0,100)
        else:
            unit_economics_df['Unit_Profit_Margin_Heatmap_Score'] = 0

        efficiency_matrix_ue = unit_economics_df[['Prod_Efficiency_Score', 'Mkt_Efficiency_Score', 'Unit_Profit_Margin_Heatmap_Score']].values.T
        im_heatmap_ue = ax2_5_6_ue.imshow(efficiency_matrix_ue, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)

        ax2_5_6_ue.set_xticks(np.arange(len(valid_rounds_ue)))
        ax2_5_6_ue.set_yticks(np.arange(3))
        ax2_5_6_ue.set_xticklabels(valid_rounds_ue)
        ax2_5_6_ue.set_yticklabels(['Efektywno≈õƒá produkcji', 'Efektywno≈õƒá marketingu', 'Rentowno≈õƒá jednostkowa'], fontsize=10)
        ax2_5_6_ue.set_title('Karta wynik√≥w efektywno≈õci ekonomiki jednostkowej')

        for i_heat in range(efficiency_matrix_ue.shape[0]):
            for j_heat in range(efficiency_matrix_ue.shape[1]):
                text_val = efficiency_matrix_ue[i_heat, j_heat]
                ax2_5_6_ue.text(j_heat, i_heat, f'{text_val:.0f}',
                                 ha="center", va="center", color="black" if 30 < text_val < 70 else "white", fontsize=9, fontweight='bold')

        cbar_heatmap_ue = plt.colorbar(im_heatmap_ue, ax=ax2_5_6_ue, orientation='vertical', fraction=0.046, pad=0.04)
        cbar_heatmap_ue.set_label('Wynik efektywno≈õci (0-100)')

        fig_ue2.suptitle('Dog≈Çƒôbna analiza ekonomiki jednostkowej - Czƒô≈õƒá 2: Struktura koszt√≥w i karta wynik√≥w',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_ue2.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

        print("\nüîç KLUCZOWE SPOSTRZE≈ªENIA Z EKONOMIKI JEDNOSTKOWEJ:")
        print("-" * 100)
        if not unit_economics_df.empty:
            best_profit_unit = unit_economics_df['Profit_Per_Unit_Sold'].max()
            best_profit_round = unit_economics_df.loc[unit_economics_df['Profit_Per_Unit_Sold'].idxmax(), 'Round']
            avg_profit_unit = unit_economics_df['Profit_Per_Unit_Sold'].mean()
            min_mkt_cost_unit = unit_economics_df['Marketing_Cost_Per_Unit_Sold'].min()
            max_mkt_cost_unit = unit_economics_df['Marketing_Cost_Per_Unit_Sold'].max()
            stable_prod_cost_unit = unit_economics_df['Production_Cost_Per_Unit'].iloc[-3:].mean()
            last_round_ue = unit_economics_df.iloc[-1]

            print(f"‚Ä¢ Najlepszy zysk/jednostkƒô: ‚Ç¨{best_profit_unit:.2f} w Rundzie {best_profit_round}")
            print(f"‚Ä¢ ≈öredni zysk/jednostkƒô (w rundach sprzeda≈ºowych): ‚Ç¨{avg_profit_unit:.2f}")
            print(f"‚Ä¢ Zakres koszt√≥w marketingowych/sprzedanƒÖ jednostkƒô: od ‚Ç¨{min_mkt_cost_unit:.2f} do ‚Ç¨{max_mkt_cost_unit:.2f}")
            print(f"‚Ä¢ Koszt produkcji/jednostkƒô ustabilizowa≈Ç siƒô wok√≥≈Ç: ‚Ç¨{stable_prod_cost_unit:.2f} (p√≥≈∫niejsze rundy)")
            print(f"‚Ä¢ Ekonomika jednostkowa w ostatniej rundzie (R{last_round_ue['Round']}): "
                  f"Przych√≥d/jedn. ‚Ç¨{last_round_ue['Revenue_Per_Unit_Sold']:.2f}, "
                  f"Ca≈Çkowity koszt/jedn. ‚Ç¨{last_round_ue['Total_Cost_Per_Unit_Sold']:.2f}, "
                  f"Zysk/jedn. ‚Ç¨{last_round_ue['Profit_Per_Unit_Sold']:.2f}")
        else:
            print("‚Ä¢ Brak danych sprzeda≈ºowych do obliczenia ekonomiki jednostkowej.")

    def _plot_operational_excellence(self):
        """Generates plots for operational excellence analysis."""
        print("\n\n‚öôÔ∏è SEKCJA 3: DOSKONA≈ÅO≈öƒÜ OPERACYJNA - DROGA DO 200%")
        print("="*100)

        print("\nKAMIENIE MILOWE EFEKTYWNO≈öCI OPERACYJNEJ:")
        print("-" * 100)
        print(f"{'Runda':<8} {'Efektywno≈õƒá linii Fash (%)':<20} {'Efektywno≈õƒá linii Geeks (%)':<20} {'Pojemno≈õƒá (jednostki)':<18} {'Produkcja (jednostki)':<20} {'Wska≈∫nik sprzeda≈ºy (%)':<15} {'Jednostki/prac. prod.':<15}")
        print("-" * 100)
        for _, row in self.operations_data[self.operations_data['Round'] >= 2].iterrows():
            sales_rate_ops = (row['Units_Sold'] / row['Units_Produced'] * 100) if row['Units_Produced'] > 0 else 0
            print(f"{row['Round']:<8} {row['Fashionistas_Eff']:>18.1f} {row['Geeks_Eff']:>18.1f} "
                  f"{row['Production_Capacity']:>16,} {row['Units_Produced']:>18,} {sales_rate_ops:>14.1f} "
                  f"{row['Units_Per_Production_Worker']:>14.1f}")

        print("\nüèÜ OSIƒÑGNIƒòCIA W ZAKRESIE EFEKTYWNO≈öCI:")
        print("-" * 100)
        print(f"‚Ä¢ Runda 9: Pierwsza linia (Fashionistas) osiƒÖgnƒô≈Ça 200% efektywno≈õci.")
        print(f"‚Ä¢ Runda 10: Obie linie produkcyjne osiƒÖgnƒô≈Çy teoretyczne maksimum 200% efektywno≈õci.")
        print(f"‚Ä¢ Ko≈Ñcowa zdolno≈õƒá produkcyjna: {self.operations_data['Production_Capacity'].iloc[-1]:,} jednostek (zoptymalizowana ze szczytu {self.operations_data['Production_Capacity'].max():,}).")
        if not self.operations_data[self.operations_data['Round']==2].empty and self.operations_data[self.operations_data['Round']==2]['Units_Per_Production_Worker'].iloc[0] > 0 :
            prod_worker_increase = (self.operations_data[self.operations_data['Round']==10]['Units_Per_Production_Worker'].iloc[0] / self.operations_data[self.operations_data['Round']==2]['Units_Per_Production_Worker'].iloc[0] -1) *100
            print(f"‚Ä¢ Wydajno≈õƒá pracownik√≥w (jednostki/pracownik produkcji): Wzros≈Ça z {self.operations_data[self.operations_data['Round']==2]['Units_Per_Production_Worker'].iloc[0]:.1f} (R2) do {self.operations_data[self.operations_data['Round']==10]['Units_Per_Production_Worker'].iloc[0]:.1f} (R10), ~{prod_worker_increase:.0f}% poprawy.")
        else:
            print(f"‚Ä¢ Wydajno≈õƒá pracownik√≥w (jednostki/pracownik produkcji): OsiƒÖgnƒô≈Ça {self.operations_data[self.operations_data['Round']==10]['Units_Per_Production_Worker'].iloc[0]:.1f} (R10).")

        # Figure 3.1 (4 plots)
        fig_ops1, axes_ops1 = plt.subplots(2, 2, figsize=(18, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 3.1 Ewolucja efektywno≈õci do perfekcji
        ax3_1_ops = axes_ops1[0, 0]
        rounds_ops_plot = self.operations_data['Round'][self.operations_data['Round'] >= 2]
        fash_eff_plot = self.operations_data['Fashionistas_Eff'][self.operations_data['Round'] >= 2]
        geeks_eff_plot = self.operations_data['Geeks_Eff'][self.operations_data['Round'] >= 2]

        ax3_1_ops.plot(rounds_ops_plot, fash_eff_plot, 'o-', color=self.COLORS['primary'], linewidth=3,
                         markersize=8, label='Efektywno≈õƒá linii Fashionistas')
        ax3_1_ops.plot(rounds_ops_plot, geeks_eff_plot, 's-', color=self.COLORS['secondary'], linewidth=3,
                         markersize=8, label='Efektywno≈õƒá linii Geeks')
        ax3_1_ops.axhline(y=200, color=self.COLORS['success'], linestyle='--', linewidth=2,
                          alpha=0.7, label='200% Maks. efektywno≈õƒá')
        ax3_1_ops.fill_between(rounds_ops_plot, fash_eff_plot, 200, where=(fash_eff_plot >= 180),
                                 alpha=0.2, color=self.COLORS['success'], interpolate=True)
        ax3_1_ops.fill_between(rounds_ops_plot, geeks_eff_plot, 200, where=(geeks_eff_plot >= 180),
                                 alpha=0.2, color=self.COLORS['success'], interpolate=True)

        ax3_1_ops.set_xlabel('Runda')
        ax3_1_ops.set_ylabel('Efektywno≈õƒá linii (%)')
        ax3_1_ops.set_title('Podr√≥≈º do doskona≈Ço≈õci operacyjnej (Efektywno≈õƒá linii)')
        ax3_1_ops.grid(True, alpha=0.3)
        ax3_1_ops.set_ylim(90, 210)
        ax3_1_ops.set_xticks(rounds_ops_plot)

        ax3_1_ops.scatter([9], [200], s=250, color=self.COLORS['success'], marker='*',
                          zorder=5, label='Fashionistas osiƒÖga 200% (R9)', edgecolors='black')
        ax3_1_ops.scatter([10], [200], s=350, color=self.COLORS['purple'], marker='*',
                          zorder=5, label='Obie linie na 200% (R10)', edgecolors='black')
        handles, labels = ax3_1_ops.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        ax3_1_ops.legend(by_label.values(), by_label.keys(), frameon=True, loc='lower right')

        # 3.2 Wykorzystanie zdolno≈õci produkcyjnych
        ax3_2_ops = axes_ops1[0, 1]
        utilization_ops = (self.operations_data['Units_Produced'] / self.operations_data['Production_Capacity'] * 100)
        utilization_ops_plot = utilization_ops[self.operations_data['Round'] >= 2]

        bars_util_ops = ax3_2_ops.bar(rounds_ops_plot, utilization_ops_plot, color=self.COLORS['info'], alpha=0.7)
        ax3_2_ops.set_xlabel('Runda')
        ax3_2_ops.set_ylabel('Wykorzystanie zdolno≈õci (%)')
        ax3_2_ops.set_title('Wykorzystanie zdolno≈õci produkcyjnych')
        ax3_2_ops.grid(True, axis='y', alpha=0.3)
        ax3_2_ops.axhline(y=100, color='red', linestyle='--', alpha=0.5, label='100% wykorzystania')
        ax3_2_ops.set_xticks(rounds_ops_plot)
        ax3_2_ops.legend(frameon=True)

        for bar_item, util_val in zip(bars_util_ops, utilization_ops_plot):
            if util_val > 95:
                bar_item.set_color(self.COLORS['danger'])
            elif util_val > 80:
                bar_item.set_color(self.COLORS['warning'])
            else:
                bar_item.set_color(self.COLORS['info'])
            ax3_2_ops.text(bar_item.get_x() + bar_item.get_width()/2, util_val + 1,
                            f'{util_val:.0f}%', ha='center', va='bottom', fontsize=9)

        # 3.3 Rozw√≥j i efektywno≈õƒá sieci sprzeda≈ºy
        ax3_3_ops = axes_ops1[1, 0]
        offices_plot = self.operations_data['Sales_Offices'][self.operations_data['Round'] >= 2]
        office_eff_plot = self.operations_data['Office_Efficiency_Avg'][self.operations_data['Round'] >= 2]

        ax3_3_ops_twin = ax3_3_ops.twinx()
        bars_offices_ops = ax3_3_ops.bar(rounds_ops_plot, offices_plot, alpha=0.6, color=self.COLORS['purple'], label='Biura sprzeda≈ºy (#)')
        line_office_eff_ops = ax3_3_ops_twin.plot(rounds_ops_plot, office_eff_plot, 'go-', linewidth=3, markersize=8,
                                                     label='≈örednia efektywno≈õƒá biura (%)')

        ax3_3_ops.set_xlabel('Runda')
        ax3_3_ops.set_ylabel('Liczba biur sprzeda≈ºy', color=self.COLORS['purple'])
        ax3_3_ops_twin.set_ylabel('≈örednia efektywno≈õƒá biura (%)', color='green')
        ax3_3_ops.set_title('Rozw√≥j i efektywno≈õƒá sieci sprzeda≈ºy')
        ax3_3_ops.grid(True, alpha=0.3, axis='y', which='major', linestyle='dotted')
        ax3_3_ops_twin.grid(True, alpha=0.3, axis='y', which='major', linestyle=':')
        ax3_3_ops.set_xticks(rounds_ops_plot)

        lines_sales_net = list(bars_offices_ops[0:1]) + line_office_eff_ops
        labels_sales_net = [l.get_label() for l in lines_sales_net]
        ax3_3_ops.legend(lines_sales_net, labels_sales_net, loc='upper left', frameon=True)

        expansion_round_series = self.operations_data[self.operations_data['Sales_Offices'] > self.operations_data['Sales_Offices'].shift(1).fillna(0)]['Round']
        if not expansion_round_series.empty:
            expansion_round = expansion_round_series.iloc[0]
            if pd.notna(expansion_round) and expansion_round in rounds_ops_plot.values:
                ax3_3_ops.axvline(x=expansion_round, color='red', linestyle='--', alpha=0.7)
                ax3_3_ops.text(expansion_round + 0.1, offices_plot.max() * 0.8, f'Ekspansja\ndo {offices_plot[rounds_ops_plot == expansion_round].iloc[0]} biur',
                                 ha='left', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", fc="white", ec='red'))

        # 3.4 Katastrofa dystrybucyjna (fokus na rundy 4-6)
        ax3_4_ops = axes_ops1[1, 1]
        disaster_rounds_ops = [4, 5, 6]
        produced_disaster_ops = self.operations_data[self.operations_data['Round'].isin(disaster_rounds_ops)]['Units_Produced'].values
        sold_disaster_ops = self.operations_data[self.operations_data['Round'].isin(disaster_rounds_ops)]['Units_Sold'].values

        if produced_disaster_ops.any() > 0:
            success_rate_disaster_ops = np.where(produced_disaster_ops > 0, (sold_disaster_ops / produced_disaster_ops * 100), 0)
        else:
            success_rate_disaster_ops = np.array([0.0] * len(produced_disaster_ops))

        x_disaster_plot = np.arange(len(disaster_rounds_ops))
        width_disaster_plot = 0.35
        ax3_4_ops.bar(x_disaster_plot - width_disaster_plot/2, produced_disaster_ops, width_disaster_plot,
                      label='Wyprodukowane jednostki', color=self.COLORS['primary'], alpha=0.7)
        ax3_4_ops.bar(x_disaster_plot + width_disaster_plot/2, sold_disaster_ops, width_disaster_plot,
                      label='Sprzedane jednostki', color=self.COLORS['success'], alpha=0.7)

        for i_dis, (p_dis, s_dis, rate_dis) in enumerate(zip(produced_disaster_ops, sold_disaster_ops, success_rate_disaster_ops)):
            ax3_4_ops.text(i_dis, max(p_dis, s_dis) + 500, f'{rate_dis:.1f}% sprzedanych', ha='center',
                            fontweight='bold', fontsize=10, color=self.COLORS['danger'] if rate_dis < 10 else self.COLORS['info'])

        ax3_4_ops.set_xlabel('Runda')
        ax3_4_ops.set_ylabel('Liczba jednostek')
        ax3_4_ops.set_title('Katastrofa dystrybucyjna (Rundy 4-6)')
        ax3_4_ops.set_xticks(x_disaster_plot)
        ax3_4_ops.set_xticklabels(disaster_rounds_ops)
        ax3_4_ops.legend(frameon=True)
        ax3_4_ops.grid(True, axis='y', alpha=0.3)

        if 5 in disaster_rounds_ops:
            disaster_round_idx = disaster_rounds_ops.index(5)
            max_produced_val = produced_disaster_ops.max() if produced_disaster_ops.size > 0 else 1
            ax3_4_ops.add_patch(plt.Rectangle((x_disaster_plot[disaster_round_idx] - width_disaster_plot, 0),
                                             width_disaster_plot*2, max_produced_val * 1.05,
                                             fill=True, color='red', alpha=0.15, zorder=0))
            ax3_4_ops.text(x_disaster_plot[disaster_round_idx], max_produced_val * 0.5,
                            f'KATASTROFA!\nTylko {success_rate_disaster_ops[disaster_round_idx]:.1f}% sprzedanych (R5)',
                            ha='center', fontsize=12, fontweight='bold', color='red',
                            bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="red"))

        fig_ops1.suptitle('Analiza doskona≈Ço≈õci operacyjnej - Czƒô≈õƒá 1: Efektywno≈õƒá linii i wykorzystanie',
                           fontsize=20, fontweight='bold', y=0.98)
        fig_ops1.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

        # Figure 3.2 (2 plots)
        fig_ops2, axes_ops2 = plt.subplots(1, 2, figsize=(18, 8))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 3.5 Optymalizacja si≈Çy roboczej: Ca≈Çkowita liczba pracownik√≥w vs. ≈õrednia efektywno≈õƒá linii
        ax3_5_ops = axes_ops2[0]
        total_workers_ops = (self.operations_data['Production_Workers'] + self.operations_data['Sales_Workers'])[self.operations_data['Round'] >= 2]
        avg_line_eff_ops = ((self.operations_data['Fashionistas_Eff'] + self.operations_data['Geeks_Eff']) / 2)[self.operations_data['Round'] >= 2]

        ax3_5_ops_twin = ax3_5_ops.twinx()
        line1 = ax3_5_ops.plot(rounds_ops_plot, total_workers_ops, 'o-', color=self.COLORS['primary'], linewidth=3,
                                 markersize=8, label='Ca≈Çkowita liczba pracownik√≥w')
        line2 = ax3_5_ops_twin.plot(rounds_ops_plot, avg_line_eff_ops, 's-', color=self.COLORS['success'], linewidth=3,
                                     markersize=8, label='≈örednia efektywno≈õƒá linii (%)')

        ax3_5_ops.set_xlabel('Runda')
        ax3_5_ops.set_ylabel('Ca≈Çkowita liczba pracownik√≥w', color=self.COLORS['primary'])
        ax3_5_ops_twin.set_ylabel('≈örednia efektywno≈õƒá linii (%)', color=self.COLORS['success'])
        ax3_5_ops.set_title('Optymalizacja si≈Çy roboczej: Robiƒá wiƒôcej mniejszymi ≈õrodkami')
        ax3_5_ops.grid(True, alpha=0.3, axis='y', linestyle='dotted')
        ax3_5_ops_twin.grid(True, alpha=0.3, axis='y', linestyle=':')
        ax3_5_ops.set_xticks(rounds_ops_plot)

        lines_workforce = line1 + line2
        labels_workforce = [l.get_label() for l in lines_workforce]
        ax3_5_ops.legend(lines_workforce, labels_workforce, loc='center right', frameon=True)
        ax3_5_ops.axvspan(8, 10, alpha=0.1, color='green', label='Faza optymalizacji')

        # 3.6 ROI inwestycji w szkolenia
        ax3_6_ops = axes_ops2[1]
        training_inv_ops = self.operations_data['Training_Investment'][self.operations_data['Round'] >= 2] / 1000
        avg_line_eff_ops_series = avg_line_eff_ops.reset_index(drop=True)
        training_inv_ops_series = training_inv_ops.reset_index(drop=True)

        training_roi_ops = []
        rounds_for_roi_plot = []

        for i in range(len(training_inv_ops_series) - 1):
            current_round_num = rounds_ops_plot.iloc[i]
            eff_gain_next_round = avg_line_eff_ops_series.iloc[i+1] - avg_line_eff_ops_series.iloc[i]
            current_training_spend = training_inv_ops_series.iloc[i]

            if current_training_spend > 0:
                roi_calc_ops = eff_gain_next_round / current_training_spend
                training_roi_ops.append(roi_calc_ops)
                rounds_for_roi_plot.append(current_round_num)

        if training_roi_ops:
            min_len = min(len(training_inv_ops_series), len(training_roi_ops), len(avg_line_eff_ops_series), len(rounds_for_roi_plot))
            if min_len > 0:
                spend_for_roi_rounds = training_inv_ops_series.iloc[:min_len]
                training_roi_ops_filtered = training_roi_ops[:min_len]
                eff_for_roi_rounds = avg_line_eff_ops_series.iloc[:min_len]
                rounds_for_roi_plot_filtered = rounds_for_roi_plot[:min_len]

                scatter_training_roi = ax3_6_ops.scatter(spend_for_roi_rounds,
                                                         training_roi_ops_filtered,
                                                         s=eff_for_roi_rounds * 2,
                                                         c=rounds_for_roi_plot_filtered,
                                                         cmap='viridis', alpha=0.7, edgecolors='black')

                ax3_6_ops.set_xlabel('Inwestycja w szkolenia w rundzie (tys. ‚Ç¨)')
                ax3_6_ops.set_ylabel('Wzrost efektywno≈õci w nast. rundzie na ‚Ç¨1k szkolenia')
                ax3_6_ops.set_title('Zwroty z inwestycji w szkolenia (ROI)')
                ax3_6_ops.grid(True, alpha=0.3)
                ax3_6_ops.axhline(y=0, color='black', linestyle='--', alpha=0.5)

                cbar_training_roi = plt.colorbar(scatter_training_roi, ax=ax3_6_ops)
                cbar_training_roi.set_label('Runda inwestycji w szkolenia')

                if training_roi_ops_filtered:
                    max_roi_val = max(training_roi_ops_filtered)
                    max_roi_idx = training_roi_ops_filtered.index(max_roi_val)
                    round_of_max_roi_spend = rounds_for_roi_plot_filtered[max_roi_idx]
                    spend_at_max_roi = spend_for_roi_rounds.iloc[max_roi_idx]

                    ax3_6_ops.annotate(f'Najlepsze ROI:\n{max_roi_val:.2f} pkt/‚Ç¨k\n(szkolenie R{round_of_max_roi_spend})',
                                         xy=(spend_at_max_roi, max_roi_val),
                                         xytext=(spend_at_max_roi + 10 if spend_at_max_roi < 80 else spend_at_max_roi - 30 , max_roi_val + (0.1 if max_roi_val < 1 else 0.5) ),
                                         arrowprops=dict(arrowstyle='->', color='red', lw=1.5),
                                         fontsize=9, fontweight='bold', bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red"))
                else:
                    ax3_6_ops.text(0.5, 0.5, 'Problem z wyr√≥wnaniem danych dla wykresu ROI szkole≈Ñ', ha='center', va='center', fontsize=10)
            else:
                ax3_6_ops.text(0.5, 0.5, 'Brak wystarczajƒÖcych danych dla wykresu ROI szkole≈Ñ', ha='center', va='center', fontsize=12)
        else:
            ax3_6_ops.text(0.5, 0.5, 'Brak wystarczajƒÖcych danych dla wykresu ROI szkole≈Ñ', ha='center', va='center', fontsize=12)

        fig_ops2.suptitle('Analiza doskona≈Ço≈õci operacyjnej - Czƒô≈õƒá 2: Optymalizacja zasob√≥w',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_ops2.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_competitive_dynamics(self):
        """Generates plots for competitive dynamics and market positioning."""
        print("\n\nüèÅ SEKCJA 4: DYNAMIKA KONKURENCYJNA - STRATEGIA POZYCJONOWANIA PREMIUM")
        print("="*100)

        print("\nEWOLUCJA POZYCJI NA RYNKU:")
        print("-" * 120)
        print(f"{'Runda':<7} {'Nasz udzia≈Ç (%)':<15} {'Ranking':<6} {'Udzia≈Ç lidera (%)':<25} {'R√≥≈ºnica do lidera (%)':<20} {'Faza strategii':<30}")
        print("-" * 120)
        for _, row in self.market_data[self.market_data['Round'] >= 3].iterrows():
            leader_share_val = row['market_leader_share']
            leader_col_name = self.market_data[self.competitor_columns_plot].idxmax(axis=1)[row.name] if leader_share_val > 0 else None
            leader_display_name = self.plot_teams_map.get(leader_col_name, ('N/A', ''))[0] if leader_col_name else "N/A"
            gap_to_leader_val = row['V4EScooters_share'] - leader_share_val

            phase_strat = ""
            if row['Round'] <= 3: phase_strat = "Wej≈õcie na rynek i wczesne prowadzenie"
            elif row['Round'] <= 5: phase_strat = "Ekspansja i zmienno≈õƒá/kryzys"
            elif row['Round'] <= 7: phase_strat = "Repozycjonowanie premium i odbudowa"
            else: phase_strat = "Optymalizacja zysku i stabilno≈õƒá"

            print(f"{row['Round']:<7} {row['V4EScooters_share']:>13.2f} {int(row['V4EScooters_rank']):>4} "
                  f"{leader_share_val:>8.2f} ({leader_display_name:<15}) {gap_to_leader_val:>+18.2f}  {phase_strat:<30}")

        fig_comp, axes_comp = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 4.1 Wy≈õcig o udzia≈Ç w rynku
        ax4_1_comp = axes_comp[0, 0]
        rounds_comp_plot_data = self.market_data[self.market_data['Round'] >= 3]['Round']
        for col_name, (plot_label, color_val) in self.plot_teams_map.items():
            data_plot = self.market_data[self.market_data['Round'] >= 3][col_name]
            linewidth = 4 if col_name == 'V4EScooters_share' else 2.5
            marker = 'o-' if col_name == 'V4EScooters_share' else 's--'
            ax4_1_comp.plot(rounds_comp_plot_data, data_plot, marker, linewidth=linewidth, markersize=8 if col_name == 'V4EScooters_share' else 6,
                            label=plot_label, color=color_val, alpha=0.9 if col_name == 'V4EScooters_share' else 0.7)

        ax4_1_comp.set_xlabel('Runda')
        ax4_1_comp.set_ylabel('Udzia≈Ç w rynku (%)')
        ax4_1_comp.set_title('Ewolucja udzia≈Çu w rynku - Wy≈õcig konkurencyjny')
        ax4_1_comp.legend(frameon=True, loc='upper left', ncol=2)
        ax4_1_comp.grid(True, alpha=0.3)
        ax4_1_comp.set_xticks(rounds_comp_plot_data)

        ax4_1_comp.axvspan(4.5, 5.5, alpha=0.1, color=self.COLORS['danger'], label='Punkt kryzysowy V4E')
        ax4_1_comp.axvspan(6.5, 7.5, alpha=0.1, color=self.COLORS['success'], label='Zwrot V4E na premium')

        # 4.2 Macierz strategicznego pozycjonowania
        ax4_2_comp = axes_comp[0, 1]
        final_shares_comp = [self.market_data[col].iloc[-1] for col in self.competitor_columns_plot]
        v4e_final_profit_per_unit = self.df_finance[self.df_finance['Round']==10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 0
        v4e_final_rev_per_unit = self.df_finance[self.df_finance['Round']==10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 1
        v4e_final_margin_est = (v4e_final_profit_per_unit / v4e_final_rev_per_unit * 100) if v4e_final_rev_per_unit > 0 else 0

        estimated_margins_comp = [
            v4e_final_margin_est,
            15.0,
            25.0,
            -10.0
        ]
        companies_labels_comp = [
            f"{self.plot_teams_map['team_5'][0]}\n(Zysk Premium)",
            f"{self.plot_teams_map['KKKompany'][0]}\n(Lider Wolumenu)",
            f"{self.plot_teams_map['V4EScooters_share'][0]}\n(Zr√≥wnowa≈ºone podej≈õcie)",
            f"{self.plot_teams_map['team_8'][0]}\n(Wyj≈õcie z rynku)"
        ]
        colors_comp_plot = [self.plot_teams_map[col][1] for col in self.competitor_columns_plot]
        bubble_sizes_comp = [1200, 1000, 800, 400]

        ax4_2_comp.scatter(final_shares_comp, estimated_margins_comp, s=bubble_sizes_comp,
                           c=colors_comp_plot, alpha=0.75, edgecolors='black', linewidth=1.5)

        for i, company_label in enumerate(companies_labels_comp):
            ax4_2_comp.annotate(company_label, (final_shares_comp[i], estimated_margins_comp[i]),
                                 textcoords="offset points", xytext=(0,10 if bubble_sizes_comp[i]>500 else -15),
                                 ha='center', fontsize=9, fontweight='bold')

        ax4_2_comp.set_xlabel('Ko≈Ñcowy udzia≈Ç w rynku (%)')
        ax4_2_comp.set_ylabel('Szacowana mar≈ºa zysku (%)')
        ax4_2_comp.set_title('Ko≈Ñcowe pozycjonowanie strategiczne - Jako≈õƒá vs. Ilo≈õƒá')
        ax4_2_comp.grid(True, alpha=0.3, linestyle='--')
        ax4_2_comp.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        ax4_2_comp.text(self.market_data['V4EScooters_share'].iloc[-1], v4e_final_margin_est + 5, 'Nisza zysku', ha='center', fontsize=10, style='italic', color=self.COLORS['purple'])
        ax4_2_comp.text(self.market_data['KKKompany'].iloc[-1], 15 - 5, 'Fokus na wolumen', ha='center', fontsize=10, style='italic', color=self.COLORS['secondary'])

        # 4.3 Ewolucja rankingu rynkowego V4EScooters
        ax4_3_comp = axes_comp[1, 0]
        ranks_v4e_plot = self.market_data[self.market_data['Round'] >= 3]['V4EScooters_rank']
        rounds_rank_plot = self.market_data[self.market_data['Round'] >= 3]['Round']

        max_rank = len(self.competitor_columns_plot)
        plot_rank_values = max_rank - ranks_v4e_plot + 1

        colors_rank_plot = ['gold' if r == 1 else 'silver' if r == 2 else '#CD7F32' if r == 3 else 'lightgray'
                            for r in ranks_v4e_plot]
        bars_rank_plot = ax4_3_comp.bar(rounds_rank_plot, plot_rank_values, color=colors_rank_plot, alpha=0.85, edgecolor='black')

        ax4_3_comp.set_xlabel('Runda')
        ax4_3_comp.set_ylabel('Pozycja na rynku')
        ax4_3_comp.set_title(f"Ewolucja rankingu rynkowego {self.plot_teams_map['V4EScooters_share'][0]}")
        ax4_3_comp.set_yticks(np.arange(1, max_rank + 1))
        ax4_3_comp.set_yticklabels([f'{max_rank-i+1}. miejsce' for i in np.arange(1, max_rank + 1)])
        ax4_3_comp.grid(True, axis='y', alpha=0.3)
        ax4_3_comp.set_xticks(rounds_rank_plot)

        for bar_rank_item, rank_val_item in zip(bars_rank_plot, ranks_v4e_plot):
            ax4_3_comp.text(bar_rank_item.get_x() + bar_rank_item.get_width()/2,
                             bar_rank_item.get_height() - 0.2, f'{int(rank_val_item)}{". miejsce" if int(rank_val_item) in [1,2,3] else ". miejsce"}',
                             ha='center', va='top', fontweight='bold', color='black', fontsize=10)

        # 4.4 Dynamika rynku: Zmienno≈õƒá i udzia≈Ç lidera
        ax4_4_comp = axes_comp[1, 1]
        market_volatility_plot = self.market_data[self.market_data['Round'] >= 3][self.competitor_columns_plot].std(axis=1)
        leader_share_plot = self.market_data[self.market_data['Round'] >= 3]['market_leader_share']

        ax4_4_comp_twin = ax4_4_comp.twinx()
        line1 = ax4_4_comp.plot(rounds_comp_plot_data, market_volatility_plot, 'o-', color=self.COLORS['primary'],
                                 linewidth=3, markersize=8, label='Zmienno≈õƒá udzia≈Çu w rynku (odch. std.)')
        line2 = ax4_4_comp_twin.plot(rounds_comp_plot_data, leader_share_plot, 's--', color=self.COLORS['danger'],
                                      linewidth=3, markersize=8, label="Udzia≈Ç lidera w rynku (%)")

        ax4_4_comp.set_xlabel('Runda')
        ax4_4_comp.set_ylabel('Zmienno≈õƒá udzia≈Çu w rynku (odch. std.)', color=self.COLORS['primary'])
        ax4_4_comp_twin.set_ylabel("Udzia≈Ç lidera w rynku (%)", color=self.COLORS['danger'])
        ax4_4_comp.set_title('Dynamika rynku - Zmienno≈õƒá i koncentracja')
        ax4_4_comp.grid(True, alpha=0.3, axis='y', linestyle='dotted')
        ax4_4_comp_twin.grid(True, alpha=0.3, axis='y', linestyle=':')
        ax4_4_comp.set_xticks(rounds_comp_plot_data)

        lines_dyn_plot_combined = line1 + line2
        labels_dyn_plot_combined = [l.get_label() for l in lines_dyn_plot_combined]
        ax4_4_comp.legend(lines_dyn_plot_combined, labels_dyn_plot_combined, loc='upper center', frameon=True, ncol=1, bbox_to_anchor=(0.5, -0.15))
        ax4_4_comp.axvline(x=5, color='gray', linestyle='--', alpha=0.7, label='WstrzƒÖs na rynku (R5)')

        fig_comp.suptitle('Dynamika konkurencyjna i pozycjonowanie rynkowe V4EScooters',
                           fontsize=20, fontweight='bold', y=0.98)
        fig_comp.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_product_strategy(self):
        """Generates plots for product strategy and market segmentation."""
        print("\n\nüì¶ SEKCJA 5: STRATEGIA PRODUKTOWA - CELOWANIE W DOSKONA≈ÅO≈öƒÜ")
        print("="*100)

        print("\nV4EScooters MIKSU PRODUKTOWEGO I EWOLUCJI CEN:")
        print("-" * 120)
        print(f"{'Runda':<7} {'Fashionistas Sprzedane%':<12} {'Geeks Sprzedane%':<13} {'Savers Sprzedane%':<14} {'≈örednia cena (‚Ç¨)':<15} {'Faza strategii produktowej':<35}")
        print("-" * 120)
        for _, row in self.product_data_v4e.iterrows():
            total_sold = row['Total_Sold_V4E']
            fash_pct_sold = (row['Fashionistas_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0
            geeks_pct_sold = (row['Geeks_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0
            savers_pct_sold = (row['Savers_Sold_V4E'] / total_sold * 100) if total_sold > 0 else 0

            prod_strat_phase = ""
            if row['Round'] <= 4: prod_strat_phase = "PoczƒÖtkowy miks i wej≈õcie na rynek"
            elif row['Round'] <= 6: prod_strat_phase = "Dostosowanie i testowanie kryzysowe"
            else: prod_strat_phase = "Fokus na portfolio premium"

            print(f"{row['Round']:<7} {fash_pct_sold:>10.1f} {geeks_pct_sold:>11.1f} {savers_pct_sold:>12.1f} "
                  f"{row['Avg_Selling_Price_V4E']:>14.2f} {prod_strat_phase:<35}")

        fig_prod, axes_prod = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.4, wspace=0.3)

        # 5.1 Ewolucja miksu produktowego V4EScooters
        ax5_1_prod = axes_prod[0, 0]
        rounds_prod_plot_v4e = self.product_data_v4e['Round']
        width_prod_plot = 0.8

        fash_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0, self.product_data_v4e['Fashionistas_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)
        geeks_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0, self.product_data_v4e['Geeks_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)
        savers_pct_plot = np.where(self.product_data_v4e['Total_Sold_V4E'] > 0, self.product_data_v4e['Savers_Sold_V4E'] / self.product_data_v4e['Total_Sold_V4E'] * 100, 0)

        ax5_1_prod.bar(rounds_prod_plot_v4e, fash_pct_plot, width_prod_plot, label='Fashionistas (% sprzedanych)', color=self.COLORS['purple'])
        ax5_1_prod.bar(rounds_prod_plot_v4e, geeks_pct_plot, width_prod_plot, bottom=fash_pct_plot, label='Geeks/Ekolodzy (% sprzedanych)', color=self.COLORS['success'])
        ax5_1_prod.bar(rounds_prod_plot_v4e, savers_pct_plot, width_prod_plot, bottom=fash_pct_plot+geeks_pct_plot, label='Oszczƒôdni (% sprzedanych)', color=self.COLORS['warning'])

        ax5_1_prod.set_xlabel('Runda')
        ax5_1_prod.set_ylabel('Miks produktowy (% sprzedanych jednostek V4E)')
        ax5_1_prod.set_title('Ewolucja portfolio produkt√≥w V4EScooters (Miks sprzeda≈ºy)')
        ax5_1_prod.legend(frameon=True, loc='upper center', ncol=3, bbox_to_anchor=(0.5, -0.12))
        ax5_1_prod.grid(True, axis='y', alpha=0.3)
        ax5_1_prod.set_xticks(rounds_prod_plot_v4e)
        ax5_1_prod.set_ylim(0, 100)

        # 5.2 Ewolucja strategii cenowej V4EScooters
        ax5_2_prod = axes_prod[0, 1]
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Fashionistas_Price_V4E'], 'o-', linewidth=3,
                         label='Cena Fashionistas', color=self.COLORS['purple'], markersize=8)
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Geeks_Price_V4E'], 's-', linewidth=3,
                         label='Cena Geeks/Ekolodzy', color=self.COLORS['success'], markersize=8)
        ax5_2_prod.plot(rounds_prod_plot_v4e, self.product_data_v4e['Savers_Price_V4E'], '^-', linewidth=3,
                         label='Cena Oszczƒôdni', color=self.COLORS['warning'], markersize=8)

        ax5_2_prod.set_xlabel('Runda')
        ax5_2_prod.set_ylabel('Cena za jednostkƒô (‚Ç¨)')
        ax5_2_prod.set_title('Strategia cenowa V4EScooters wed≈Çug segmentu')
        ax5_2_prod.legend(frameon=True, loc='upper left')
        ax5_2_prod.grid(True, alpha=0.3)
        ax5_2_prod.set_xticks(rounds_prod_plot_v4e)

        ax5_2_prod.axvline(x=6.5, color='red', linestyle='--', alpha=0.7, linewidth=2)
        ax5_2_prod.text(6.3, self.product_data_v4e['Savers_Price_V4E'].min() + 20, 'Zmiana cen\nPremium\n(Po kryzysie)',
                         ha='right', fontsize=10, fontweight='bold',
                         bbox=dict(boxstyle="round,pad=0.5", fc="yellow", alpha=0.7))

        # 5.3 Radar wydajno≈õci produktu V4EScooters (R10)
        ax5_3_prod = fig_prod.add_subplot(2, 2, 3, projection='polar')
        categories_radar_prod = ['Wolumen sprzeda≈ºy (% ca≈Çkowitego V4E)', 'Poziom ceny (‚Ç¨)', 'Wska≈∫nik sukcesu sprzeda≈ºy (%)',
                                 'Szacowany wk≈Çad w zysk (%)', 'Wynik dopasowania strategicznego (1-5)']
        num_vars_radar = len(categories_radar_prod)
        angles_radar = np.linspace(0, 2 * np.pi, num_vars_radar, endpoint=False).tolist()
        angles_radar += angles_radar[:1]

        r10_data_prod = self.product_data_v4e[self.product_data_v4e['Round'] == 10].iloc[0]
        total_sold_r10_v4e = r10_data_prod['Total_Sold_V4E']

        fash_vol_r10 = (r10_data_prod['Fashionistas_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        fash_price_r10 = r10_data_prod['Fashionistas_Price_V4E']
        fash_success_r10 = (r10_data_prod['Fashionistas_Sold_V4E'] / r10_data_prod['Fashionistas_Produced_V4E'] * 100) if r10_data_prod['Fashionistas_Produced_V4E'] > 0 else 0
        fash_profit_contr_est_r10 = 40
        fash_strat_fit_r10 = 5
        fash_scores_r10 = [fash_vol_r10, fash_price_r10, fash_success_r10, fash_profit_contr_est_r10, fash_strat_fit_r10*20]
        fash_scores_r10 += fash_scores_r10[:1]

        geeks_vol_r10 = (r10_data_prod['Geeks_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        geeks_price_r10 = r10_data_prod['Geeks_Price_V4E']
        geeks_success_r10 = (r10_data_prod['Geeks_Sold_V4E'] / r10_data_prod['Geeks_Produced_V4E'] * 100) if r10_data_prod['Geeks_Produced_V4E'] > 0 else 0
        geeks_profit_contr_est_r10 = 45
        geeks_strat_fit_r10 = 5
        geeks_scores_r10 = [geeks_vol_r10, geeks_price_r10, geeks_success_r10, geeks_profit_contr_est_r10, geeks_strat_fit_r10*20]
        geeks_scores_r10 += geeks_scores_r10[:1]

        savers_vol_r10 = (r10_data_prod['Savers_Sold_V4E'] / total_sold_r10_v4e * 100) if total_sold_r10_v4e > 0 else 0
        savers_price_r10 = r10_data_prod['Savers_Price_V4E']
        savers_success_r10 = (r10_data_prod['Savers_Sold_V4E'] / r10_data_prod['Savers_Produced_V4E'] * 100) if r10_data_prod['Savers_Produced_V4E'] > 0 else 0
        savers_profit_contr_est_r10 = 15
        savers_strat_fit_r10 = 3
        savers_scores_r10 = [savers_vol_r10, savers_price_r10, savers_success_r10, savers_profit_contr_est_r10, savers_strat_fit_r10*20]
        savers_scores_r10 += savers_scores_r10[:1]

        ax5_3_prod.plot(angles_radar, fash_scores_r10, 'o-', linewidth=2, label='Fashionistas (R10)', color=self.COLORS['purple'], markersize=6)
        ax5_3_prod.fill(angles_radar, fash_scores_r10, self.COLORS['purple'], alpha=0.25)
        ax5_3_prod.plot(angles_radar, geeks_scores_r10, 's-', linewidth=2, label='Geeks (R10)', color=self.COLORS['success'], markersize=6)
        ax5_3_prod.fill(angles_radar, geeks_scores_r10, self.COLORS['success'], alpha=0.25)
        ax5_3_prod.plot(angles_radar, savers_scores_r10, '^-', linewidth=2, label='Oszczƒôdni (R10)', color=self.COLORS['warning'], markersize=6)
        ax5_3_prod.fill(angles_radar, savers_scores_r10, self.COLORS['warning'], alpha=0.25)

        ax5_3_prod.set_xticks(angles_radar[:-1])
        ax5_3_prod.set_xticklabels(categories_radar_prod, fontsize=9)
        ax5_3_prod.set_yticks(np.arange(0, 301, 50))
        ax5_3_prod.set_title('Profil wydajno≈õci produktu V4E (R10 - Skale ilustracyjne)', pad=20)
        ax5_3_prod.legend(loc='upper right', bbox_to_anchor=(1.4, 1.1), fontsize=10)

        # 5.4 Wk≈Çad przychod√≥w V4EScooters wed≈Çug produktu
        ax5_4_prod = axes_prod[1, 1]
        revenue_rounds_prod_plot = [7, 8, 9, 10]
        revenue_data_list_prod = []

        for r_rev_prod in revenue_rounds_prod_plot:
            row_rev_prod = self.product_data_v4e[self.product_data_v4e['Round'] == r_rev_prod].iloc[0]
            fash_rev_prod = row_rev_prod['Fashionistas_Sold_V4E'] * row_rev_prod['Fashionistas_Price_V4E'] / 1000
            geeks_rev_prod = row_rev_prod['Geeks_Sold_V4E'] * row_rev_prod['Geeks_Price_V4E'] / 1000
            savers_rev_prod = row_rev_prod['Savers_Sold_V4E'] * row_rev_prod['Savers_Price_V4E'] / 1000
            revenue_data_list_prod.append([fash_rev_prod, geeks_rev_prod, savers_rev_prod])

        revenue_data_np_prod = np.array(revenue_data_list_prod).T
        x_rev_prod_plot = np.arange(len(revenue_rounds_prod_plot))
        width_rev_prod_plot = 0.25

        product_names_rev_plot = ['Przych√≥d Fashionistas', 'Przych√≥d Geeks/Eko', 'Przych√≥d Oszczƒôdni']
        colors_rev_prod_plot = [self.COLORS['purple'], self.COLORS['success'], self.COLORS['warning']]

        for i, (product_rev_item, color_rev_item) in enumerate(zip(product_names_rev_plot, colors_rev_prod_plot)):
            ax5_4_prod.bar(x_rev_prod_plot + i*width_rev_prod_plot - width_rev_prod_plot, revenue_data_np_prod[i],
                            width_rev_prod_plot, label=product_rev_item, color=color_rev_item, alpha=0.85)

        ax5_4_prod.set_xlabel('Runda (Faza Premium)')
        ax5_4_prod.set_ylabel('Przych√≥d (tys. ‚Ç¨)')
        ax5_4_prod.set_title('Wk≈Çad przychod√≥w produkt√≥w V4EScooters')
        ax5_4_prod.set_xticks(x_rev_prod_plot)
        ax5_4_prod.set_xticklabels(revenue_rounds_prod_plot)
        ax5_4_prod.legend(frameon=True, loc='upper left')
        ax5_4_prod.grid(True, axis='y', alpha=0.3)

        fig_prod.suptitle('Analiza strategii produktowej V4EScooters - Wydajno≈õƒá portfolio premium',
                           fontsize=20, fontweight='bold', y=0.98)
        fig_prod.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_geographical_strategy(self):
        """Generates plots for geographical market strategy."""
        print("\n\nüåç SEKCJA 6: EKSPANSJA GEOGRAFICZNA - Opanowanie rynku")
        print("="*100)

        print("\nWYNIKI V4EScooters NA RYNKACH GEOGRAFICZNYCH (Fokus na fazƒô premium R7-R10):")
        print("-" * 130)
        print(f"{'Miasto':<12} {'≈örednia skuteczno≈õƒá V4E (%)':<22} {'Cena referencyjna rynku (‚Ç¨)':<22} {'Rozmiar rynku':<15} {'Konkurencja':<15} {'Fokus strategii V4E':<25}")
        print("-" * 130)
        for _, row in self.df_geo_v4e.iterrows():
            v4e_strat_focus = "Celuj w wysoki potencja≈Ç" if row['Avg_Success_V4E_R7_R10'] > 85 and row['Competition_Level'] != 'Wysoka' else \
                              "Gra na niszy premium" if row['Avg_Success_V4E_R7_R10'] > 75 and row['Competition_Level'] == 'Wysoka' else \
                              "Rynek rozwojowy"
            print(f"{row['City']:<12} {row['Avg_Success_V4E_R7_R10']:>20.1f} {row['Ref_Avg_Price_Market']:>20} "
                  f"{row['Market_Size_Category']:<15} {row['Competition_Level']:<15} {v4e_strat_focus:<25}")

        fig_geo, axes_geo = plt.subplots(2, 2, figsize=(20, 14))
        plt.subplots_adjust(hspace=0.45, wspace=0.3)

        # 6.1 Sukces rynkowy V4EScooters wed≈Çug miasta (R7-R10)
        ax6_1_geo = axes_geo[0, 0]
        cities_geo_plot = self.df_geo_v4e['City']
        x_geo_indices = np.arange(len(cities_geo_plot))
        width_geo_bar = 0.20

        rounds_geo_success_cols = ['R7_Success_V4E', 'R8_Success_V4E', 'R9_Success_V4E', 'R10_Success_V4E']
        round_labels_geo = ['Sukces R7', 'Sukces R8', 'Sukces R9', 'Sukces R10']
        colors_geo_rounds = [self.COLORS['primary'], self.COLORS['success'], self.COLORS['warning'], self.COLORS['purple']]

        for i, (col_name, label) in enumerate(zip(rounds_geo_success_cols, round_labels_geo)):
            ax6_1_geo.bar(x_geo_indices + i*width_geo_bar - 1.5*width_geo_bar, self.df_geo_v4e[col_name],
                          width_geo_bar, label=label, color=colors_geo_rounds[i], alpha=0.85)

        ax6_1_geo.set_xlabel('Miasto')
        ax6_1_geo.set_ylabel('Wska≈∫nik sukcesu sprzeda≈ºy V4EScooters (%)')
        ax6_1_geo.set_title('Wyniki geograficzne V4EScooters wed≈Çug rundy (R7-R10)')
        ax6_1_geo.set_xticks(x_geo_indices)
        ax6_1_geo.set_xticklabels(cities_geo_plot, rotation=45, ha="right")
        ax6_1_geo.legend(frameon=True, loc='upper center', ncol=2, bbox_to_anchor=(0.5, -0.25))
        ax6_1_geo.grid(True, axis='y', alpha=0.3)
        ax6_1_geo.axhline(y=100, color='red', linestyle='--', alpha=0.5, label='Cel 100%')
        ax6_1_geo.set_ylim(0, 110)

        # 6.2 Mapa ciep≈Ça wydajno≈õci miasta V4EScooters
        ax6_2_geo = axes_geo[0, 1]
        performance_matrix_geo = self.df_geo_v4e[rounds_geo_success_cols].values.T
        im_geo_heatmap = ax6_2_geo.imshow(performance_matrix_geo, cmap='RdYlGn', aspect='auto', vmin=0, vmax=100)

        ax6_2_geo.set_xticks(np.arange(len(cities_geo_plot)))
        ax6_2_geo.set_yticks(np.arange(len(round_labels_geo)))
        ax6_2_geo.set_xticklabels(cities_geo_plot, rotation=45, ha="right")
        ax6_2_geo.set_yticklabels(round_labels_geo)
        ax6_2_geo.set_title('Mapa ciep≈Ça wydajno≈õci miasta V4EScooters (R7-R10 Sukces %)')

        for i_heat_geo in range(performance_matrix_geo.shape[0]):
            for j_heat_geo in range(performance_matrix_geo.shape[1]):
                text_val_geo = performance_matrix_geo[i_heat_geo, j_heat_geo]
                ax6_2_geo.text(j_heat_geo, i_heat_geo, f'{text_val_geo:.0f}%',
                                 ha="center", va="center", color="black" if 30 < text_val_geo < 70 else "white",
                                 fontsize=9, fontweight='bold')

        cbar_geo_heatmap = plt.colorbar(im_geo_heatmap, ax=ax6_2_geo, orientation='vertical', fraction=0.046, pad=0.04)
        cbar_geo_heatmap.set_label('Wska≈∫nik sukcesu sprzeda≈ºy V4E (%)')

        # 6.3 Mapa atrakcyjno≈õci rynku vs. wydajno≈õci V4EScooters
        ax6_3_geo = axes_geo[1, 0]
        market_size_numeric_map = {'Ma≈Çy': 1, '≈öredni': 2, 'Du≈ºy': 3}
        self.df_geo_v4e['Market_Size_Num'] = self.df_geo_v4e['Market_Size_Category'].map(market_size_numeric_map)
        competition_numeric_map = {'Wysoka': 1, '≈örednia': 2, 'Niska': 3}
        self.df_geo_v4e['Competition_Attractive_Num'] = self.df_geo_v4e['Competition_Level'].map(competition_numeric_map)

        scatter_market_map = ax6_3_geo.scatter(
            self.df_geo_v4e['Market_Size_Num'],
            self.df_geo_v4e['Competition_Attractive_Num'],
            s=self.df_geo_v4e['Avg_Success_V4E_R7_R10'] * 8,
            c=self.df_geo_v4e['Ref_Avg_Price_Market'],
            cmap='coolwarm', alpha=0.75, edgecolors='black', linewidth=1.5
        )

        for i, city_label in enumerate(self.df_geo_v4e['City']):
            ax6_3_geo.annotate(city_label,
                                 (self.df_geo_v4e['Market_Size_Num'].iloc[i], self.df_geo_v4e['Competition_Attractive_Num'].iloc[i]),
                                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)

        ax6_3_geo.set_xlabel('Kategoria rozmiaru rynku (Wiƒôkszy ‚Üí)')
        ax6_3_geo.set_ylabel('Atrakcyjno≈õƒá rynku (Ni≈ºsza konkurencja ‚Üí)')
        ax6_3_geo.set_title('Pozycjonowanie rynkowe V4E: Atrakcyjno≈õƒá vs. Wydajno≈õƒá')
        ax6_3_geo.set_xticks([1, 2, 3])
        ax6_3_geo.set_xticklabels(['Ma≈Çy', '≈öredni', 'Du≈ºy'])
        ax6_3_geo.set_yticks([1, 2, 3])
        ax6_3_geo.set_yticklabels(['Wysoka konk.', '≈örednia konk.', 'Niska konk.'])
        ax6_3_geo.grid(True, alpha=0.3, linestyle='--')

        cbar_market_map_price = plt.colorbar(scatter_market_map, ax=ax6_3_geo)
        cbar_market_map_price.set_label('Referencyjna ≈õrednia cena rynkowa (‚Ç¨)')
        for success_level in [70, 85, 95]:
            ax6_3_geo.scatter([], [], s=success_level*8, label=f'{success_level}% Sukcesu', color='gray', alpha=0.7)
        ax6_3_geo.legend(scatterpoints=1, frameon=True, title='Rozmiar bƒÖbelka (≈örednia skuteczno≈õƒá V4E R7-10)', loc='lower right')

        # 6.4 Strategiczne grupy rynkowe V4EScooters
        ax6_4_geo = axes_geo[1, 1]
        self.df_geo_v4e['Strategic_Group'] = 'RozwijajƒÖcy siƒô'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 85) & (self.df_geo_v4e['Competition_Attractive_Num'] >= 2), 'Strategic_Group'] = 'Gwiazdy wydajno≈õci'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 75) & (self.df_geo_v4e['Avg_Success_V4E_R7_R10'] < 85) & (self.df_geo_v4e['Competition_Attractive_Num'] >= 2), 'Strategic_Group'] = 'Solidny wzrost'
        self.df_geo_v4e.loc[(self.df_geo_v4e['Avg_Success_V4E_R7_R10'] >= 70) & (self.df_geo_v4e['Competition_Attractive_Num'] == 1), 'Strategic_Group'] = 'Trudna nisza'

        group_counts = self.df_geo_v4e['Strategic_Group'].value_counts()
        group_colors_map = {'Gwiazdy wydajno≈õci': self.COLORS['success'], 'Solidny wzrost': self.COLORS['info'],
                            'Trudna nisza': self.COLORS['warning'], 'RozwijajƒÖcy siƒô': self.COLORS['danger']}

        bars_strat_groups = ax6_4_geo.bar(group_counts.index, group_counts.values,
                                           color=[group_colors_map.get(g, 'gray') for g in group_counts.index], alpha=0.85)

        ax6_4_geo.set_xlabel('Strategiczna grupa rynkowa dla V4EScooters')
        ax6_4_geo.set_ylabel('Liczba miast w grupie')
        ax6_4_geo.set_title('Strategiczna segmentacja rynkowa V4EScooters (fokus R7-R10)')
        ax6_4_geo.tick_params(axis='x', rotation=15)
        ax6_4_geo.grid(True, axis='y', alpha=0.3)
        for bar_group in bars_strat_groups:
            yval = bar_group.get_height()
            ax6_4_geo.text(bar_group.get_x() + bar_group.get_width()/2, yval + 0.05, int(yval), ha='center', va='bottom', fontsize=10)

        fig_geo.suptitle('Strategia rynkowa geograficzna V4EScooters - OsiƒÖgniƒôcie regionalnego mistrzostwa',
                          fontsize=20, fontweight='bold', y=0.98)
        fig_geo.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _plot_strategic_lessons(self):
        """Generates plots for key strategic lessons and success factors."""
        print("\n\nüéì SEKCJA 7: LEKCJE STRATEGICZNE - KSIƒòGA SUKCESU (V4EScooters)")
        print("="*100)

        final_balance_v4e = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi_v4e = (final_balance_v4e - self.initial_investment) / self.initial_investment * 100
        final_market_share_v4e = self.market_data['V4EScooters_share'].iloc[-1]
        final_rank = int(self.market_data['V4EScooters_rank'].iloc[-1])
        avg_office_eff_final = self.operations_data['Office_Efficiency_Avg'].iloc[-1]
        capacity_util_r10 = (self.operations_data['Units_Produced'].iloc[-1] / self.operations_data['Production_Capacity'].iloc[-1] * 100) if self.operations_data['Production_Capacity'].iloc[-1] > 0 else 0
        avg_price_premium_phase = self.product_data_v4e[self.product_data_v4e['Round']>=7]['Avg_Selling_Price_V4E'].mean()
        sales_success_r10_v4e = self.marketing_full[self.marketing_full['Round']==10]['Sales_Success_Rate'].iloc[0]
        v4e_final_profit_per_unit = self.df_finance[self.df_finance['Round']==10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 0
        v4e_final_rev_per_unit = self.df_finance[self.df_finance['Round']==10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 1
        v4e_final_margin_est = (v4e_final_profit_per_unit / v4e_final_rev_per_unit * 100) if v4e_final_rev_per_unit > 0 else 0

        summary_metrics_v4e = {
            'FINANSE': {
                'Inwestycja poczƒÖtkowa': f'‚Ç¨{self.initial_investment/1e6:.1f}M',
                'Ko≈Ñcowe saldo': f'‚Ç¨{final_balance_v4e/1e6:.2f}M',
                'Ca≈Çkowity ROI': f'{total_roi_v4e:.2f}%',
                'Mar≈ºa zysku (szac. jedn. R10)': f'{v4e_final_margin_est:.1f}%',
                'Odbudowa po kryzysie': 'Szybka (2 rundy od dna do silnego zysku)'
            },
            'OPERACJE': {
                'Ko≈Ñcowa efektywno≈õƒá linii': '200% (linie Fashionistas i Geeks)',
                'Efektywno≈õƒá biura sprzeda≈ºy (R10)': f'{avg_office_eff_final:.1f}% ({self.operations_data["Sales_Offices"].iloc[-1]} biur)',
                'Zdolno≈õƒá produkcyjna (R10)': f'{self.operations_data["Production_Capacity"].iloc[-1]:,} jednostek',
                'Wykorzystanie zdolno≈õci (R10)': f"{capacity_util_r10:.1f}%",
                'Optymalizacja si≈Çy roboczej': 'ZnaczƒÖce wzrosty produktywno≈õci przy zoptymalizowanej liczbie pracownik√≥w'
            },
            'RYNEK I PRODUKT': {
                'Ko≈Ñcowy udzia≈Ç w rynku': f"{final_market_share_v4e:.2f}% ({final_rank}{'. miejsce' if final_rank in [1,2,3] else '. miejsce'})",
                '≈örednia cena sprzeda≈ºy (faza premium R7-10)': f"‚Ç¨{avg_price_premium_phase:.2f}",
                'Wska≈∫nik sukcesu sprzeda≈ºy (R10)': f"{sales_success_r10_v4e:.1f}%",
                'Strategia produktowa': 'Udane przej≈õcie na zr√≥wnowa≈ºone portfolio premium',
                'Si≈Ça geograficzna': 'Wysoki sukces w kluczowych miastach, takich jak Sofia, Budapeszt, Warszawa (R10)'
            },
            'ZWINNO≈öƒÜ STRATEGICZNA': {
                'Odkrycie efektywno≈õci marketingowej': 'Bud≈ºet ‚Ç¨0 (R8) przyni√≥s≈Ç 92.7% sukcesu sprzeda≈ºy',
                'ZarzƒÖdzanie kryzysowe': 'Skutecznie zarzƒÖdzano katastrofƒÖ dystrybucyjnƒÖ w R5',
                'Zdolno≈õƒá adaptacji': 'Zmieniono strategiƒô po kryzysie, aby skupiƒá siƒô na zysku zamiast wolumenu',
                'Kultura uczenia siƒô': 'Widoczna z test√≥w marketingowych i ulepsze≈Ñ operacyjnych'
            }
        }

        print("\nüèÜ KLUCZOWE WSKA≈πNIKI SUKCESU I OSIƒÑGNIƒòCIA V4EScooters:")
        for category, metrics in summary_metrics_v4e.items():
            print(f"\n{category.upper()}:")
            print("-" * 60)
            for metric_name, value in metrics.items():
                print(f"  ‚Ä¢ {metric_name:<40} {value}")

        # Figure 7.1 (4 plots)
        fig_lessons1 = plt.figure(figsize=(22, 16))
        fig_lessons1.suptitle('Lekcje strategiczne i czynniki sukcesu V4EScooters - Czƒô≈õƒá 1: O≈õ czasu i filary',
                              fontsize=24, fontweight='bold', y=0.97)

        gs_new1 = fig_lessons1.add_gridspec(2, 2, hspace=0.5, wspace=0.3, top=0.9, bottom=0.05, left=0.05, right=0.95)

        # 7.1 O≈õ czasu strategiczna dla V4EScooters
        ax7_1_lsn = fig_lessons1.add_subplot(gs_new1[0, :])
        timeline_events = [
            (1, 'Fundamenty', 'PoczƒÖtkowe ustawienia i infrastruktura', self.COLORS['info']),
            (3, 'Wej≈õcie na rynek', 'Silna poczƒÖtkowa sprzeda≈º (58% udzia≈Çu)', self.COLORS['success']),
            (5, 'Punkt kryzysowy', 'Katastrofa dystrybucyjna, saldo spada do -‚Ç¨144k', self.COLORS['danger']),
            (6, 'PoczƒÖtek odbudowy', 'Zwrot strategiczny, fokus na operacje', self.COLORS['warning']),
            (7, 'Transformacja', 'Wdro≈ºono ceny premium, zysk ro≈õnie', self.COLORS['purple']),
            (8, 'Test marketingowy', 'Bud≈ºet ‚Ç¨0 przynosi 92.7% sukcesu sprzeda≈ºy!', self.COLORS['success']),
            (9, 'Szczyt operacyjny', 'Linia Fashionistas osiƒÖga 200% efektywno≈õci', self.COLORS['primary']),
            (10, 'Opanowanie zysku', 'Ko≈Ñcowe saldo ‚Ç¨5.98M, obie linie 200%', self.COLORS['purple'])
        ]

        ax7_1_lsn.plot([t[0] for t in timeline_events], [0]*len(timeline_events), '-o',
                       color='gray', markersize=12, markerfacecolor='white', zorder=1, linewidth=2)

        level = 0.6
        for i, (round_num, phase, event, color) in enumerate(timeline_events):
            ax7_1_lsn.scatter(round_num, 0, s=700, c=color, zorder=5, alpha=0.8, edgecolors='black', linewidth=1.5)
            current_level = level if i % 2 == 0 else -level
            ax7_1_lsn.annotate(f'R{round_num}:\n{phase}\n{event}',
                               (round_num, 0),
                               xytext=(round_num, current_level),
                               ha='center', va='center', fontsize=9, arrowprops=dict(arrowstyle='-', connectionstyle="arc3,rad=0.15", color='gray'),
                               bbox=dict(boxstyle="round,pad=0.4", fc=color, alpha=0.2, ec=color))

        ax7_1_lsn.set_xlim(0.5, 10.5)
        ax7_1_lsn.set_ylim(-1.2, 1.2)
        ax7_1_lsn.set_xlabel('Runda gry')
        ax7_1_lsn.set_title('O≈õ czasu strategicznej podr√≥≈ºy V4EScooters', pad=15)
        ax7_1_lsn.set_yticks([])
        ax7_1_lsn.grid(True, axis='x', alpha=0.3, linestyle='--')

        # 7.2 Piramida czynnik√≥w sukcesu dla V4EScooters
        ax7_2_lsn = fig_lessons1.add_subplot(gs_new1[1, 0])
        pyramid_levels = [
            ('Strategia Premium oparta na zysku', 4, self.COLORS['purple']),
            ('Doskona≈Ço≈õƒá Operacyjna (200% Efektywno≈õci)', 3.5, self.COLORS['primary']),
            ('Zwinny Marketing i Eksperymentowanie', 3, self.COLORS['success']),
            ('Odporno≈õƒá Finansowa i ZarzƒÖdzanie Kryzysowe', 2.5, self.COLORS['warning'])
        ]
        ax7_2_lsn.set_title('Piramida czynnik√≥w sukcesu V4EScooters')
        ax7_2_lsn.axis('off')
        base_width = 4
        height_step = 1
        current_y = 0.5
        for label, width, color in reversed(pyramid_levels):
            points = [
                [(base_width - width) / 2, current_y],
                [(base_width + width) / 2, current_y],
                [(base_width + width*0.8) / 2, current_y + height_step],
                [(base_width - width*0.8) / 2, current_y + height_step]
            ]
            polygon = plt.Polygon(points, closed=True, facecolor=color, alpha=0.8, edgecolor='black')
            ax7_2_lsn.add_patch(polygon)
            ax7_2_lsn.text(base_width / 2, current_y + height_step / 2, label.replace('(200% Efektywno≈õci)','\n(200% Efektywno≈õci)'),
                            ha='center', va='center', fontsize=9, fontweight='bold', color='white', wrap=True)
            current_y += height_step
        ax7_2_lsn.set_xlim(0, base_width)
        ax7_2_lsn.set_ylim(0, current_y + height_step*0.5)

        # 7.3 Macierz wp≈Çywu decyzji strategicznych dla V4EScooters
        ax7_3_lsn = fig_lessons1.add_subplot(gs_new1[1, 1])
        decisions = {
            'Zmiana cen premium (R7)': {'Impact': 9, 'Execution': 9, 'Timing': 8},
            'Test marketingowy ‚Ç¨0 (R8)': {'Impact': 8, 'Execution': 10, 'Timing': 9},
            'Optymalizacja si≈Çy roboczej (R8-10)': {'Impact': 9, 'Execution': 8, 'Timing': 7},
            'Ekspansja biur sprzeda≈ºy (R5)': {'Impact': 6, 'Execution': 6, 'Timing': 5},
            'Fokus na efektywno≈õƒá linii': {'Impact': 10, 'Execution': 9, 'Timing': 10}
        }
        decision_names = list(decisions.keys())
        impact_scores = [decisions[d]['Impact'] for d in decision_names]
        execution_scores = [decisions[d]['Execution'] for d in decision_names]
        timing_scores_bubble = [decisions[d]['Timing'] * 30 for d in decision_names]

        ax7_3_lsn.scatter(impact_scores, execution_scores, s=timing_scores_bubble,
                           c=[self.COLORS['success'], self.COLORS['info'], self.COLORS['purple'], self.COLORS['warning'], self.COLORS['primary']],
                           alpha=0.7, edgecolors='black', linewidth=1.5)
        for i, name in enumerate(decision_names):
            ax7_3_lsn.annotate(name.replace(' (R7)','\n(R7)').replace(' (R8)','\n(R8)').replace(' (R8-10)','\n(R8-10)').replace(' (R5)','\n(R5)'),
                                 (impact_scores[i], execution_scores[i]),
                                 textcoords="offset points", xytext=(0,5 if timing_scores_bubble[i]>150 else -10),
                                 ha='center', fontsize=8, wrap=True)

        ax7_3_lsn.set_xlabel('Wp≈Çyw strategiczny (1-10)')
        ax7_3_lsn.set_ylabel('Jako≈õƒá wykonania (1-10)')
        ax7_3_lsn.set_title('Kluczowe decyzje strategiczne V4EScooters')
        ax7_3_lsn.grid(True, alpha=0.3, linestyle='--')
        ax7_3_lsn.set_xlim(4.5, 10.5)
        ax7_3_lsn.set_ylim(4.5, 10.5)
        ax7_3_lsn.axhline(y=7.5, color='gray', linestyle=':', alpha=0.5)
        ax7_3_lsn.axvline(x=7.5, color='gray', linestyle=':', alpha=0.5)
        for timing_val in [5, 7, 10]:
            ax7_3_lsn.scatter([], [], s=timing_val*30, label=f'Wynik czasu {timing_val}', color='gray', alpha=0.7)
        ax7_3_lsn.legend(scatterpoints=1, frameon=True, title='Rozmiar bƒÖbelka (Czas)', loc='lower right', fontsize=8)

        fig_lessons1.tight_layout(rect=[0.02, 0.02, 0.98, 0.93])
        plt.show()

        # Figure 7.2 (2 plots)
        fig_lessons2 = plt.figure(figsize=(22, 10))
        fig_lessons2.suptitle('Lekcje strategiczne i czynniki sukcesu V4EScooters - Czƒô≈õƒá 2: Przewagi i wyniki',
                              fontsize=24, fontweight='bold', y=0.97)

        gs_new2 = fig_lessons2.add_gridspec(1, 2, hspace=0.5, wspace=0.3, top=0.9, bottom=0.05, left=0.05, right=0.95)

        # 7.4 Radar przewagi konkurencyjnej dla V4EScooters
        ax7_4_lsn = fig_lessons2.add_subplot(gs_new2[0, 0], projection='polar')
        advantages = ['Maks. efektywno≈õƒá linii', 'Ceny premium', 'Jako≈õƒá produktu (postrzegana)',
                      'Zdolno≈õƒá adaptacji', 'Uczenie siƒô rynku', 'Odbudowa po kryzysie']
        scores_adv = [100, 90, 95, 95, 90, 95]

        angles_adv = np.linspace(0, 2*np.pi, len(advantages), endpoint=False).tolist()
        scores_adv_plot = scores_adv + scores_adv[:1]
        angles_adv_plot = angles_adv + angles_adv[:1]

        ax7_4_lsn.plot(angles_adv_plot, scores_adv_plot, 'o-', linewidth=3, color=self.COLORS['purple'], markersize=8, label='Si≈Ça V4E')
        ax7_4_lsn.fill(angles_adv_plot, scores_adv_plot, self.COLORS['purple'], alpha=0.3)
        ax7_4_lsn.set_xticks(angles_adv)
        ax7_4_lsn.set_xticklabels([adv.replace(' (postrzegana)','\n(postrzegana)') for adv in advantages], fontsize=9)
        ax7_4_lsn.set_yticks(np.arange(0, 101, 20))
        ax7_4_lsn.set_title('Przewagi konkurencyjne V4EScooters', pad=20)
        ax7_4_lsn.legend(loc='upper right', bbox_to_anchor=(1.3, 1.15))

        # 7.5 Ko≈Ñcowa karta wynik√≥w strategicznych dla V4EScooters
        ax7_5_lsn = fig_lessons2.add_subplot(gs_new2[0, 1])
        ax7_5_lsn.axis('off')
        scorecard_data = [
            ['Obszar wska≈∫nika', 'Wynik (szac.)', 'Ocena'],
            ['Wyniki finansowe', '95/100', 'A+'],
            ['Doskona≈Ço≈õƒá operacyjna', '100/100', 'A+'],
            ['Pozycjonowanie na rynku (fokus na zysk)', '90/100', 'A'],
            ['Zdolno≈õƒá adaptacji strategicznej i uczenie siƒô', '98/100', 'A+'],
            ['ZarzƒÖdzanie kryzysowe i odporno≈õƒá', '92/100', 'A'],
            ['Og√≥lna wydajno≈õƒá w grze', '95/100', 'A+']
        ]
        table_score = ax7_5_lsn.table(cellText=scorecard_data, cellLoc='center', loc='center',
                                       colWidths=[0.5, 0.25, 0.25])
        table_score.auto_set_font_size(False)
        table_score.set_fontsize(11)
        table_score.scale(1, 2.5)
        for i_sc in range(3):
            table_score[(0, i_sc)].set_facecolor(self.COLORS['primary'])
            table_score[(0, i_sc)].set_text_props(weight='bold', color='white')
        grade_colors_map = {'A+': '#2ecc71', 'A': '#27ae60', 'A-': '#f1c40f'}
        for i_sc in range(1, len(scorecard_data)):
            grade = scorecard_data[i_sc][2]
            table_score[(i_sc, 2)].set_facecolor(grade_colors_map.get(grade, 'lightgray'))
            table_score[(i_sc, 2)].set_text_props(weight='bold', color='white' if grade in ['A+', 'A'] else 'black')
        ax7_5_lsn.set_title('Ko≈Ñcowa karta wynik√≥w strategicznych V4EScooters', pad=15)

        fig_lessons2.tight_layout(rect=[0.02, 0.02, 0.98, 0.93])
        plt.show()

        # Figure 7.3 (Efficiency-Profitability Journey)
        fig_lessons3 = plt.figure(figsize=(12, 8))
        fig_lessons3.suptitle('Lekcje strategiczne i czynniki sukcesu V4EScooters - Czƒô≈õƒá 3: Podr√≥≈º Efektywno≈õƒá-Rentowno≈õƒá',
                              fontsize=20, fontweight='bold', y=0.98)
        ax7_6_lsn = fig_lessons3.add_subplot(1, 1, 1)

        avg_line_eff_journey = ((self.operations_data['Fashionistas_Eff'] + self.operations_data['Geeks_Eff']) / 2).fillna(0)
        profit_margin_journey = self.df_finance['Profit_Margin'].fillna(-10)
        rounds_journey = self.df_finance['Round']

        for i_jrn in range(len(rounds_journey)-1):
            ax7_6_lsn.annotate('', xy=(avg_line_eff_journey.iloc[i_jrn+1], profit_margin_journey.iloc[i_jrn+1]),
                               xytext=(avg_line_eff_journey.iloc[i_jrn], profit_margin_journey.iloc[i_jrn]),
                               arrowprops=dict(arrowstyle='->', color=self.COLORS['gray'], lw=2, alpha=0.7, connectionstyle="arc3,rad=0.1"))

        scatter_journey = ax7_6_lsn.scatter(avg_line_eff_journey, profit_margin_journey,
                                             c=rounds_journey, cmap='viridis_r', s=self.df_finance['Ending_Balance']/20000 + 50,
                                             edgecolors='black', linewidth=1.5, zorder=5, alpha=0.8)
        for i_jrn, r_jrn in enumerate(rounds_journey):
            ax7_6_lsn.annotate(f'R{r_jrn}', (avg_line_eff_journey.iloc[i_jrn], profit_margin_journey.iloc[i_jrn]),
                                 ha='center', va='center', fontweight='bold', fontsize=8, color='white')

        ax7_6_lsn.set_xlabel('≈örednia efektywno≈õƒá operacyjna linii (%)')
        ax7_6_lsn.set_ylabel('Ca≈Çkowita mar≈ºa zysku (%)')
        ax7_6_lsn.set_title('Podr√≥≈º V4EScooters: Efektywno≈õƒá vs. Rentowno≈õƒá')
        ax7_6_lsn.grid(True, alpha=0.3, linestyle='--')
        ax7_6_lsn.set_xlim(0, 210)
        ax7_6_lsn.set_ylim(min(profit_margin_journey.min(), -20)-5, profit_margin_journey.max()+10)
        ax7_6_lsn.axhline(y=0, color='black', linestyle='-', alpha=0.5)
        cbar_journey = plt.colorbar(scatter_journey, ax=ax7_6_lsn, label="Runda")

        fig_lessons3.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

    def _print_final_conclusions(self):
        """Prints the final conclusions and key takeaways."""
        print("\n\nüèÅ SEKCJA 8: KO≈ÉCOWE WNIOSKI - TRIUMF STRATEGII PREMIUM V4EScooters")
        print("="*100)

        final_balance_v4e = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi_v4e = (final_balance_v4e - self.initial_investment) / self.initial_investment * 100
        avg_price_premium_phase = self.product_data_v4e[self.product_data_v4e['Round']>=7]['Avg_Selling_Price_V4E'].mean()
        final_market_share_v4e = self.market_data['V4EScooters_share'].iloc[-1]

        print("\nüìä LICZBY, KT√ìRE OPOWIADAJƒÑ HISTORIƒò V4EScooters:")
        print("-" * 100)
        print(f"‚Ä¢ Kapita≈Ç poczƒÖtkowy: ‚Ç¨{self.initial_investment/1e6:.1f}M  |  Ko≈Ñcowe saldo: ‚Ç¨{final_balance_v4e/1e6:.2f}M  |  Ca≈Çkowity ROI: {total_roi_v4e:.2f}%")
        print(f"‚Ä¢ ZarzƒÖdzanie kryzysowe: Prze≈ºyto saldo -‚Ç¨{abs(self.df_finance['Ending_Balance'].min()/1e3):.0f}k (R5), aby osiƒÖgnƒÖƒá rekordowe zyski do R7.")
        print(f"‚Ä¢ Szczyt operacyjny: OsiƒÖgniƒôto i utrzymano 200% efektywno≈õci na obu liniach produkcyjnych (R9-R10).")
        print(f"‚Ä¢ Si≈Ça cen premium: Skutecznie utrzymywano ≈õredniƒÖ cenƒô sprzeda≈ºy ~‚Ç¨{avg_price_premium_phase:.0f} w fazie premium (R7-10).")
        print(f"‚Ä¢ Zysk ponad udzia≈Ç: Zabezpieczono szacunkowe 1. miejsce pod wzglƒôdem rentowno≈õci z {final_market_share_v4e:.2f}% udzia≈Çu w rynku (3. miejsce).")

        print("\nüéØ FORMU≈ÅA SUKCESU STRATEGICZNEGO V4EScooters:")
        print("-" * 100)
        print("1. NIEZACHWIANE ZAANGA≈ªOWANIE W DOSKONA≈ÅO≈öƒÜ OPERACYJNƒÑ jako nienegocjowalny fundament.")
        print("    ‚Üí Umo≈ºliwi≈Ço to przyw√≥dztwo kosztowe w segmencie premium, napƒôdzajƒÖc wysokie mar≈ºe.")
        print("2. ODWA≈ªNE I KONSEKWENTNE POZYCJONOWANIE PREMIUM, priorytetyzujƒÖc rentowno≈õƒá nad wolumenem udzia≈Çu w rynku.")
        print("    ‚Üí Zaowocowa≈Ço to lepszymi wynikami finansowymi, pomimo braku prowadzenia w og√≥lnym wolumenie sprzeda≈ºy.")
        print("3. ZWINNY I OPARTE NA WGLƒÑDZIE MARKETING, uczƒÖcy siƒô z eksperyment√≥w w celu maksymalizacji ROI.")
        print("    ‚Üí Wykorzystano sukces bud≈ºetu ‚Ç¨0 (R8), aby zrozumieƒá prawdziwƒÖ si≈Çƒô marki i zoptymalizowaƒá przysz≈Çe wydatki.")
        print("4. ODPORNE I ADAPTACYJNE WYKONANIE STRATEGII, zw≈Çaszcza w obliczu kryzys√≥w.")
        print("    ‚Üí Skutecznie zarzƒÖdzano katastrofƒÖ dystrybucyjnƒÖ w R5, zmieniajƒÖc produkcjƒô, ceny i fokus sprzeda≈ºy.")
        print("5. KULTURA CIƒÑG≈ÅEGO DOSKONALENIA I UCZENIA SIƒò STRATEGICZNEGO.")
        print("    ‚Üí Konsekwentnie kwestionowano za≈Ço≈ºenia, testowano hipotezy i udoskonalano strategiƒô z rundy na rundƒô.")

        print("\nüí° KLUCZOWE ODKRYCIA I WALIDACJE ZMIENIAJƒÑCE GRƒò V4EScooters:")
        print("-" * 100)
        print("‚Ä¢ Rewelacja efektywno≈õci marketingowej: Ogromne wydatki na reklamƒô nie sƒÖ warunkiem wstƒôpnym wysokiego sukcesu sprzeda≈ºy, je≈õli marka i produkt sƒÖ silne (bud≈ºet ‚Ç¨0 w R8 przyni√≥s≈Ç 92.7% sukcesu sprzeda≈ºy).")
        print("‚Ä¢ Supremacja operacyjna to decydujƒÖca przewaga konkurencyjna: OsiƒÖgniƒôcie 200% efektywno≈õci linii stworzy≈Ço znaczƒÖcƒÖ przewagƒô kosztowƒÖ i jako≈õciowƒÖ.")
        print("‚Ä¢ Rentowno≈õƒá strategii premium: Klienci na tym symulowanym rynku sƒÖ sk≈Çonni p≈Çaciƒá premium za postrzeganƒÖ wysokƒÖ jako≈õƒá, silnƒÖ markƒô i doskona≈Çe cechy produktu.")
        print("‚Ä¢ Lean mo≈ºe byƒá potƒô≈ºny: Strategiczna optymalizacja si≈Çy roboczej w po≈ÇƒÖczeniu z wysokƒÖ efektywno≈õciƒÖ mo≈ºe prowadziƒá do wiƒôkszej produkcji przy mniejszych zasobach.")

        print("\nüèÜ OSTATECZNY WERDYKT W SPRAWIE PODR√ì≈ªY V4EScooters:")
        print("-" * 100)
        print("V4EScooters mistrzowsko zademonstrowa≈Ç, ≈ºe w tym z≈Ço≈ºonym ≈õrodowisku symulacji biznesowej:")
        print("‚Ä¢ Zr√≥wnowa≈ºona rentowno≈õƒá jest cenniejszym celem ni≈º samo przyw√≥dztwo w udziale w rynku.")
        print("‚Ä¢ Efektywno≈õƒá operacyjna i postrzegana jako≈õƒá sƒÖ silniejszymi d≈Çugoterminowymi motorami napƒôdowymi ni≈º tylko skala czy wojny cenowe.")
        print("‚Ä¢ Zwinno≈õƒá strategiczna, uczenie siƒô oparte na danych i zdecydowana zdolno≈õƒá adaptacji przewy≈ºszajƒÖ sztywne, z g√≥ry okre≈õlone plany.")
        print("‚Ä¢ Odporno≈õƒá i zdolno≈õƒá do uczenia siƒô na b≈Çƒôdach sƒÖ kluczowe do zarzƒÖdzania zmienno≈õciƒÖ i osiƒÖgniƒôcia ostatecznego sukcesu.")
        print("\nStrategia V4EScooters w zakresie E-Hulajn√≥g Premium by≈Ça nie tylko udana; by≈Ça ≈õwiadectwem strategicznej wizji, dyscypliny wykonawczej i adaptacyjnego przyw√≥dztwa, kulminujƒÖc w wybitnym triumfie finansowym i operacyjnym.")

        print("\n" + "="*100)
        print(" " * 10 + "üéâ V4EScooters - KONIEC GRY, OSIƒÑGNIƒòTO DOSKONA≈ÅO≈öƒÜ STRATEGICZNƒÑ! üéâ")
        print("="*100)

    def _plot_final_summary_dashboard(self):
        """Generates the final summary dashboard plots."""
        final_balance_v4e = self.df_finance['Ending_Balance'].iloc[-1]
        total_roi_v4e = (final_balance_v4e - self.initial_investment) / self.initial_investment * 100
        final_market_share_v4e = self.market_data['V4EScooters_share'].iloc[-1]
        final_rank_v4e = int(self.market_data['V4EScooters_rank'].iloc[-1])
        avg_office_eff_final = self.operations_data['Office_Efficiency_Avg'].iloc[-1]
        capacity_util_r10 = (self.operations_data['Units_Produced'].iloc[-1] / self.operations_data['Production_Capacity'].iloc[-1] * 100) if self.operations_data['Production_Capacity'].iloc[-1] > 0 else 0
        avg_price_premium_phase = self.product_data_v4e[self.product_data_v4e['Round']>=7]['Avg_Selling_Price_V4E'].mean()
        sales_success_r10_v4e = self.marketing_full[self.marketing_full['Round']==10]['Sales_Success_Rate'].iloc[0]
        v4e_final_profit_per_unit = self.df_finance[self.df_finance['Round']==10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 0
        v4e_final_rev_per_unit = self.df_finance[self.df_finance['Round']==10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 1
        v4e_final_margin_est = (v4e_final_profit_per_unit / v4e_final_rev_per_unit * 100) if v4e_final_rev_per_unit > 0 else 0

        # Figure 8.1 (4 plots)
        fig_summary1, axs_summary1 = plt.subplots(2, 2, figsize=(20, 14))
        fig_summary1.suptitle('V4EScooters: E-HULAJNOGI PREMIUM - PULPIT NAWIGACYJNY ZWYCIƒòSKIEJ STRATEGII - Czƒô≈õƒá 1',
                              fontsize=24, fontweight='bold', color=self.COLORS['primary'], y=0.98)
        fig_summary1.patch.set_facecolor('#f0f4f8')

        # Panel kluczowych statystyk
        axs_summary1[0, 0].axis('off')
        axs_summary1[0, 0].set_title('Kluczowe wska≈∫niki wydajno≈õci', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=20)
        stats_summary_v4e_final = [
            (f'Inwestycja poczƒÖtkowa: ‚Ç¨{self.initial_investment/1e6:.1f}M', 0.85),
            (f'Ko≈Ñcowe saldo: ‚Ç¨{final_balance_v4e/1e6:.2f}M', 0.75),
            (f'Ca≈Çkowity ROI: {total_roi_v4e:.2f}%', 0.65),
            (f'Szac. mar≈ºa zysku (jedn. R10): {v4e_final_margin_est:.1f}%', 0.55),
            (f'Ko≈Ñcowy udzia≈Ç w rynku: {final_market_share_v4e:.2f}% ({final_rank_v4e}. miejsce)', 0.45),
            (f'Efektywno≈õƒá operacyjna: 200% Szczyt', 0.35),
            (f'≈örednia cena premium (R7-10): ‚Ç¨{avg_price_premium_phase:.0f}', 0.25),
            (f'Sukces sprzeda≈ºy (R10): {sales_success_r10_v4e:.1f}%', 0.15)
        ]
        for kpi_text, y_pos_kpi in stats_summary_v4e_final:
            metric_name, metric_value = kpi_text.split(': ')
            axs_summary1[0, 0].text(0.05, y_pos_kpi, metric_name + ":", transform=axs_summary1[0, 0].transAxes,
                                     fontsize=12, fontweight='bold', color=self.COLORS['gray'], va='center')
            axs_summary1[0, 0].text(0.95, y_pos_kpi, metric_value, transform=axs_summary1[0, 0].transAxes,
                                     fontsize=14 if "‚Ç¨" in metric_value or "%" in metric_value else 12,
                                     fontweight='bold', color=self.COLORS['primary'] if "‚Ç¨" in metric_value or "%" in metric_value else self.COLORS['secondary'],
                                     ha='right', va='center')
        axs_summary1[0,0].add_patch(plt.Rectangle((0,0), 1, 1, facecolor='#ffffff', edgecolor=self.COLORS['primary'], lw=2, alpha=0.1, transform=axs_summary1[0,0].transAxes))

        # Podsumowanie wzrostu finansowego
        ax_fin_sum = axs_summary1[0, 1]
        ax_fin_sum.set_title('Trajektoria finansowa', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=15)
        ax_fin_sum.plot(self.df_finance['Round'], self.df_finance['Ending_Balance']/1e6, color=self.COLORS['success'], linewidth=3, marker='o', markersize=8, label='Ko≈Ñcowe saldo (mln ‚Ç¨)')
        ax_fin_sum.axhline(y=self.initial_investment/1e6, color=self.COLORS['gray'], linestyle='--', linewidth=2, label=f'Inwestycja poczƒÖtkowa (‚Ç¨{self.initial_investment/1e6:.1f}M)')
        ax_fin_sum.set_xlabel('Runda')
        ax_fin_sum.set_ylabel('Saldo (mln ‚Ç¨)')
        ax_fin_sum.grid(True, linestyle=':', alpha=0.7)
        ax_fin_sum.legend(fontsize=10, loc='upper left')
        ax_fin_sum.set_facecolor('#f9f9f9')
        for spine in ax_fin_sum.spines.values():
            spine.set_edgecolor(self.COLORS['gray'])
        ax_fin_sum.set_xticks(self.df_finance['Round'])

        # Najwa≈ºniejsze aspekty doskona≈Ço≈õci operacyjnej
        ax_ops_sum = axs_summary1[1, 0]
        ax_ops_sum.set_title('Szczyty operacyjne', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=15)
        op_labels = ['Efektywno≈õƒá linii (R10)', 'Efektywno≈õƒá biur sprzeda≈ºy (R10)', 'Wykorzystanie zdolno≈õci (R10)']
        op_values = [200, avg_office_eff_final, capacity_util_r10]
        op_colors = [self.COLORS['success'], self.COLORS['info'], self.COLORS['warning'] if capacity_util_r10 < 80 else self.COLORS['success']]
        bars_ops_sum = ax_ops_sum.barh(op_labels, op_values, color=op_colors, alpha=0.8, edgecolor='black')
        ax_ops_sum.set_xlabel('Wska≈∫nik wydajno≈õci (%)')
        ax_ops_sum.set_xlim(0, 220)
        for i, bar in enumerate(bars_ops_sum):
            width = bar.get_width()
            ax_ops_sum.text(width + 5, bar.get_y() + bar.get_height()/2, f'{width:.1f}%',
                             ha='left', va='center', fontsize=11, fontweight='bold')
        ax_ops_sum.invert_yaxis()
        ax_ops_sum.grid(True, axis='x', linestyle=':', alpha=0.7)
        ax_ops_sum.set_facecolor('#f9f9f9')
        for spine in ax_ops_sum.spines.values():
            spine.set_edgecolor(self.COLORS['gray'])

        # Marketing i sukces sprzeda≈ºy
        ax_mkt_sum = axs_summary1[1, 1]
        ax_mkt_sum.set_title('Wp≈Çyw na rynek i sprzeda≈º', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=15)
        mkt_kpis = {
            'Ko≈Ñcowy udzia≈Ç w rynku': final_market_share_v4e,
            'Sukces sprzeda≈ºy (R10)': sales_success_r10_v4e,
            'Ranking udzia≈Çu w rynku (ko≈Ñcowy)': float(final_rank_v4e)
        }
        labels_mkt_sum = list(mkt_kpis.keys())
        values_mkt_sum = list(mkt_kpis.values())
        y_pos_mkt = np.arange(len(labels_mkt_sum))

        bars_mkt_kpi = ax_mkt_sum.barh(y_pos_mkt, values_mkt_sum, align='center', color=[self.COLORS['purple'], self.COLORS['success'], self.COLORS['info']], alpha=0.8, edgecolor='black')
        ax_mkt_sum.set_yticks(y_pos_mkt)
        ax_mkt_sum.set_yticklabels(labels_mkt_sum, fontsize=10)
        ax_mkt_sum.invert_yaxis()
        ax_mkt_sum.set_xlabel('Warto≈õƒá / Ranking')
        for i, bar in enumerate(bars_mkt_kpi):
            label_val = f'{values_mkt_sum[i]:.1f}%' if "Ranking" not in labels_mkt_sum[i] else \
                        (f'{int(values_mkt_sum[i])}{". miejsce" if int(values_mkt_sum[i]) in [1,2,3] else ". miejsce"}')
            ax_mkt_sum.text(bar.get_width() + (2 if "Udzia≈Ç" in labels_mkt_sum[i] or "Sukces" in labels_mkt_sum[i] else 0.1), bar.get_y() + bar.get_height()/2,
                            label_val, ha='left', va='center', fontsize=11, fontweight='bold')
        ax_mkt_sum.grid(True, axis='x', linestyle=':', alpha=0.7)
        ax_mkt_sum.set_facecolor('#f9f9f9')
        for spine in ax_mkt_sum.spines.values():
            spine.set_edgecolor(self.COLORS['gray'])

        fig_summary1.text(0.5, 0.02, f"V4EScooters | Business Game Final Report | Generated: {datetime.now().strftime('%B %d, %Y')}",
                          ha='center', va='bottom', fontsize=10, color=self.COLORS['gray'], style='italic')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.subplots_adjust(hspace=0.4, wspace=0.3)
        plt.show()

        # Figure 8.2 (2 plots)
        fig_summary2, axs_summary2 = plt.subplots(1, 2, figsize=(20, 8))
        fig_summary2.suptitle('V4EScooters: E-HULAJNOGI PREMIUM - PULPIT NAWIGACYJNY ZWYCIƒòSKIEJ STRATEGII - Czƒô≈õƒá 2',
                              fontsize=24, fontweight='bold', color=self.COLORS['primary'], y=0.98)
        fig_summary2.patch.set_facecolor('#f0f4f8')

        # Filary strategiczne
        ax_strat_sum = axs_summary2[0]
        ax_strat_sum.set_title('Kluczowe filary strategiczne', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=20)
        ax_strat_sum.axis('off')
        pillars = [
            ("1. FOKUS PREMIUM", "Rentowno≈õƒá ponad wolumen", self.COLORS['purple']),
            ("2. DOSKONA≈ÅO≈öƒÜ OPERACYJNA", "Cel 200% efektywno≈õci", self.COLORS['primary']),
            ("3. ZWINNY MARKETING", "Eksperymenty oparte na danych", self.COLORS['success']),
            ("4. ODPORNO≈öƒÜ", "Nawigacja kryzysowa i odbudowa", self.COLORS['warning'])
        ]
        y_pillar = 0.85
        for i, (title, subtitle, color) in enumerate(pillars):
            ax_strat_sum.text(0.05, y_pillar - i*0.22, title, transform=ax_strat_sum.transAxes, fontsize=13, fontweight='bold', color=color)
            ax_strat_sum.text(0.1, y_pillar - i*0.22 - 0.08, subtitle, transform=ax_strat_sum.transAxes, fontsize=10, style='italic', color=self.COLORS['gray'])
        ax_strat_sum.add_patch(plt.Rectangle((0,0), 1, 1, facecolor='#ffffff', edgecolor=self.COLORS['primary'], lw=2, alpha=0.1, transform=ax_strat_sum.transAxes))

        # Ko≈Ñcowy przeglƒÖd konkurent√≥w
        ax_comp_sum = axs_summary2[1]
        ax_comp_sum.set_title('Ko≈Ñcowa pozycja na rynku', fontsize=16, fontweight='bold', color=self.COLORS['primary'], pad=15)

        final_shares_comp = [self.market_data[col].iloc[-1] for col in self.competitor_columns_plot]
        v4e_final_profit_per_unit = self.df_finance[self.df_finance['Round']==10]['Profit_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 0
        v4e_final_rev_per_unit = self.df_finance[self.df_finance['Round']==10]['Revenue_Per_Unit_Sold'].iloc[0] if not self.df_finance[self.df_finance['Round']==10].empty else 1
        v4e_final_margin_est = (v4e_final_profit_per_unit / v4e_final_rev_per_unit * 100) if v4e_final_rev_per_unit > 0 else 0

        estimated_margins_comp = [
            v4e_final_margin_est,
            15.0,
            25.0,
            -10.0
        ]
        comp_labels_sum = [self.plot_teams_map['team_5'][0], self.plot_teams_map['KKKompany'][0], self.plot_teams_map['V4EScooters_share'][0], self.plot_teams_map['team_8'][0]]
        comp_colors_sum = [self.plot_teams_map['team_5'][1], self.plot_teams_map['KKKompany'][1], self.plot_teams_map['V4EScooters_share'][1], self.plot_teams_map['team_8'][1]]

        scatter_comp_sum = ax_comp_sum.scatter(final_shares_comp, estimated_margins_comp, s=[s*15 for s in [6,5,4,2]],
                                               c=comp_colors_sum, alpha=0.8, edgecolors='black')
        for i, txt in enumerate(comp_labels_sum):
            ax_comp_sum.annotate(txt.split('\n')[0], (final_shares_comp[i], estimated_margins_comp[i]),
                                 textcoords="offset points", xytext=(0,10 if i==0 else -15), ha='center', fontsize=9)

        ax_comp_sum.set_xlabel('Udzia≈Ç w rynku (%)')
        ax_comp_sum.set_ylabel('Szac. mar≈ºa zysku (%)')
        ax_comp_sum.grid(True, linestyle=':', alpha=0.7)
        ax_comp_sum.axhline(0, color='black', lw=0.5, ls='--')
        ax_comp_sum.set_facecolor('#f9f9f9')
        for spine in ax_comp_sum.spines.values():
            spine.set_edgecolor(self.COLORS['gray'])

        fig_summary2.text(0.5, 0.02, f"V4EScooters | Business Game Final Report | Generated: {datetime.now().strftime('%B %d, %Y')}",
                          ha='center', va='bottom', fontsize=10, color=self.COLORS['gray'], style='italic')

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.subplots_adjust(hspace=0.4, wspace=0.3)
        plt.show()

    def generate_report(self):
        """Generates the complete business game report."""
        self._load_and_process_data()
        self._print_header()
        self._plot_financial_performance()
        self._plot_marketing_strategy()
        self._plot_unit_economics()
        self._plot_operational_excellence()
        self._plot_competitive_dynamics()
        self._plot_product_strategy()
        self._plot_geographical_strategy()
        self._plot_strategic_lessons()
        self._print_final_conclusions()
        self._plot_final_summary_dashboard()

# To run the report:
if __name__ == "__main__":
    reporter = BusinessGameReporter()
    reporter.generate_report()
```
